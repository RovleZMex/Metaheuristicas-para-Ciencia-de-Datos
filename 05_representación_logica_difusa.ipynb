## 1. El papel de la lógica como lenguaje

El razonamiento humano es, por naturaleza, variable e inconsistente. Cada persona desarrolla una concepción del mundo y un marco de razonamiento que depende de las experiencias que va acumulando, el contexto en el que ocurren y la secuencia de esos eventos. Con esa herramienta imprecisa y altamente subjetiva, el ser humano pretende comprender y actuar sobre un mundo que es también incierto y cambiante. Para lograrlo —y para poder colaborar con otras personas— necesitábamos un mecanismo que permitiera determinar la validez de un juicio independientemente de quién lo emitiera.

---

### La lógica clásica

Uno de los logros más importantes del pensamiento humano fue el desarrollo de la lógica clásica, que formalizó los mecanismos del razonamiento y los liberó de la subjetividad del hablante. Este esquema se apoya en tres principios fundamentales:

1. **Principio de Identidad:** Una cosa es lo que es ($A = A$).
2. **Principio de No Contradicción:** Algo no puede ser y no ser al mismo tiempo ($\neg(A \land \neg A)$).
3. **Principio del Tercero Excluido:** Una proposición es verdadera o es falsa; no existe una tercera opción ($A \lor \neg A$).

<p align="center">
  <img src="images/logica_clasica.png" width="800">
</p>


Estos tres principios son pilares del razonamiento matemático formal y, al mismo tiempo, bases epistemológicas del método científico.

El poder de la lógica clásica proviene de una simplificación radical: obliga a las *proposiciones sobre* la realidad a expresarse en términos bivaluados. La realidad es un flujo continuo de experiencias; la lógica binaria ofrece una regla de decisión que poda esa complejidad para permitirnos actuar. Decir "este hongo es venenoso (1) o no lo es (0)" es una simplificación vital que ignora grados de toxicidad en favor de la supervivencia.

Pero esa misma simplificación introduce una tensión profunda: la realidad es continua, ambigua y contextual, mientras que la lógica clásica es discreta, rígida y absoluta. Podemos entenderla como un instrumento perfectamente diseñado para la consistencia formal, pero que le queda demasiado ajustado a la naturaleza. Esa tensión es el punto de partida de la lógica difusa.

---

### Los límites de la lógica clásica

Los límites de la lógica clásica han sido expuestas a través de las llamadas "*paradojas*". Estas narraciones breves no son curiosidades filosóficas marginales (aunque según se dice Séneca las consideraba una pérdida de tiempo): Las paradojas son señales estructurales que indican dónde el sistema de representación es insuficiente.

<p align="center">
  <img src="images/Escher_Hands.png" width="400">
</p>

Una paradoja aparece cuando la realidad no cabe dentro del sistema lógico que intentamos imponerle. Formalmente, muchas paradojas pueden expresarse como ecuaciones sin solución en lógica binaria:

$$x = \neg x$$

$$\text{No existe solución en } \{0, 1\}$$

Esto no significa que el problema esté mal planteado. Significa que el sistema lógico es insuficiente. 


#### Paradojas clásicas

Las siguientes paradojas comparten una característica común: introducen autorreferencia, continuidad o vaguedad, y la lógica clásica responde con contradicción. La contradicción no está en la realidad; está en la representación.

**La Paradoja del Mentiroso**

Esta es la más antigua y directa de todas, atribuida a Eubúlides de Mileto (siglo IV a.C.). 

> Un cretense dice: *"Todos los cretenses mienten"*. Si la afirmación es verdadera, entonces él también miente, por lo que es falsa. Si es falsa, entonces él dice la verdad, por lo que es verdadera. La proposición colapsa en un bucle sin fondo.


**La Paradoja del Barbero (Russell, 1901)**

> En un pueblo hay un barbero que sigue una regla estricta: *afeita exactamente a quienes no se afeitan a sí mismos*.

La pregunta inevitable es: ¿entonces, quién afeita al barbero?

- Si se afeita a sí mismo, viola su regla.
- Si no se afeita, su regla lo obliga a hacerlo.

La proposición no puede ser 0 ni 1.


**La Paradoja de la Horca**

> Un marajá condena a un prisionero, pero le ofrece una última oportunidad: *"Di una sentencia. Si es verdadera, morirás en la horca. Si es falsa, morirás decapitado."* El prisionero, después de reflexionar, responde: *"Moriré decapitado."*

¿Cuál es la decisión correcta del marajá? 

- Si lo manda decapitar, la sentencia se comprueba como verdadera, y debió colgarlo (no decapitarlo).
- Si lo cuelga, la sentencia resulta falsa, pero entonces no debió ahorcarlo (debió decapitarlo).

El sistema binario es incapaz de procesar esta situación. No porque sea irracional, sino porque la lógica clásica no tiene categorías para la dependencia circular entre una afirmación y su consecuencia.

**La Paradoja de Sorites (el montón de arena)**

> Un millón de granos de arena forman un montón. Retirar un grano no destruye el montón. Aplicando ese razonamiento repetidamente debemos concluir que un solo grano es un montón.

El problema aquí es que la lógica clásica no tiene forma de trazar el límite porque el límite no existe de manera precisa. La paradoja de Sorites no es un bucle lógico: es una frontera borrosa, y eso es algo cualitativamente diferente.

> **Nota histórica:** También atribuida a Eubúlides, su motivación era epistemológica: demostrar que el lenguaje ordinario no puede capturar la realidad con precisión. Durante siglos fue tratada como un problema menor o un juego de palabras. Su verdadero peso solo se hizo evidente en el siglo XX, cuando Lotfi Zadeh reconoció en ella no un defecto del lenguaje, sino una característica del mundo. La paradoja de Sorites es, en cierto sentido, el punto de partida implícito de toda la lógica difusa.


#### ¿Incertidumbre o vaguedad? El gato de Schrödinger revisitado

Las paradojas anteriores sugieren que hay al menos dos tipos distintos de problemas que la lógica clásica no puede manejar: la **autorreferencia** y la **vaguedad**. Otro problema que la lógica clásica no resuelve satisfactoriamente es la **incertidumbre**, un concepto que es importante distinguir del de *vaguedad*, antes de entrar en detalles de la Lógica difusa.

Una metáfora clásica para entender esta diferencia es la de las dos botellas de agua en el desierto. Imaginemos que estamos en el desierto y tenemos sed y nos ofrecen dos botellas de agua:

<ul>
    <li>La primera botella <em>tiene un 5% de probabilidad de que esté totalmente envenenada.</em> La presencia del veneno es binario: está o no está. En este caso tenemos incertidumbre acerca de si la botella pertenece al conjunto de botellas de agua potable o al conjunto de botellas de agua envenenada.</li><br>

<li>La segunda botella <em>está ligeramente envenenada</em> ($\mu=0.05$). En este caso, sabemos que la botella pertenece <em>un poco</em> al conjunto de botellas de agua envenenada. El veneno está diluido en toda el agua; sabemos exactamente qué hay adentro, pero la definición de <em>agua potable</em> se ha vuelto borrosa.

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/Nuka_Cola2.jpg" width="400">
</p></li>
</ul>

> **La probabilidad mide la frecuencia de que algo ocurra. La lógica difusa mide el grado en que algo pertenece a un concepto.**

Podemos llevar este ejercicio al famoso experimento mental de Erwin Schrödinger. Él lo propuso para criticar la interpretación ontológica de la escuela de Copenhague, pero nosotros lo usaremos para mapear tres formas de ver la realidad. El problema se enuncia de la siguiente manera:

>Un gato es encerrado en una caja con un dispositivo que incluye una fuente radiactiva, un detector y veneno. Si el átomo se desintegra (asumiendo una probabilidad de este evento del 50%), el veneno se libera y el gato muere; si no, vive.

- El enfoque Cuántico (**Superposición**): Si el átomo realmente se encuentra en una superposición de estados, porque la naturaleza "traslapa" diferentes estados del átomo y muestra al observador uno de esos estados (a capricho) en el momento de interactuar con el átomo, entonces **el gato también debería compartir esa indefinición y, por lo tanto, debía estar vivo y muerto al mismo tiempo**. Para Schrödinger, esto era un absurdo que demostraba que algo fallaba en la interpretación cuántica de la época.
  
- El enfoque Probabilístico (**Incertidumbre Epistemológica**): Aquí, mientras la caja está cerrada, decimos que **el gato tiene un 50% de probabilidad de estar vivo**. Sin embargo, asumimos que dentro de la caja el gato ya es un cadáver o un ser vivo. La probabilidad mide nuestra falta de información. Al abrir la caja, la incertidumbre "colapsa" porque nuestra ignorancia desaparece.
  
- El enfoque de Lógica Difusa (**Vaguedad Semántica**): Aquí el problema no es el estado físico del gato, sino la rigidez del predicado "**estar vivo**". En lugar de preguntar si el gato está vivo o muerto, preguntamos *en qué grado podemos afirmar que lo está*. En ausencia de observación, o ante estados intermedios (agonía, coma), el concepto mismo admite fronteras difusas. La vaguedad no describe una falla en el conocimiento (como la probabilidad), sino una falta de precisión en nuestras categorías lingüísticas para capturar la complejidad de la realidad.

> La incertidumbre es una falta de conocimiento sobre un hecho nítido; la vaguedad es una falta de nitidez en la definición misma del hecho.

### El camino hacia lo no-binario

Antes de la llegada de la lógica difusa en 1965, la humanidad ya sospechaba que el blanco o negro era insuficiente. Diferentes pensadores intentaron añadir matices al pensamiento formal para rescatar la complejidad de la realidad. El recorrido es más largo de lo que suele reconocerse, y empieza, con cierta ironía, en el mismo lugar donde comenzó la lógica clásica.


#### Aristóteles y el futuro contingente

Aristóteles, quien sistematizó los principios de la lógica clásica, fue el primero en notar que estos principios —diseñados para un mundo de certezas— fallaban al mirar hacia el mañana. Este es el problema de los futuros contingentes. Consideremos el enunciado

> *"Mañana habrá una batalla naval."*

- **El riesgo del determinismo**: Si hoy decimos que es verdadera, entonces la batalla tiene que ocurrir; el destino ya está escrito y el esfuerzo humano es irrelevante.
- **El riesgo de la imposibilidad**: Si decimos que es falsa, estamos decretando hoy que la batalla es imposible, negando que el azar o la voluntad puedan cambiar las cosas.

Él propuso una excepción elegante: aunque es necesariamente cierto que mañana habrá o no habrá una batalla (la suma de las opciones es 100%), ninguna de las dos partes es verdadera todavía. Al hacer esto, Aristóteles rompió el binomio "verdadero/falso" para los eventos inciertos, sembrando la semilla de lo que siglos después llamaríamos lógicas polivalentes o de sistemas donde la verdad no es estática, sino que "llega a ser". En resumen: el futuro no es un dato, es una potencia.


#### Lógicas multivaluadas y modales

En el siglo XX, el lógico polaco **Jan Łukasiewicz** retomó la intuición aristotélica y la formalizó. Su contribución fue doble:

1. **Lógica trivaluada ($Ł_3$):** Introdujo un tercer valor de verdad, el $\frac{1}{2}$, para representar proposiciones **indeterminadas** o "posibles". Esto representó el primer desafío formal al monopolio del conjunto $\{0, 1\}$ (Verdadero/Falso) dentro de la lógica matemática.

2. **Lógicas de $n$ valores:** Łukasiewicz también desarrolló sistemas con cualquier número finito de valores e incluso exploró versiones **infinito-valuadas**. En estas, la verdad comienza a entenderse no como una categoría rígida, sino como un **grado** o una distancia.

En paralelo, las **lógicas modales** enriquecieron el vocabulario formal mediante operadores que distinguen la realidad de la potencialidad:
*   $\Box P$ : Es **necesario** que $P$.
*   $\Diamond P$ : Es **posible** que $P$.

Esta distinción permite capturar matices que la lógica clásica ignora: no es lo mismo afirmar "mañana habrá una batalla" que admitir que "es **posible** ($\Diamond$) que la haya". 

Sin embargo, a pesar de su sofisticación, estas lógicas seguían siendo esencialmente **estructuras de peldaños**. Incluso en los modelos de infinitos valores, la verdad se trataba como una sucesión de puntos aislados en una escala. Faltaba el paso definitivo: dejar de ver la verdad como una escalera de grises y empezar a verla como un **espectro continuo** y fluido, donde las fronteras entre los estados se disuelven.


#### Lógica paraconsistente

Las lógicas multivaluadas atacaron el problema de la indeterminación. Pero las paradojas de autorreferencia — el Mentiroso, el Barbero, la Horca — plantean un problema distinto: no ausencia de valor de verdad, sino *contradicción activa*. 

En la lógica clásica, una contradicción es una catástrofe absoluta. Debido al **Principio de Explosión**, si se acepta una sola premisa contradictoria, se puede derivar lógicamente cualquier otra afirmación. El sistema se vuelve "trivial": si "A" y "no A" son ciertos, entonces cualquier cosa puede ser. El sistema colapsa y deja de ser útil.

La **lógica paraconsistente**, desarrollada formalmente por **Newton da Costa** en los años 60 y expandida por **Graham Priest** con su sistema **LP** (*Logic of Paradox*), propone una solución radical: **abandonar el principio de explosión**. 

Un sistema paraconsistente permite:
*   **Contradicciones locales:** El sistema puede contener "focos" de inconsistencia sin que estos infecten al resto del razonamiento.
*   **Dialetismo:** La aceptación de que algunas proposiciones (como la paradoja del Mentiroso) son simplemente **verdaderas y falsas a la vez**. 

Lejos de ser un ejercicio abstracto, esta lógica es la herramienta natural para dominios donde la coherencia perfecta es imposible:
*   **Bases de datos masivas:** Donde fuentes de información contradictorias deben coexistir.
*   **Sistemas jurídicos:** Donde dos normas en conflicto no pueden anular la validez de todo el código legal.
*   **Inteligencia Artificial:** Para agentes que deben razonar a partir de datos ruidosos o creencias humanas inconsistentes.

Si la lógica de Łukasiewicz nos enseñó a manejar el "quizás", la paraconsistencia de Da Costa nos enseñó a sobrevivir al "sí y no al mismo tiempo".


#### Lógica difusa: el gradiente continuo

En 1965, **Lotfi A. Zadeh** dio el salto definitivo. En lugar de añadir uno o dos valores extra, propuso que el grado de pertenencia de un elemento a un conjunto puede ser **cualquier número real en el intervalo $[0, 1]$**:

$$\mu_A(x) \in [0, 1]$$

La pertenencia deja de ser una puerta — abierta o cerrada — y se convierte en una rampa. Un elemento puede pertenecer *parcialmente* a un conjunto con grado $\mu = 0.3$, sin que eso implique nada automático sobre su relación con el conjunto complementario. Esto permite que un sistema formal opere con conceptos lingüísticos matizados como *"muy"*, *"poco"* o *"más o menos"* — los llamados **modificadores lingüísticos** — que la lógica clásica simplemente no puede representar.

1. La paradoja de Sorites, que durante siglos pareció un juego de palabras sin solución, encuentra aquí su respuesta natural: no existe un límite preciso entre "montón" y "no montón" porque *montón* es un concepto inherentemente difuso. La frontera no está ausente por falta de información — está ausente porque el concepto mismo no la tiene.
2. Bajo esta óptica, el barbero puede pertenecer al conjunto de hombres afeitados por el barbero y también pertenecer al conjunto de hombres que afeitan a sí mismos, aunque un conjunto sea el complemento del otro.

> En lógica difusa, la verdad ya no está limitada a los valores estrictos de "cierto/falso", sino que presenta verdades parciales.


#### Conjuntos ásperos (*Rough Sets*)

Vale mencionar un desarrollo paralelo que complementa, sin reemplazar, a la lógica difusa. En 1982, **Zdzisław Pawlak** propuso los *Rough Sets* desde un ángulo distinto. Para entender la diferencia, conviene recuperar la distinción que establecimos con el gato de Schrödinger:

- La **lógica difusa** habita el lado *ontológico*: la vaguedad está en el concepto mismo. "Alto" no tiene una frontera precisa porque la altura es un continuo y el lenguaje no la corta en ningún punto privilegiado.
- Los **conjuntos ásperos** habitan el lado *epistemológico*: el concepto puede estar perfectamente bien definido, pero nuestra granularidad de datos no es suficiente para determinar si un objeto pertenece o no.

Cuando la información disponible no permite distinguir entre dos objetos, Pawlak propuso representar el conjunto mediante dos aproximaciones:

- **Aproximación inferior:** los elementos que *con certeza* pertenecen al conjunto.
- **Aproximación superior:** los elementos que *posiblemente* pertenecen.

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/rough_set.png" width="800">
</p>

La región entre ambas es la *zona de frontera* — la huella de nuestra ignorancia sobre los datos. En ciencia de datos, los conjuntos ásperos son especialmente útiles para reducción de atributos y manejo de datos inconsistentes en tablas de decisión.


#### Una mirada de conjunto

Cada sistema que hemos visto ataca un tipo distinto de límite de la lógica clásica:

| Paradoja / Problema | Naturaleza | Respuesta |
| :--- | :--- | :--- |
| Futuro contingente | Indeterminación temporal | Łukasiewicz — introduce $\frac{1}{2}$ |
| Mentiroso / Horca / Barbero | Autorreferencia contradictoria | Lógica paraconsistente — tolera contradicciones locales |
| Sorites | Vaguedad ontológica del concepto | Lógica difusa — disuelve la frontera en un gradiente |
| Indiscernibilidad de datos | Imprecisión epistemológica | Conjuntos ásperos — aproximación inferior y superior |

La lógica clásica no está equivocada — opera con precisión dentro de sus supuestos. Lo que estas extensiones reconocen es que esos supuestos no cubren todo el territorio del razonamiento real. La lógica difusa, en particular, no corrige a Aristóteles: lo extiende hacia el dominio donde sus herramientas no alcanzan.

## 2. Conjuntos difusos: formalización

### Conceptos fundamentales

Un **conjunto difuso** $A$ sobre un universo de discurso $X$ se define mediante una **función de membresía**:

$$\mu_A : X \rightarrow [0, 1]$$

que asigna a cada elemento $x \in X$ un grado de pertenencia $\mu_A(x)$. 

Observemos que la visualización de un conjunto difuso es diferente a los conjuntos clásicos (Cantor). Si queremos forzar un diagrama de Venn para un conjunto y su complemento se varía, posiblemente, de la siguiente manera:

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/fuzzy_Venn.png" width="600">
</p>

Como vemos, esta representación ya no es de utilidad. En su lugar, representamos un conjunto difuso en términos de su fucnión de membresía.

En la representación de un conjunto difuso distinguimos los siguientes conceptos:

- **Soporte** $\text{supp}(A) = \{ x \in X \mid \mu_A(x) > 0 \}$: Es el alcance total del concepto. Incluye a todo elemento que tenga *alguna* posibilidad de pertenecer al conjunto $(\mu_A(x) > 0)$.

- **Núcleo:** $\text{core}(A) = \{ x \in X \mid \mu_A(x) = 1 \}$: Es la zona de certeza absoluta. Contiene los elementos que son prototipos perfectos del concepto ($ \mu_A(x) = 1 $).
- **Ancho de banda (Bandwidth):** Es la distancia entre los **puntos de cruce** (donde $\mu = 0.5$). Mide la extensión de la parte "sustancial" del conjunto.
- **Altura** $h(A) = \sup_{x \in X} \mu_A(x)$: Es el máximo valor de pertenencia de algún elemento en el conjunto.
- **Corte-$\alpha$** $A_\alpha = \{ x \in X \mid \mu_A(x) \geq \alpha \}$: Es un conjunto nítido (*crisp*) que actúa como umbral. Por ejemplo, un corte-$0.8$ extrae solo los elementos con alta membresía.

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/Fuzzy_set.png" width="800">
</p>

*   **Normalidad:** Un conjunto es **normal** si su altura es 1. Esto garantiza que el concepto no es vacío ni puramente hipotético; existe al menos un elemento que lo define plenamente.
*   **Convexidad:** Indica que el conjunto no tiene "huecos" ni es multimodal. Si un elemento está entre dos elementos con alta membresía, él también debe tenerla. Esto asegura que los **Cortes-$\alpha$** sean siempre intervalos continuos, facilitando el cálculo matemático.


### Funciones de membresía

La elección de la función de membresía determina cómo se modela la transición entre pertenencia y no pertenencia. Las funciones más utilizadas son:

<ul>
<li><strong>Triangular</strong> — transición lineal simétrica o asimétrica:
<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/triangular.png" width="300"><br>

### Elección de la función de membresía

La función de membresía no es un detalle técnico menor: es la hipótesis central del modelo difuso. Elegirla bien significa entender qué tipo de transición describe el fenómeno que queremos representar, con qué precisión conocemos esa transición, y qué costo computacional podemos asumir.


#### Triangular

**Forma:** rampa de subida lineal, pico en $b$, rampa de bajada lineal.

**Ventajas:**
- Interpretable: los tres parámetros $(a, b, c)$ tienen significado directo — inicio, centro y fin de la zona de transición.
- Computacionalmente trivial: solo operaciones lineales.
- Fácil de ajustar manualmente por un experto de dominio.

**Desventajas:**
- El pico es una esquina — la derivada es discontinua en $b$, lo que puede generar problemas en sistemas que requieren gradientes suaves.
- No modela bien fenómenos con transiciones asimétricas complejas.

**Usos típicos:** sistemas de control difuso clásicos, prototipado rápido, representación de variables lingüísticas simples (*"temperatura media"*, *"velocidad moderada"*).


#### Trapezoidal

**Forma:** rampa de subida, meseta central entre $b$ y $c$, rampa de bajada.

**Ventajas:**
- Permite modelar un **núcleo extendido**: una zona donde el elemento pertenece completamente al conjunto, no solo un punto.
- Más robusta que la triangular ante pequeñas variaciones del valor de entrada dentro de la meseta.

**Desventajas:**
- Cuatro parámetros hacen el ajuste manual algo más laborioso.
- Las esquinas en $b$ y $c$ mantienen el problema de discontinuidad en la derivada.

**Usos típicos:** representación de rangos aceptables en control de calidad, variables con zona de núcleo bien definida (*"temperatura confortable entre 20°C y 24°C"*).


#### Gaussiana

**Forma:** campana simétrica suave centrada en $c$, dispersión controlada por $\sigma$.

**Ventajas:**
- Infinitamente diferenciable — ideal para sistemas que requieren gradientes.
- Conexión natural con distribuciones de probabilidad y modelos estadísticos.
- Transición suave y continua en todo su dominio.

**Desventajas:**
- Soporte técnicamente infinito: nunca llega exactamente a cero, lo que puede generar activaciones residuales no deseadas.
- Siempre simétrica — no puede modelar transiciones asimétricas sin modificaciones.
- Más costosa computacionalmente que las funciones lineales por tramos.

**Usos típicos:** sistemas neuro-difusos (ANFIS), modelado de variables continuas con comportamiento unimodal, cualquier aplicación donde la suavidad del gradiente es prioritaria.



#### Función S

**Forma:** transición suave monotónica creciente de 0 a 1.

**Ventajas:**
- Modela conceptos abiertos por la derecha: *"muy alto"*, *"muy rápido"*, *"suficientemente caliente"*.
- Transición suave sin las esquinas de la trapezoidal.
- Solo dos parámetros: punto de inicio y punto de saturación.

**Desventajas:**
- No tiene núcleo acotado — una vez que alcanza 1, permanece ahí indefinidamente.
- No sirve para modelar conceptos con zona central de pertenencia máxima y decaimiento posterior.

**Usos típicos:** umbrales de activación, representación de extremos en escalas lingüísticas, preprocesamiento de señales.


#### Función Z

**Forma:** complemento exacto de la función S — transición suave monotónica decreciente.

**Ventajas y desventajas:** simétricas a las de la función S, con el comportamiento invertido.

**Usos típicos:** mismos contextos que la función S, pero para el extremo opuesto de la escala lingüística. En la práctica, S y Z suelen aparecer juntas como los flancos de una partición difusa.



#### Función Pi

**Forma:** campana construida como concatenación de una función S y una función Z.

**Ventajas:**
- Núcleo extendido con transiciones suaves — combina lo mejor de la trapezoidal y la gaussiana.
- Control independiente sobre los flancos de subida y bajada mediante cuatro parámetros.
- Puede ser asimétrica, a diferencia de la gaussiana.

**Desventajas:**
- Cuatro parámetros requieren más cuidado en el ajuste.
- Menos común en la literatura (los árbitros suelen cuestionarla).

**Usos típicos:** modelado de variables lingüísticas centrales con transiciones suaves (*"temperatura media"*, *"presión normal"*), aplicaciones donde la simetría de la gaussiana es demasiado restrictiva.


#### Sigmoide

**Forma:** curva en S suave con pendiente controlada por $a$ y punto de inflexión en $c$.

**Ventajas:**
- Diferenciable en todo su dominio — compatible con optimización por gradiente.
- Ampliamente disponible en frameworks de aprendizaje automático.
- El parámetro $a$ controla directamente la *sharpness* de la transición: valores grandes aproximan una función escalón, valores pequeños producen transiciones muy suaves.

**Desventajas:**
- El soporte es técnicamente infinito en ambas direcciones.

**Usos típicos:** sistemas neuro-difusos donde los parámetros se aprenden por retropropagación, activaciones en redes difusas, modelado de umbrales graduales.


#### Campana sigmoide (aproximación gaussiana)

**Forma:** diferencia de dos sigmoides — campana asimétrica controlable.

**Ventajas:**
- Hereda la diferenciabilidad de la sigmoide — apta para gradiente.
- Permite **asimetría** en los flancos ajustando independientemente $c_1$ y $c_2$.
- Puede aproximar una gaussiana cuando $c_1$ y $c_2$ son simétricamente equidistantes del centro.
- Útil cuando se quiere una forma gaussiana pero con transiciones más abruptas en uno de los flancos.

**Desventajas:**
- Menos interpretable que la gaussiana — los parámetros $c_1$ y $c_2$ no tienen el mismo significado intuitivo que $c$ y $\sigma$.
- La normalización a $[0,1]$ requiere cuidado cuando $a$ es pequeño.

**Usos típicos:** sistemas neuro-difusos con variables asimétricas, modelado de respuestas sensoriales o perceptuales donde el umbral de entrada y el de saturación no son equidistantes del centro.



#### Guía de selección rápida

| Criterio | Función recomendada |
| :--- | :--- |
| Máxima interpretabilidad | Triangular, Trapezoidal |
| Núcleo puntual, gradiente suave | Gaussiana |
| Núcleo extendido, gradiente suave | Pi |
| Concepto abierto (extremo de escala) | S / Z |
| Compatible con aprendizaje por gradiente | Gaussiana, Sigmoide, Bell sigmoide |
| Asimetría en los flancos | Pi, Bell sigmoide |
| Prototipado rápido | Triangular |
| Recursos computacionales limitados | Triangular, Trapezoidal, S / Z |

## 3. Operaciones sobre conjuntos difusos

Las operaciones sobre conjuntos difusos generalizan las operaciones clásicas de la teoría de conjuntos. El principio rector es el mismo que guió a Zadeh en 1965: cuando los grados de membresía se restringen a $\{0,1\}$, todas las operaciones difusas deben colapsar exactamente a sus equivalentes clásicos.



### A. Operaciones estándar de Zadeh

Sean $A$ y $B$ dos conjuntos difusos sobre $X$. Las operaciones estándar se definen puntualmente:

**Complemento:**
$$\mu_{\neg A}(x) = 1 - \mu_A(x)$$

**Unión:**
$$\mu_{A \cup B}(x) = \max(\mu_A(x),\ \mu_B(x))$$

**Intersección:**
$$\mu_{A \cap B}(x) = \min(\mu_A(x),\ \mu_B(x))$$

Estas tres operaciones satisfacen la mayoría de las propiedades de la lógica clásica — conmutatividad, asociatividad, distributividad, leyes de De Morgan — con una excepción notable:

$$\mu_{A \cup \neg A}(x) \neq 1 \qquad \mu_{A \cap \neg A}(x) \neq 0$$

El **principio del tercero excluido** y el **principio de no contradicción** no se sostienen en general para conjuntos difusos. Esto no es un defecto — es precisamente lo que permite modelar la vaguedad.

---

### B. T-normas y T-conormas: generalizando la intersección y la unión

Las operaciones de Zadeh son solo una elección posible. Una **T-norma** $T: [0,1]^2 \rightarrow [0,1]$ es cualquier función que generaliza la intersección satisfaciendo:

1. **Conmutatividad:** $T(a, b) = T(b, a)$
3. **Asociatividad:** $T(a, T(b, c)) = T(T(a, b), c)$
4. **Monotonía:** $b \leq c \Rightarrow T(a, b) \leq T(a, c)$
5. **Condición de frontera:** $T(a, 1) = a$

Las **T-conormas** (o S-normas) generalizan la unión de forma análoga, con condición de frontera $S(a, 0) = a$.

> El término **T** en **T-normas** y **T-conormas** significa "*Triangle*" haciendo refeencia a la desigualdad del triángulo; Karl Menger diseñó estas funciones en 1942 para generalizar el cálculo de distancias en espacios métricos probabilísticos.

Las T-normas más utilizadas son:

| Nombre | $T(a, b)$ | Interpretación |
| :--- | :--- | :--- |
| Mínimo (Zadeh) | $\min(a, b)$ | Conservadora — solo lo que ambos comparten |
| Producto algebraico | $a \cdot b$ | Penaliza fuertemente membresías bajas  $a \cdot b \le \min(a, b)$  |
| Łukasiewicz | $\max(0,\ a + b - 1)$ | La más restrictiva — exige alta membresía en ambos conjuntos|
| Drástica | $
T_D(a, b) = 
\begin{cases} 
\min(a, b) & \text{si } \max(a, b)=1 \\ 
0 & \text{en otro caso} 
\end{cases}
$| Caso límite — solo activa si uno de los dos vale 1 |

Y las T-conormas correspondientes:

| Nombre | $S(a, b)$ | Interpretación |
| :--- | :--- | :--- |
| Máximo (Zadeh) | $\max(a, b)$ | El más permisivo |
| Suma probabilística | $a + b - a \cdot b$ | Análoga a la probabilidad de unión |
| Łukasiewicz | $\min(1,\ a + b)$ | Saturación en 1 |
| Drástica | $S_D(a,b) = 
\begin{cases} 
\max(a, b) & \text{si } \min(a, b)=0 \\ 
1 & \text{en otro caso} 
\end{cases}
$| Solo activa si uno de los dos vale 0 |

La elección de T-norma y T-conorma define el **comportamiento lógico** del sistema difuso. En aplicaciones de control, el producto algebraico suaviza las transiciones; en clasificación, el mínimo de Zadeh es más interpretable.

---

### C. Modificadores lingüísticos

Los **modificadores lingüísticos** (o *hedónicos*) son operadores que transforman un conjunto difuso para representar matices lingüísticos, sin cambiar el conjunto base. Actúan como adverbios (*muy*, *poco*, *más o menos*), ajustando la precisión y la intensidad del conocimiento experto en sistemas difusos. 

Se aplican directamente sobre la función de membresía:

| Modificador | Operación | Efecto |
| :--- | :--- | :--- |
| *very* (muy) | $\mu^2(x)$ | Concentra — reduce membresías intermedias |
| *slightly* (ligeramente) | $\mu^{0.5}(x)$ | Dilata — amplía la zona de transición |
| *more or less* (más o menos) | $\mu^{1/3}(x)$ | Dilata más agresivamente |
| *extremely* (extremadamente) | $\mu^3(x)$ | Concentra más agresivamente |
| *not* (no) | $1 - \mu(x)$ | Complemento estándar |

En general, un modificador hedónico de potencia $p$ actúa como:

$$\mu_{\text{hedge}}(x) = [\mu_A(x)]^p$$

- $p > 1$ **concentra** el conjunto — exige mayor grado de membresía para pertenecer.
- $p < 1$ **dilata** el conjunto — acepta membresías más bajas.

Esto permite construir automáticamente la representación de *"muy caliente"* a partir de *"caliente"*, sin necesidad de definir una nueva función de membresía desde cero.

---

### D. Principio de extensión

El **principio de extensión de Zadeh** permite generalizar cualquier función clásica $f: X \rightarrow Y$ para que opere sobre conjuntos difusos en lugar de elementos puntuales.

Dada $f: X \rightarrow Y$ y un conjunto difuso $A$ sobre $X$, el conjunto difuso imagen $f(A)$ sobre $Y$ se define como:

$$ \mu_{f(A)}(y) = \sup_{x : f(x) = y} \mu_A(x) $$

Es decir, para saber qué tan "cierto" es un resultado $y$, buscamos todos los valores de $x$ que podrían haberlo generado a través de la función. Si varios valores de $x$ apuntan al mismo resultado $y$, nos quedamos con el **mayor grado de pertenencia** (el supremo) entre ellos. 

**¿Por qué importa?** Porque permite propagar incertidumbre difusa a través de modelos matemáticos. Si la entrada de un sistema es un conjunto difuso — por ejemplo, *"temperatura aproximadamente 30°C"* — el principio de extensión define cómo calcular la salida difusa correspondiente. Es la base teórica de la **aritmética difusa** y de los sistemas de inferencia difusa que veremos en la siguiente sección.

```python

import matplotlib.gridspec as gridspec

# ── Paleta Clara (Vista Estándar) ───────────────────────────────────────────
BG       = "#FFFFFF"  # Fondo blanco
TEXT     = "#1A1A1A"  # Texto oscuro
CYAN     = "#007BFF"  # Azul vibrante
AMBER    = "#D97706"  # Naranja/Ámbar oscuro
LAVENDER = "#6D28D9"  # Violeta
GREEN    = "#059669"  # Esmeralda
PINK     = "#DB2777"  # Rosa fuerte
GRAY     = "#6B7280"  # Gris para ejes
GRID     = "#E5E7EB"  # Gris muy claro para guías

x = np.linspace(0, 10, 500)

# ── Funciones de Membresía ────────────────────────────────────────────────────
def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

def funcion_S(x, a, b):
    m = (a + b) / 2
    y = np.where(x <= a, 0,
        np.where(x <= m, 2 * ((x - a) / (b - a)) ** 2,
        np.where(x <= b, 1 - 2 * ((x - b) / (b - a)) ** 2, 1)))
    return y.astype(float)

def funcion_Z(x, a, b):
    return 1 - funcion_S(x, a, b)

A = triangular(x, 1, 3, 5)
B = triangular(x, 4, 6, 8)

# ── Operaciones y T-normas ───────────────────────────────────────────────────
complement_A   = 1 - A
union_AB       = np.maximum(A, B)
intersect_AB   = np.minimum(A, B)
tnorm_prod     = A * B
tnorm_luka     = np.maximum(0, A + B - 1)
tconorm_prob   = A + B - A * B
tconorm_luka   = np.minimum(1, A + B)

# ── Hedges y Extensión ───────────────────────────────────────────────────────
very_A         = A ** 2
slightly_A     = A ** 0.5
y_range        = np.linspace(0, 100, 1000)

def extension_principle(x, mu_A, f, y_range):
    mu_fA = np.zeros_like(y_range)
    f_vals = f(x)
    for i, y in enumerate(y_range):
        mask = np.abs(f_vals - y) < 0.5
        mu_fA[i] = np.max(mu_A[mask]) if np.any(mask) else 0.0
    return mu_fA

mu_ext = extension_principle(x, A, lambda x: x**2, y_range)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(16, 18), facecolor=BG)
gs  = gridspec.GridSpec(4, 3, figure=fig, hspace=0.5, wspace=0.3)

def styled_ax(ax, title):
    ax.set_facecolor(BG)
    ax.set_title(title, color=TEXT, fontsize=10, fontweight="bold", pad=10)
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(-0.05, 1.15)
    ax.set_xlabel("$x$", color=GRAY, fontsize=9)
    ax.set_ylabel("$\\mu(x)$", color=GRAY, fontsize=9)
    ax.tick_params(colors=GRAY, labelsize=8)
    ax.grid(True, color=GRID, linestyle='-', linewidth=0.5)
    for spine in ax.spines.values():
        spine.set_edgecolor(GRID)
    return ax

def plot_base(ax):
    ax.plot(x, A, color=GRAY,  linewidth=1, linestyle="--", alpha=0.4, label="$A$")
    ax.plot(x, B, color=GRAY,  linewidth=1, linestyle=":", alpha=0.4, label="$B$")

# Paneles de ejemplo (Operaciones)
panels = [
    (gs[0,0], "Complemento $\\neg A$", complement_A, LAVENDER),
    (gs[0,1], "Unión (Máx)", union_AB, GREEN),
    (gs[0,2], "Intersección (Mín)", intersect_AB, PINK),
    (gs[1,0], "T-norma Producto", tnorm_prod, AMBER),
    (gs[1,1], "T-norma Łukasiewicz", tnorm_luka, LAVENDER),
    (gs[2,0], "T-conorma Suma Prob.", tconorm_prob, AMBER),
    (gs[2,1], "T-conorma Łukasiewicz", tconorm_luka, LAVENDER),
]

for spec, title, yd, col in panels:
    ax = styled_ax(fig.add_subplot(spec), title)
    plot_base(ax)
    ax.plot(x, yd, color=col, linewidth=2, label="Resultado")
    ax.fill_between(x, yd, color=col, alpha=0.1)
    ax.legend(fontsize=8, loc='center right', frameon=True, facecolor='white')

# Hedges
ax_h = styled_ax(fig.add_subplot(gs[3, 0:2]), "Modificadores (Hedges)")
ax_h.plot(x, A, color=TEXT, linewidth=1, linestyle="--", label="Base")
ax_h.plot(x, very_A, color=CYAN, linewidth=2, label="Very A")
ax_h.plot(x, slightly_A, color=AMBER, linewidth=2, label="Slightly A")
ax_h.legend(fontsize=8)

# Extensión
ax_ext = fig.add_subplot(gs[3, 2])
ax_ext.set_title("Principio de Extensión\n$f(x)=x^2$", fontweight="bold", fontsize=10)
ax_ext.plot(y_range, mu_ext, color=PINK, linewidth=2)
ax_ext.fill_between(y_range, mu_ext, color=PINK, alpha=0.1)
ax_ext.set_xlim(0, 100); ax_ext.set_ylim(-0.05, 1.15)
ax_ext.grid(True, color=GRID)

fig.suptitle("Manual de Operaciones Difusas (Vista Estándar)", color=TEXT, fontsize=14, y=0.92)
plt.show()
```

## 4. Base de reglas difusas

### El lenguaje del conocimiento experto

Un sistema difuso no opera sobre ecuaciones numéricas: opera sobre **reglas lingüísticas**. Una base de reglas difusas es la formalización del *conocimiento experto* en forma de proposiciones condicionales del tipo:

$$\text{SI } x_1 \text{ es } A_1 \text{ Y } x_2 \text{ es } A_2 \quad \text{ENTONCES} \quad y \text{ es } B$$

donde $A_1$, $A_2$ y $B$ son conjuntos difusos definidos sobre los universos de discurso de las variables de entrada $x_1, x_2$ y la variable de salida $y$, respectivamente.

Cada regla tiene tres componentes:

- **Antecedente:** una o varias proposiciones difusas sobre las entradas, combinadas con conectivos lógicos (*Y*, *O*). Su evaluación produce un escalar $\alpha \in [0, 1]$ llamado **grado de activación** o **nivel de disparo** (*firing strength*).

- **Consecuente:** una proposición difusa sobre la salida. En el modelo de Mamdani es un conjunto difuso $B$; en Takagi–Sugeno es una función numérica de las entradas.

- **Mecanismo de implicación:** el operador que transfiere el grado de activación $\alpha$ del antecedente al consecuente, produciendo un **consecuente recortado** (o ponderado).

A diferencia de las reglas en la lógica de predicados clásica —que se activan completamente o no se activan, de acuerdo a un esquema binario—, una regla difusa se **activa en grado**, dependiendo de qué tanto se cumple su antecedentes.

### Evaluación del antecedente

Para una regla con múltiples antecedentes conectados por *Y*, el grado de activación se computa como la **T-norma** aplicada a los grados de membresía individuales:

$$\alpha = T\bigl(\mu_{A_1}(x_1),\ \mu_{A_2}(x_2),\ \ldots \bigr)$$

Con la T-norma mínimo de Zadeh:

$$\alpha = \min\bigl(\mu_{A_1}(x_1),\ \mu_{A_2}(x_2)\bigr)$$

Esta elección es conservadora: el grado de activación queda limitado por el antecedente más débil, exactamente como en la lógica clásica el *AND* booleano devuelve 0 si cualquier operando es 0.

Con el **producto algebraico** $T(a,b) = a \cdot b$, la activación distribuye el efecto de ambos antecedentes en lugar de ser dominada por el mínimo. La elección de T-norma es, por tanto, una decisión de diseño con consecuencias sobre el comportamiento del sistema.

Si los antecedentes están conectados por *O*, se utiliza la **T-conorma** correspondiente (máximo de Zadeh, suma probabilística, etc.).


### Implicación: del antecedente al consecuente

Una vez calculado $\alpha$, el consecuente difuso $B$ es modificado para reflejar que la regla se cumple solo parcialmente. Las dos estrategias más comunes son:

| Método | Operación sobre $\mu_B(y)$ | Explicación |
| :--- | :--- | :--- |
| **Recorte** (*clipping*) | $\mu_{B'}(y) = \min(\alpha,\ \mu_B(y))$ | Se "trunca" el conjunto $B$ a la altura de $\alpha$. Es el más común porque mantiene la forma original del conjunto pero limita su certidumbre. |
| **Escalado** (*scaling*) | $\mu_{B'}(y) = \alpha \cdot \mu_B(y)$ | Se "encoge" el conjunto $B$ multiplicando todos sus puntos por $\alpha$. Este método preserva mejor la proporcionalidad de la forma. |

En ambos casos, $B'$ es el **consecuente activado**, cuya forma refleja tanto el conocimiento codificado en $B$ como el grado en que la regla se disparó.


### Base de reglas como partición del espacio de entrada

En la práctica, el sistema difuso no contiene una sola regla sino una **base de reglas** $\mathcal{R} = \{R_1, R_2, \ldots, R_K\}$ que cubre el espacio de entrada. Para que el sistema sea bien comportado, esta cobertura debe satisfacer dos propiedades:

- **Completitud:** para cualquier combinación posible de entradas, al menos una regla debe activarse con $\alpha > 0$. Sin esto, hay regiones del espacio de entrada donde el sistema no produce respuesta.


- **Consistencia:** Las reglas deben guardar armonía entre sí. Si dos reglas tienen antecedentes muy similares (se activan ante las mismas entradas), sus consecuentes deben ser coherentes. 
    - **El problema del conflicto:** Una inconsistencia ocurre cuando una regla dice "Girar a la izquierda" y otra similar dice "Girar a la derecha". 
    - **Efecto en el sistema:** Al promediar o agregar estas respuestas opuestas, el sistema podría quedar "paralizado" en el centro o presentar oscilaciones bruscas y erráticas en la salida, perdiendo la suavidad característica del control difuso.


La completitud de un sistema difuso se garantiza asegurando que las funciones de membresía cubran completamente el universo de discurso. Esto implica que ningún punto quede sin representación semántica dentro del sistema.

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/completitud.png" width="400">
</p>

Sea un conjunto de funciones de membresía:

$$
\{\mu_1(x), \mu_2(x), \dots, \mu_n(x)\}, \quad x \in X
$$


$$
\forall x \in X,\; \exists i \in \{1,\dots,n\} \text{ tal que } \mu_i(x) > 0
$$

Esta condición garantiza que **todo punto del universo de discurso pertenece al menos parcialmente a alguna etiqueta lingüística**, evitando "huecos semánticos".

La definición más común y robusta, sin embargo, es la **partición de unidad de Ruspini**:

$$
\forall x \in X,\; \sum_{i=1}^{n} \mu_i(x) = 1
$$

Esta condición indica que en cualquier punto $x$, la "certeza" total está distribuida entre los conjuntos. La partición difusa introduce una estructura más regular que favorece la estabilidad en procesos de inferencia y optimización.

En muchos sistemas también se exige que cada función de membresía alcance el valor máximo en algún punto:

$$
\forall i,\; \exists x \in X \text{ tal que } \mu_i(x) = 1
$$

Esto asegura que cada etiqueta lingüística esté plenamente representada en el universo de discurso.

## 5. Sistemas de Inferencia Difusa (FIS)

Un **sistema de inferencia difusa** (FIS, *Fuzzy Inference System*) es el "motor" del modelo. Es un marco computacional que traduce el conocimiento experto expresado en lenguaje natural —como *"si la epidemia crece rápido, la tasa de contagio es alta"*— en una función matemática de mapeo no lineal. Su objetivo es transformar entradas numéricas (datos de sensores o estadísticas) en salidas numéricas (decisiones o estimaciones).

Si las funciones de membresía están bien diseñadas (formando una partición de la unidad), el FIS garantiza una transición suave y continua entre reglas, evitando saltos bruscos en la salida cuando las entradas cambian ligeramente.

### Arquitectura funcional

El proceso se divide en tres etapas críticas que permiten pasar del entorno numérico estricto, proveniente de sensores, a un contexto conceptual, y regresar al contexto numérico para generar una respuesta accionable:

$$\text{Entrada numérica} \xrightarrow{\text{Fuzzificación}} \text{Grados de membresía} \xrightarrow{\text{Inferencia}} \text{Conjuntos difusos de salida} \xrightarrow{\text{Defuzzificación}} \text{Salida numérica}$$

<ol>
<li><strong>Fuzzificación:</strong> Traduce las entradas numéricas estrictas en grados de membresía $\mu \in [0, 1]$. Aquí es donde los datos adquieren significado semántico (por ejemplo, una temperatura de $25^\circ C$ puede tener una <strong>pertenencia</strong> de $0.7$ en el conjunto <strong>Templado</strong>).</li>

<li><strong>Inferencia:</strong> Esta etapa es un proceso de tres pasos:
    <ol>
        <li><strong>Evaluación del antecedente:</strong> Aquí se aplican las <strong>T-normas</strong>/<strong>T-conormas</strong> para 
            combinar los grados de pertenencia de las entradas y obtener el <strong>grado de activación</strong> ($\alpha_k$) de cada 
            regla $k$. En 
            esta etapa, el sistema decide qué tan relevante es cada fragmento del conocimiento experto para la situación actual. </li>
        <li><strong>Implicación:</strong> El grado $\alpha_k$ se transfiere al consecuente de la regla. Mediante <strong>recorte</strong>
            o <strong>escalado</strong>, transformamos el conjunto difuso teórico en un <strong>consecuente activado</strong> $\mu_{C_k'}
            (y)$.</li>
        <li><strong>Agregación:</strong> Combina las conclusiones de todas las reglas activas. Dado que un FIS opera bajo una lógica 
            disyuntiva, la agregación  unifica los resultados parciales en una única distribución difusa global (usualmente vía <strong>T-
                conormas</strong> como el <emph>máximo</emph>). Esta etapa unifica las "opiniones" de todas las reglas disparadas.</li>
    </ol>
<li><strong>Defuzzificación:</strong> Realiza la síntesis final. Convierte la distribución difusa agregada en un valor numérico único y accionable, utilizando criterios como el <strong>Centroide</strong> o la <strong>Media de Máximos</strong>.
</ol>


### Taxonomía de los sistemas FIS

Aunque el pipeline general es compartido, los sistemas clásicos difieren fundamentalmente en la naturaleza de sus **consecuentes** y en su estrategia de **agregación**:

* **Mamdani (1975):** El más cercano al razonamiento humano. Sus consecuencias son conjuntos difusos. Es altamente **interpretable** pero computacionalmente más costoso debido a la necesidad de integrar áreas en la defuzzificación.
* **Takagi-Sugeno (1985):** Diseñado para control automatizado y optimización. Sus consecuencias son funciones matemáticas (lineales o constantes). Es **eficiente** y facilita el análisis matemático de estabilidad.
* **Tsukamoto (1979):** Un punto medio donde las consecuencias son conjuntos difusos con funciones de membresía monótonas, lo que simplifica la inferencia a una suma ponderada.
---

### A. Sistema de Mamdani

El modelo de Mamdani (1975) es el más cercano al razonamiento humano: tanto las entradas como las salidas se describen mediante **etiquetas lingüísticas**. Su estructura sigue la forma:

$$\text{SI } x_1 \text{ es } A_i \text{ Y } x_2 \text{ es } B_i \text{ ENTONCES } y \text{ es } C_i$$

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/Mandami.png" width="800" alt="Esquema Mamdani">
</p>

**Etapa 1 — Fuzzificación**
Se transforman las entradas numéricas (escalares) en grados de pertenencia $\mu \in [0, 1]$.  
*Ejemplo base: Control de un motor.*
*   **Entrada 1 (Error):** *Negativo, Cero, Positivo.*
*   **Entrada 2 (Variación del error; derivada):** *Lento, Rápido.*

**Etapa 2 — Base de Reglas**
El conocimiento se organiza en una matriz lógica. Cada celda representa una decisión experta:


| # | Si el Error es... | Y la Variación es... | Entonces la Potencia es... |
|:--|:--|:--|:--|
| **R1** | Positivo | Rápido | Muy Alta |
| **R2** | Positivo | Lento | Alta |
| **R3** | Cero | Lento | Media |
| **R4** | Negativo | Lento | Baja |

**Etapa 3 — Motor de Inferencia**
1. **Activación:** Se calcula el nivel de disparo $\alpha_i$ usando una T-norma (usualmente el mínimo):
   $$\alpha_i = \min(\mu_{A_i}(x_1), \mu_{B_i}(x_2))$$
2. **Implicación:** Se obtiene el **consecuente recortado** $\mu_{C_i'}(y)$ para cada regla:
   $$\mu_{C_i'}(y) = \min(\alpha_i, \mu_{C_i}(y))$$
3. **Agregación:** Se combinan todas las respuestas en un único conjunto difuso global mediante la unión (máximo):
   $$\mu_{\text{out}}(y) = \max_i \mu_{C_i'}(y)$$

**Etapa 4 — Defuzzificación**
Para obtener una acción de control numérica ($y^*$), se procesa el conjunto agregado:

*   **Centroide (COG):** Calcula el centro de masa de la figura resultante. Es el método más robusto y produce transiciones suaves.
    $$y^* = \frac{\int y \cdot \mu_{\text{out}}(y)\, dy}{\int \mu_{\text{out}}(y)\, dy}$$
*   **Media de Máximos (MOM):** Promedia los valores de $y$ que tienen la membresía más alta. Es útil cuando solo importa el valor más plausible, aunque puede generar cambios bruscos.

Un portafolio de inversión (o cartera de inversión) es el conjunto total de activos financieros que posee una persona o una institución. En lugar de poner todo el dinero en un solo lugar, el portafolio agrupa diferentes instrumentos para intentar alcanzar objetivos financieros específicos. 

Sus componentes principales suelen ser:

- **Bonos**: Instrumentos de deuda (créditos, pagarés, etc.).
- **Efectivo**: Dinero líquido o fondos de mercado monetario (divisas).

2.  **Perfil de Riesgo:** El diseño del portafolio depende de la tolerancia a la pérdida del inversor y de su horizonte temporal (cuánto tiempo planea mantener la inversión).
3.  **Asignación de Activos (*Asset Allocation*):** Es la decisión estratégica de qué porcentaje del capital total se destina a cada categoría de inversión. 
    *   En un portafolio tradicional, esto podría ser **60%** acciones y **40%** bonos. 
    *   En productos estructurados como los **CDO** (*Collateralized Debt Obligations*), la diversificación se realiza con base en los
      **tramos** de riesgo; por ejemplo, asignar un **80%** al segmento de menor riesgo (Supersenior), un **15%** al segmento de riesgo
        medio (Mezzanine) y un **5%** al tramo de máximo riesgo (Equity/Junior).

#### El Modelo Clásico: Selección de Portafolios de Markowitz

La piedra angular de la gestión de carteras moderna es la **Teoría de Portafolios de Harry Markowitz** (1952). Su premisa fundamental es que un inversor no debe evaluar un activo de forma aislada, sino por cómo su riesgo y retorno contribuyen al conjunto de la cartera.

El modelo asume que los inversores son racionales y tienen aversión al riesgo, buscando maximizar el **retorno esperado** para un nivel determinado de **riesgo** (medido como volatilidad). Las componentes del modelo son:

| Variable | Definición Matemática | Significado |
| :--- | :--- | :--- |
| **$E[R_p]$** | $\sum w_i E[R_i]$ | Retorno esperado del portafolio (media ponderada). |
| **$\sigma_p^2$** | $\sum \sum w_i w_j \sigma_{ij}$ | Riesgo del portafolio (considera la covarianza entre activos). |
| **$w_i$** | $w_i \in [0,1], \sum w_i = 1$ | Pesos o proporción de capital asignado a cada activo. |

> Aquí, $E[R_p]$, leído como "el retorno esperado de la cartera p", no es una "promesa" de ganancia, sino el primer momento de la distribución de probabilidad de los rendimientos, es decir, el término "retorno esperado" hace referencia al valor esperado en estadística.

El objetivo matemático es encontrar la **Frontera Eficiente**: el conjunto de portafolios que ofrecen el máximo retorno para cada nivel de riesgo. 

<p align="center" style="padding-top: 10px; padding-bottom: 10px;">
  <img src="images/frontera_eficiente.png" width="500">
</p>

Para evaluar la calidad de un punto en esta frontera, se utiliza habitualmente la **Tasa de Sharpe ($S$):**

$$
S = \frac{E[R_p] - R_f}{\sigma_p}
$$

donde $R_f$ es la tasa libre de riesgo. El **Ratio de Sharpe** no tiene límites teóricos: un $S$ más alto indica una mayor eficiencia en el uso del riesgo. En la práctica su rango útil suele acotarse a **[0, 4]** para cubrir la mayoría de los casos reales:

- **< 0**: El activo tiene un rendimiento menor que la tasa libre de riesgo (no deseable y, por lo tanto, no atractivo para nuestro universo de selección).
- **0 a 1**: Activos con bajo o nulo premio por riesgo (ej.: bonos corporativos de bajo riesgo).
- **1 a 2**: Activos aceptables, típicos en fondos de inversión tradicionales o estrategias pasivas.
- **2 a 3**: Activos muy buenos, asociados a estrategias activas exitosas o mercados con oportunidades claras.
- **> 3**: Activos excepcionales, poco comunes y generalmente limitados a estrategias de alto rendimiento (ej.: *hedge funds* especializados) o mercados emergentes con alto potencial.
  **Nota**: Valores en este rango pueden requerir validación adicional para descartar errores en el cálculo del riesgo.

#### Validez y Limitaciones

Aunque el modelo de Markowitz es el estándar académico, su implementación en escenarios reales en analítica presenta retos significativos que motivan el uso de alternativas como la **Lógica difusa**.

##### Puntos fuertes

- **Diversificación cuantificable:** Proporciona un marco matemático riguroso para reducir el riesgo sistémico mediante la correlación de activos.
- **Optimización convexa:** El problema de selección se puede resolver eficientemente mediante programación cuadrática, lo cual es computacionalmente atractivo.

##### Limitaciones

1. **Inestabilidad de los parámetros:** El modelo es extremadamente sensible a pequeñas variaciones en las estimaciones de $E[R]$ y $\sigma$, lo que se conoce como "maximización de errores".
2. **Supuesto de normalidad:** Markowitz asume que los rendimientos siguen una distribución normal. En la realidad, los mercados presentan "colas pesadas" (eventos extremos frecuentes), lo que subestima el riesgo real.
3. **Falta de sentido común (interpretación lingüística):** Un optimizador numérico puede recomendar una asignación del 99% a un solo activo por una diferencia de 0.001 en la taza de Sharpe. Los sistemas **Mamdani** permiten inyectar "reglas de cordura" (p. ej., "si la volatilidad es extrema, el atractivo es bajo sin importar el retorno").
4. **Métricas heterogéneas:** A menudo, los analistas quieren incluir datos no financieros (sentimiento de noticias, ESG -restricciones ambientales, sociales y de gobierno-, momentum). Combinar estas dimensiones con métricas de varianza en un solo optimizador lineal es complejo.

---

#### Modelado del Sistema de Inferencia Difusa (FIS)

En la gestión de carteras profesional, antes de aplicar el álgebra matricial de Markowitz para hallar los pesos óptimos ($w_i$), existe una etapa crítica: la **selección de activos** (*Asset Selection*). Un optimizador es tan bueno como los datos que recibe (*Garbage In, Garbage Out*). Si alimentamos un modelo de varianza media con activos excesivamente volátiles o con retornos distorsionados por ruido estadístico, la "frontera eficiente" resultante será inestable y poco práctica.

Para abordar las limitaciones del modelo clásico, diseñaremos un **Sistema de Inferencia Difusa (FIS) tipo Mamdani** que actúe como un **filtro experto**. En lugar de intentar resolver la asignación del portafolio completo de una sola vez, utilizaremos el sistema difuso para evaluar el **atractivo individual** de cada instrumento. 

Este enfoque nos permite:
1. **Cuantificar el "sentido común":** Traducir reglas heurísticas de analistas financieros (que suelen ser no lineales) en un sistema matemático determinista.
2. **Reducir sensibilidad al ruido:** A diferencia de un KPI numérico que cambia drásticamente ante variaciones mínimas, los conjuntos difusos "absorben" la volatilidad de los datos de entrada mediante los grados de pertenencia.
3. **Pre-seleccionar activos atractivos:** Solo los activos cuyo atractivo supere un puntaje umbral determinado por el sistema FIS serán considerados como candidatos para la fase final de optimización de pesos.

Para alimentar nuestro sistema, utilizaremos la API de **Yahoo Finance** (`yfinance`) para obtener precios históricos de diferentes activos que eventualmente pudiéramos incluir en nuestro portafolio. 

> **Nota:** Los resultados se estarán discutiendo en base a los datos de dos años, con corte al **17 de abril de 2026**. 

```python
# DATOS DE PRUEBA
# =====================

def generar_datos(ticker="NVDA", n_points=1000): 
    """
    Genera y grafica df_crisp_data.
    """
    plt.figure(figsize=(12, 5))
    
    raw = yf.download(ticker, period="2y", interval="1d", auto_adjust=True)

    # 3. Limpieza de MultiIndex (Para que funcione con cualquier versión de yfinance)
    if isinstance(raw.columns, pd.MultiIndex):
        raw.columns = raw.columns.get_level_values(0)

    # Verificación de datos
    if raw.empty:
        print(f"Error: No se pudieron descargar datos para {ticker_symbol}. Revisa tu conexión.")
        return np.nan
    else:
        # 4. Cálculo de Retornos Logarítmicos            
        raw['Returns'] = np.log(raw['Close'] / raw['Close'].shift(1))
        df_crisp_data = raw.dropna().copy()
    
        df_crisp_data['Close'].plot(color='royalblue')
        plt.title(f"Serie de Tiempo: Precios Históricos {ticker} (Yahoo Finance)")
        plt.ylabel("Precio")
    
        plt.grid(True, alpha=0.3)
        plt.show()
        return df_crisp_data

# --- SELECCIÓN DEL CONJUNTO DE DATOS ---
ticker = "HG=F"   # "NVDA", "AAPL", "BTC-USD", "GC=F" (Oro), KO (Coca-Cola), "SPAX.PVT", "INTC"
                      # "AMD", "PLTR", "MSFT", "NFLX", "AMZN"...
df_crisp_data = generar_datos(ticker=ticker)
```

<br>

A partir de las series de tiempo, calculamos los valores numéricos (*crisp*) que servirán como entradas al modelo:

1.  **Retorno esperado ($E[R]$):** Media anualizada de los rendimientos logarítmicos diarios.
2.  **Volatilidad ($\sigma$):** Desviación estándar anualizada de los rendimientos (medida de riesgo de Markowitz).
3.  **Ratio de Sharpe ($S$):** Calculado como $(E[R] - R_f) / \sigma$.

Estos datos crudos representan la "fotografía" actual del mercado, la cual suele ser ruidosa. El Sistema de Inferencia Difusa se encargará de interpretar estos valores bajo una lógica de grados de pertenencia en lugar de umbrales rígidos.

```python
# ==============================================================================
# CÁLCULO DE MÉTRICAS FINANCIERAS (VALORES CRISP)
# ==============================================================================

# Configuración de mercado
DIAS_COMERCIALES = 252
TASA_LIBRE_RIESGO = 0.04  # 4% anual (ej. Bonos del Tesoro)

# Cálculo de variables de entrada para el sistema difuso
# ------------------------------------------------------------------------------
# 1. Volatilidad (Anualizada)
vol_anual = df_crisp_data['Returns'].std() * np.sqrt(DIAS_COMERCIALES) * 100

# 2. Retorno Esperado (Anualizado)
ret_anual = df_crisp_data['Returns'].mean() * DIAS_COMERCIALES * 100

# 3. Sharpe Ratio
sharpe = (ret_anual/100 - TASA_LIBRE_RIESGO) / (vol_anual/100) if vol_anual != 0 else 0

# Almacenamos los valores crisp para el motor de inferencia
input_values = {
    "volatilidad": vol_anual,
    "retorno": ret_anual,
    "sharpe": sharpe
}

# --- Reporte de Métricas ---
print(f"{'='*40}")
print(f"VARIABLE {ticker}")
print(f"{'='*40}")

print(f"{'Métrica':<25} | {'Valor Crisp':>10}")
print(f"{'-'*40}")
print(f"{'Retorno Esperado Anual':<25} | {ret_anual:>9.2f}%")
print(f"{'Volatilidad Anualizada':<25} | {vol_anual:>9.2f}%")
print(f"{'Sharpe Ratio (Rf=4%)':<25} | {sharpe:>10.2f}")
print(f"{'='*40}\n")

# --- Visualización de la Distribución ---
fig, ax = plt.subplots(1, 2, figsize=(14, 4))

# Gráfico 1: Serie Temporal de Retornos
df_crisp_data['Returns'].plot(ax=ax[0], color='navy', alpha=0.7)
ax[0].set_title(f"Log-Retornos Diarios: {ticker}")
ax[0].grid(True, linestyle='--', alpha=0.5)

# Gráfico 2: Histograma de Retornos (Visualización del Riesgo)
df_crisp_data['Returns'].hist(ax=ax[1], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
ax[1].set_title(f"Distribución de Retornos (Ruido a Fuzzificar)")
ax[1].axvline(0, color='red', linestyle='--', label='Retorno 0')
ax[1].legend()

plt.tight_layout()
plt.show()
```


Estas gráficas capturan dos dimensiones del comportamiento del activo: su **dinámica temporal** y su **estructura estadística de riesgo**.

**1) Serie de Log-Retornos Diarios (Dinámica Temporal)**. Esta gráfica muestra la evolución de los retornos logarítmicos en el tiempo, permitiendo observar propiedades clave del proceso:

- **Ruido e imprevisibilidad local**: Los retornos fluctúan alrededor de cero sin un patrón determinista evidente, reflejando la naturaleza estocástica del mercado.
  
- **Clustering de volatilidad**: Los movimientos amplios, alejados de cero, son periodos de alta volatilidad.

- **Eventos extremos (*outliers*)**: Se muestran como picos aislados (positivos o negativos) que se alejan significativamente del comportamiento promedio. Estos eventos dominan el riesgo financiero real.

- **Estacionalidad**: Aunque es raro en activos financieros, la estacionalidad, si existiera, aparecería como patrones repetitivos o cambios estructurales persistentes.

Son componentes muy utilizada en pronóstico de volatilidad de activos, acciones, divisas y tasas de interés.

Siguiendo el esquema estándar de la arquitectura Mamdani, definiremos tres componentes fundamentales antes de ejecutar el motor de inferencia.

**1. Definición del Universo de Discurso y Fuzzificación**

El primer paso consiste en transformar nuestros valores estrictos (*crisp*) en grados de pertenencia. Para nuestro evaluador de activos, definiremos tres variables de entrada ($x_1, x_2, x_3$) y una de salida ($y$).

| Variable | Universo de Discurso ($U$) | Etiquetas Lingüísticas |
| :--- | :--- | :--- |
| **$x_1$: Volatilidad ($\sigma$)** | $[0\%, 100\%]$ anual | Bajo, Medio, Alto |
| **$x_2$: Retorno Esperado ($E[R]$)** | $[-20\%, 50\%]$ anual | Negativo, Neutro, Positivo |
| **$x_3$: Ratio de Sharpe ($S$)** | $[0, 4]$ | Pobre, Bueno, Excelente |
| **$y$: Atractivo de Inversión** | $[0, 100]$ puntos | Muy Bajo, Bajo, Medio, Alto, Muy Alto |

Nota: En el caso del ratio de Sharpe, un valor cercano a 4 simplemente caería en el hombro derececho del conjunto "*Excelente*".

<br>

En la arquitectura formal de un **Sistema Basado en Conocimiento (SBC)**, la inteligencia del sistema no reside únicamente en sus reglas, sino en la formalización de sus definiciones. Nuestra **base de conocimiento** se divide en dos componentes fundamentales:

1.  **Base de datos lingüística (diccionario de conceptos):** Es el conjunto de definiciones de los términos que utilizaremos, descritos mediante nuestras **funciones de membresía**.
2.  **Base de reglas (conocimiento procedimental):** El conjunto de sentencias que conectan los conceptos del diccionario para ejecutar el razonamiento.

A continuación, presentamos el panel completo de nuestra **base de datos lingüística**, integrando las tres variables de entrada (Volatilidad, Retorno y Sharpe) y la variable de salida (Atractivo).
 

```python
# ================================
# 1. UTILIDADES Y DEFINICIÓN FIS
# ================================

def membership_trapezoid(x, a, b, c, d):
    return np.maximum(
        0,
        np.minimum(
            np.minimum((x - a) / (b - a) if b > a else 1, 1),
            (d - x) / (d - c) if d > c else 1
        )
    )

def compute_membership(universe, params):
    return [membership_trapezoid(x, *params) for x in universe]


# ================================
# 2. UNIVERSOS
# ================================

universes = {
    'vol': np.linspace(0, 100, 500),
    'ret': np.linspace(-20, 50, 500),
    'sha': np.linspace(0, 4, 500),
    'atr': np.linspace(0, 100, 500)
}


# ================================
# 3. CONJUNTOS DIFUSOS
# ================================

fuzzy_sets = {
    'vol': {
        'Bajo':  [0, 0, 12, 25],
        'Medio': [15, 30, 45, 60],
        'Alto':  [50, 65, 100, 100]
    },
    'ret': {
        'Negativo': [-20, -20, -8, 2],
        'Neutro':   [-5, 0, 8, 15],
        'Positivo': [10, 20, 50, 50]
    },
    'sha': {
        'Pobre':     [0, 0, 0.6, 1.1],
        'Bueno':     [0.9, 1.4, 2.0, 2.8],
        'Excelente': [2.4, 3.0, 4.0, 4.0]
    },
    'atr': {
        'Muy Bajo': [0, 0, 10, 20],
        'Bajo':     [10, 25, 35, 50],
        'Medio':    [40, 50, 60, 70],
        'Alto':     [60, 70, 80, 90],
        'Muy Alto': [80, 90, 100, 100]
    }
}

# Configuración de las métricas reales obtenidas de Yahoo
labels= {
    'vol': 'Volatilidad $\sigma$ (%)',
    'ret': 'Retorno Esperado $E[R]$ (%)',
    'sha': 'Sharpe Ratio'
}

# ================================
# 4. VISUALIZACIÓN DE FUZZIFICACIÓN
# ================================

def plot_fuzzification(metrics, ticker):

    fig, axs = plt.subplots(2, 2, figsize=(14, 10))
    plt.subplots_adjust(hspace=0.3, wspace=0.2)

    layout = {
        'vol': axs[0, 0],
        'ret': axs[0, 1],
        'sha': axs[1, 0]
    }

    for key, ax in layout.items():
        for label, params in fuzzy_sets[key].items():
            y_vals = compute_membership(universes[key], params)
            ax.plot(universes[key], y_vals, label=label, lw=2)

        val = metrics[key]
        ax.axvline(val, color='red', linestyle='--', alpha=0.8,
                   label=f'{ticker}: {val:.2f}')

        ax.set_title(f'Entrada: {labels[key]}')
        ax.set_ylim(0, 1.1)
        ax.legend(fontsize='small')
        ax.grid(alpha=0.2)

    # Salida
    ax = axs[1, 1]
    for label, params in fuzzy_sets['atr'].items():
        y_vals = compute_membership(universes['atr'], params)
        ax.plot(universes['atr'], y_vals, lw=2.5, label=label)

    ax.set_title('Salida: Atractivo')
    ax.set_ylim(0, 1.1)
    ax.legend(fontsize='small')
    ax.grid(alpha=0.2)

    plt.suptitle(f'Diccionario Difuso ({ticker})', y=0.95)
    plt.show()

metrics = {
    'vol': vol_anual,
    'ret': ret_anual,
    'sha': sharpe
}

plot_fuzzification(metrics, ticker)
```

Al observar el diseño de estos conjuntos difusos, podemos extraer conclusiones clave para la arquitectura de nuestro evaluador de portafolios:

<ol>
<li>Naturaleza de la Membresía y Robustez (Entradas $x_1, x_2, x_3$)
    <ul>
        <li><strong>La Membresía Dual:</strong> A diferencia de la lógica clásica, un activo puede tener una membresía de $\mu=0.8$ en 
            "Medio" y $\mu=0.2$ en "Alto" simultáneamente. Esta dualidad permite que el motor de inferencia combine reglas de distintos 
            perfiles de riesgo de forma balanceada.</li>
        <li><strong>Interpretación de Hombros y Mesetas:</strong> Las <strong>meseta</strong> ($\mu=1.0$) representan el núcleo del 
            concepto (donde el experto tiene certeza total). Las <strong>pendientes</strong> representan la transición o ambigüedad entre
            categorías. Los hombros extienden el conjunto difuso con valores por afuera del rango práctico: un valor mayor/menor mantiene
            pertenencia total al conjunto.</li>
        <li><strong>Robustez ante el Ruido:</strong> Si un dato cambia ligeramente debido al ruido del mercado (ej. una volatilidad de 
            $45\%$ a $46\%$), su grado de membresía varía suavemente. Esta <strong>gradualidad</strong> evita cambios erráticos en la 
            decisión final,  una ventaja crítica frente a los modelos optimizadores rígidos.</li>
    </ul>
</li>
<li>Configuración Geométrica y Escalas
    <ul>
        <li><strong>Independencia de Unidades:</strong> Cada variable tiene su propio <strong>Universo de Discurso</strong> 
            El Sharpe  Ratio opera en unidades cortas ($0$ a $4$), mientras que el Retorno permite valores negativos. La lógica difusa nos 
            permite  "homogeneizar" estas métricas heterogéneas al convertirlas a grados de verdad ($\mu \in [0, 1]$).</li>
        <li><strong>Hombros en los Extremos:</strong> Todas las gráficas utilizan trapecios en los bordes (ej. "Bajo" y "Alto") para 
            asegurar que valores fuera del rango esperado sigan siendo capturados con membresía total.</li>
        <li><strong>Cobertura difusa:</strong>Podemos observar  que en este caso no utilizamos una partición de Ruspini, sino lo que suele 
            llamarse una cobertura difusa. En este contexto no es necesario imponer esa normalización porque el objetivo del sistema es 
            preservar la ambigüedad y los conflictos inherentes al análisis financiero. Forzar una partición de unidad introduciría una 
            normalización artificial que diluye estas tensiones y empuja al sistema a producir interpretaciones más “limpias” pero menos 
            informativas. En este sentido, la ausencia de Ruspini favorece una representación más fiel del razonamiento heurístico de un 
            analista bajo incertidumbre.
        </li>
    </ul>
</li>
<li>Dinámica de la Salida (Atractivo $y$)
    <ul>
        <li><strong>Distribución de Etiquetas:</strong> Hemos usado una mezcla de <strong>trapecios</strong> en los extremos y 
            <strong>triángulos</strong> en el centro. La resolución es mayor (5 conjuntos) para permitir una decisión final más matizada y 
            granular.</li>
        <li><strong>El Concepto de "Recorte" (Clipping):</strong> Cuando una regla se dispare con un nivel de confianza (ej. $\alpha = 
            0.7$), no tomaremos toda la figura de la salida, sino que la "rebanaremos" horizontalmente a esa altura. El área resultante 
            bajo ese corte es lo que fluye hacia el agregador.</li>
        <li><strong>Resolución de la Decisión:</strong> Los espacios entre los picos determinan la sensibilidad. Si las funciones se 
            solapan demasiado, la decisión será muy "suave" o ambigua; si están muy separadas, el sistema se volverá más agresivo y 
            tendiente a los extremos.</li>
    </ul>
</li>
</ol>

Al observar la posición de la línea roja en los conjuntos difusos, podemos anticipar el comportamiento del sistema; analizando el caso pora NVIDIA (NVDA), tenemos:

1. **Fuzzificación de la Volatilidad:** NVIDIA presenta una volatilidad de ~50% (3/abr/26)) por lo que cae plenamente en el conjunto **Medio** ($\mu=1.0$) pero está cerca de empezar a ser considerada **Alta**. El sistema ya está "preparado" para activar reglas de riesgo moderado.

2. **Fuzzificación del Retorno:** El retorno anualizado, para estos datos, es superior al 50% (frecuente en activos de alto crecimiento), la línea roja queda al extremo derecho del conjunto *Positivo*. Gracias a los "hombros" de nuestras funciones trapezoidales, el sistema sigue reconociendo este valor como máximo atractivo sin errores de desbordamiento.

3. **El Sharpe Ratio como Validador:** Observemos la intersección en el tercer panel. NVIDIA tiene un Sharpe reducido (ej. $S < 1.0$), activando con gran fuerza la etiqueta *Pobre*. Aunque el retorno es alto, la volatilidad también lo es, lo que "castiga" la eficiencia. De entrada, esto sugiere que el activo ofrece grandes retornos, pero a un costo de riesgo muy elevado.

**2. Base de Reglas FIS (Conocimiento Experto)**

Una vez que los datos han sido fuzzificados, el **Motor de Inferencia** utiliza una base de reglas lingüísticas para aproximar el juicio de un analista financiero. En un sistema Mamdani, estas reglas capturan heurísticas y relaciones no lineales entre riesgo, retorno y eficiencia.

No existe un conjunto definido de reglas, de origen teórico (esto no suele ser común en entornos financieros), sin embargo, existen principios generales ampliamente aceptados en teoría financiera y en la práctica de inversión. En particular, podemos basarnos en cuatro ideas centrales:

1. **Trade-off riesgo–retorno (Markowitz)**:
   A mayor riesgo (volatilidad), mayor retorno esperado *en equilibrio* (fuera de eventos como crisis, burbujas, existencia de información asimétrica, regulaciones estictas o comportamiento de manada).
   *Ejemplo*: Acciones vs. bonos.
2. **Eficiencia riesgo–retorno (Sharpe)**:
   Optimizar el **retorno por unidad de riesgo** (Ratio de Sharpe), no solo el retorno absoluto.
   *Diferencia*: Un activo con Sharpe alto puede tener menor retorno nominal que otro, pero ser más eficiente.
3. **Aversión a pérdidas (Kahneman & Tversky)**:
   Las pérdidas pesan psicológicamente más que las ganancias equivalentes, distorsionando decisiones "racionales".
4. **Dominancia (Pareto)**:
   Un activo es preferible a otro si tiene mayor retorno para el mismo nivel de riesgo, o menor riesgo para el mismo retorno.

Bajo estos principios, es razonable un conjunto de reglas como el siguiente (observemos que cada analista financiero enunciará reglas distintas de acuerdo con su criterio y la importancia que de a los principios previos):

| ID | Perfil de Regla | Premisa (SI...) | Consecuente (ENTONCES...) | Interpretación experta |
| :--- | :--- | :--- | :--- | :--- |
| **R1** | **Preservación** | Volatilidad es **ALTA** AND Sharpe es **POBRE** | Atractivo es **MUY BAJO** | Alta volatilidad combinada con baja eficiencia implica riesgo elevado sin compensación suficiente |
| **R2** | **Evitación** | Retorno es **NEGATIVO** | Atractivo es **MUY BAJO** | Un retorno negativo es condición suficiente para descartar el activo |
| **R3** | **Conservadora** | Volatilidad es **MEDIA** AND Sharpe es **POBRE** | Atractivo es **BAJO** | Esta regla penaliza activos que no justifican el riesgo asumido |
| **R4** | **Equilibrio** | Volatilidad es **MEDIA** AND Retorno es **NEUTRO** | Atractivo es **MEDIO** | Representa un escenario neutro (riesgo medio, retorno medio); es un caso base sin señales claras| 
| **R5** | **Crecimiento** | Volatilidad es **BAJA** AND Retorno es **POSITIVO** | Atractivo es **ALTO** | El caso ideal desde el punto de vista clásico de portafolio: Bajo riesgo con retorno positivo es una combinación favorable |
| **R6** | **Oportunista** | Volatilidad es **MEDIA** AND Retorno es **POSITIVO** | Atractivo es **ALTO** | Retornos positivos con riesgo moderado pueden ser atractivos si el inversionista está dispuesto a tomar riesgos|
| **R7** | **Eficiencia** | Sharpe es **EXCELENTE** AND Volatilidad es **BAJA** | Atractivo es **MUY ALTO** | Alto Sharpe y baja volatilidad indican un activo de alta |
| **R8** | **Especulativa** | Volatilidad es **ALTA** AND Retorno es **POSITIVO** | Atractivo es **MEDIO** | Retorno positivo con alta volatilidad implica alto potencial, pero con riesgo significativo. Es típico de inversiones altamente especulativas (*criptoactivos*, por ejemplo)|


**Notas de Diseño**

<ol>
    <li style="margin-top:15px;"><strong>El Ratio de Sharpe como filtro de eficiencia</strong>: En reglas como R1, R3 y R7, el Ratio de 
        Sharpe actúa como un indicador de eficiencia riesgo-retorno. Un Sharpe "Pobre" penaliza el atractivo incluso si otras variables son 
        favorables, mientras que un Sharpe "Excelente" refuerza escenarios de alta calidad.</li>
    <li style="margin-top:15px;"><strong>Reglas con premisas parciales (R2)</strong>: No todas las reglas requieren todas las variables de 
        entrada. La regla R2 
        establece que un retorno negativo es condición suficiente para una evaluación pesimista, introduciendo un sesgo conservador en el 
        sistema.</li>
    <li style="margin-top:15px;"><strong>Interacción riesgo-retorno (R6 vs R8)</strong>: Un mismo retorno positivo puede derivar en 
        evaluaciones distintas 
        dependiendo del nivel de volatilidad:
        <ul>
            <li>Volatilidad media → oportunidad (R6)</li>
            <li>Volatilidad alta → especulación (R8)</li>
        </ul></li>
    <li style="margin-top:15px;"><strong>Resolución de conflictos</strong>: En esta implementación:
        <ul>
            <li>El operador <strong>mínimo ($\min$)</strong> se utiliza para evaluar el operador lógico AND  </li>
            <li>El operador <strong>máximo ($\max$)</strong> se utiliza para agregar las contribuciones de múltiples reglas hacia una 
                misma etiqueta de salida</li>
        </ul>
      </li>
    <li style="margin-top:15px;"><strong>Cobertura del espacio de decisión</strong>: La base de reglas cubre escenarios típicos de 
        inversión (conservador, 
        crecimiento, especulación), aunque pueden existir combinaciones no explícitamente modeladas, lo cual introduce ambigüedad 
        controlada en la inferencia.</li>
</ol>


```python
# ================================
# 1. BASE DE REGLAS 
# ================================

base_reglas = [
    {'id': 'R1', 'perfil': 'Preservación', 'antecedentes': {'vol': 'Alto', 'sha': 'Pobre'}, 'consecuente': 'Muy Bajo'},
    {'id': 'R2', 'perfil': 'Evitación',    'antecedentes': {'ret': 'Negativo'},             'consecuente': 'Muy Bajo'},
    {'id': 'R3', 'perfil': 'Conservadora', 'antecedentes': {'vol': 'Medio', 'sha': 'Pobre'},'consecuente': 'Bajo'},
    {'id': 'R4', 'perfil': 'Equilibrio',   'antecedentes': {'vol': 'Medio', 'ret': 'Neutro'},'consecuente': 'Medio'},
    {'id': 'R5', 'perfil': 'Crecimiento',  'antecedentes': {'vol': 'Bajo', 'ret': 'Positivo'},'consecuente': 'Alto'},
    {'id': 'R6', 'perfil': 'Oportunista',  'antecedentes': {'vol': 'Medio', 'ret': 'Positivo'},'consecuente': 'Alto'},
    {'id': 'R7', 'perfil': 'Eficiencia',   'antecedentes': {'sha': 'Excelente', 'vol': 'Bajo'},'consecuente': 'Muy Alto'},
    {'id': 'R8', 'perfil': 'Especulativa', 'antecedentes': {'vol': 'Alto', 'ret': 'Positivo'},'consecuente': 'Medio'}
]

print(f"Base de reglas cargada: {len(base_reglas)}")


# ================================
# 2. VISUALIZACIÓN DE REGLAS
# ================================

def plot_rule_regions():

    fig, ax = plt.subplots(figsize=(10, 6))

    # Curvas de salida
    for label, params in fuzzy_sets['atr'].items():
        y_vals = compute_membership(universes['atr'], params)
        ax.plot(universes['atr'], y_vals, lw=2.5, label=label)

    # Colores por etiqueta
    colors = {
        'Muy Bajo': 'tab:blue',
        'Bajo': 'tab:orange',
        'Medio': 'tab:green',
        'Alto': 'tab:red',
        'Muy Alto': 'tab:purple'
    }

    # Manejo automático de traslapes (ej. R4 y R8)
    offsets = {}

    for rule in base_reglas:
        label = rule['consecuente']
        a, b, c, d = fuzzy_sets['atr'][label]

        count = offsets.get(label, 0)
        offsets[label] = count + 1

        y_offset = 1.02 + count * 0.05

        ax.axvspan(b, c, alpha=0.15, color=colors[label])

        x_pos = (b + c) / 2
        ax.text(x_pos, y_offset, rule['id'],
                ha='center', va='bottom', fontsize=8)

    ax.set_title('Regiones de Activación de Reglas')
    ax.set_xlabel('Atractivo')
    ax.set_ylabel('Grado de membresía')
    ax.set_ylim(0, 1.2)
    ax.grid(alpha=0.2)
    ax.legend(fontsize='small')

    plt.show()


plot_rule_regions()
```

**3. Motor de Inferencia y Desfuzzificación**

Una vez que las reglas se disparan, el motor realiza el **recorte** de las funciones de salida. El resultado agregado es una superficie que representa la "opinión acumulada" de todas las reglas. Finalmente, aplicamos el método del **Centroide** para obtener un número único que un algoritmo de trading pueda procesar:

$$y^* = \frac{\int \mu(y) \cdot y \, dy}{\int \mu(y) \, dy}$$

Este valor $y^*$ es nuestra **Salida Estricta**, que cuantifica el atractivo final del activo.

```python
# ==============================================================================
# MOTOR DE INFERENCIA MAMDANI
# ==============================================================================

from IPython.display import display

# ================================
# 1. FUZZIFICACIÓN
# ================================

# Mapeo de nombres de entrada → claves del sistema
input_map = {
    'vol': input_values['volatilidad'],
    'ret': input_values['retorno'],
    'sha': input_values['sharpe']
}

# Cálculo de grados de pertenencia
mu = {}

for var, value in input_map.items():
    mu[var] = {
        label: membership_trapezoid(value, *params)
        for label, params in fuzzy_sets[var].items()
    }

# ================================
# 2. EVALUACIÓN DE REGLAS
# ================================

alfa_salidas = {label: 0.0 for label in fuzzy_sets['atr'].keys()}

for regla in base_reglas:
    
    grados = []
    
    for var, etiqueta in regla['antecedentes'].items():
        grados.append(mu[var].get(etiqueta, 1.0))
    
    # AND = mínimo
    nivel_disparo = np.min(grados) if grados else 0.0
    
    # OR (agregación por etiqueta)
    salida = regla['consecuente']
    alfa_salidas[salida] = max(alfa_salidas[salida], nivel_disparo)


# ================================
# 3. AGREGACIÓN
# ================================

y_universe = universes['atr']
y_agregado = np.zeros_like(y_universe)

plt.figure(figsize=(12, 6))

for etiqueta, alfa in alfa_salidas.items():
    
    params = fuzzy_sets['atr'][etiqueta]
    func_original = np.array([
        membership_trapezoid(y, *params) for y in y_universe
    ])
    
    if alfa > 0:
        # Curva base (referencia)
        plt.plot(y_universe, func_original, ':', lw=1, color='gray', alpha=0.4)
        
        # Nivel alpha
        plt.axhline(y=alfa, color='red', linestyle=':', alpha=0.3)
        plt.text(-3, alfa, f'α={alfa:.2f}', color='red', fontsize=8, va='center')
        
        # Etiqueta
        centro = y_universe[np.argmax(func_original)]
        plt.text(centro, alfa + 0.02, etiqueta,
                 color='darkorange', fontsize=10, ha='center')
        
        # Recorte Mamdani
        recorte = np.minimum(alfa, func_original)
        y_agregado = np.maximum(y_agregado, recorte)


# ================================
# 4. TABLA DE ACTIVACIÓN
# ================================

print(f"\nResumen de Activación de Conceptos para {ticker}:")
print("-" * 45)

df_alfas = pd.DataFrame([
    {"Etiqueta de Atractivo": k, "Nivel de Confianza (Alpha)": v}
    for k, v in alfa_salidas.items()
])

df_alfas = df_alfas.sort_values(by="Nivel de Confianza (Alpha)", ascending=False)

display(df_alfas.style.format({"Nivel de Confianza (Alpha)": "{:.4f}"}))


# ================================
# 5. DEFUZZIFICACIÓN (CENTROIDE)
# ================================

numerador = np.sum(y_universe * y_agregado)
denominador = np.sum(y_agregado)

y_final = numerador / denominador if denominador > 0 else 0


# ================================
# 6. VISUALIZACIÓN FINAL
# ================================

plt.fill_between(y_universe, 0, y_agregado,
                 alpha=0.3, label='Área Agregada')

plt.plot(y_universe, y_agregado,
         lw=3, label='Envolvente Difusa')

plt.axvline(x=y_final, linestyle='--', lw=2.5,
            label=f'Centroide: {y_final:.2f}')

plt.ylim(0, 1.1)
plt.xlim(0, 100)
plt.title(f'Resultado Mamdani para {ticker}')
plt.xlabel('Atractivo')
plt.ylabel('Membresía')
plt.legend(fontsize='small')
plt.grid(alpha=0.15)

plt.show()


# ================================
# 7. CONCLUSIÓN
# ================================

top = df_alfas.iloc[0]
etiqueta_dom = top["Etiqueta de Atractivo"]
confianza = top["Nivel de Confianza (Alpha)"]

print("\n>>> CONCLUSIÓN DE MAMDANI:")
print(f"Etiqueta dominante: '{etiqueta_dom}' (α = {confianza:.2f})")
print(f"Score final: {y_final:.2f} / 100")

print(f"\n>>> CONCLUSIÓN DEL EXPERTO:")
print(f"El atractivo de {ticker} es {y_final:.2f} / 100")
```

<br>

La "envolvente difusa" (zona sombreada) en la gráfica es la unión geométrica de todas las recomendaciones activadas por nuestra base de reglas. Este gráfico nos permite "abrir la caja negra" del modelo y entender el razonamiento del experto:

1.  **Recorte de Funciones (*Clipping*):** La salida muestra trapecios en lugar de los triángulos originales. Esto ocurre porque el sistema limita su optimismo o pesimismo mediante el nivel de disparo ($\alpha$). Si una regla se cumple solo al 60%, el sistema "rebana" la punta del conjunto de salida a esa altura, reconociendo que no existe evidencia total para esa etiqueta.
2.  **Conflictos Logicos:** Es común ver múltiples áreas activas (ej. una zona en "Bajo" y otra en "Alto"). Esto refleja la naturaleza de los mercados: un activo como **NVDA** puede tener retornos excepcionales (atractivo) pero una volatilidad extrema (riesgo).
4.  **El Centroide (La Decisión Final):** A diferencia de un sistema *If-Else* tradicional que elegiría una sola etiqueta, el método Mamdani calcula el centro de gravedad de toda el área. El valor resultante (0-100) pondera todas las reglas activas, entregando una calificación robusta que no se deja llevar ni por el pánico del riesgo ni por la euforia del retorno.

### B. Sistema de Tsukamoto

El sistema de Tsukamoto es una variante de Mamdani donde los conjuntos difusos de salida son **monotónicos** — funciones S o Z exclusivamente. Esto permite invertir la función de membresía y obtener directamente un valor numérico $y_i$ para cada regla:

$$\alpha_i = \mu_{C_i}(y_i) \implies y_i = \mu_{C_i}^{-1}(\alpha_i)$$

La salida final es el promedio ponderado de las salidas desfuzificadas:

$$y^* = \frac{\sum_i \alpha_i \cdot y_i}{\sum_i \alpha_i}$$


```python
# ==============================================================================
# CONVERSIÓN AUTOMÁTICA: MAMDANI → TSUKAMOTO MONÓTONO
# ==============================================================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# ================================
# 1. GENERADOR DE CONJUNTOS MONÓTONOS
# ================================

def build_tsukamoto_sets(atr_sets):
    """
    Convierte conjuntos trapezoidales en funciones monótonas.
    """
    labels = list(atr_sets.keys())
    n = len(labels)
    
    ts_sets = {}
    
    for i, label in enumerate(labels):
        a, b, c, d = atr_sets[label]
        
        # Muy Bajo → decreciente
        if i == 0:
            ts_sets[label] = [a, a, c, d]  # hombro izquierdo
        
        # Muy Alto → creciente
        elif i == n - 1:
            ts_sets[label] = [a, b, d, d]  # hombro derecho
        
        # Intermedios → creciente
        else:
            ts_sets[label] = [a, b, d, d]
    
    return ts_sets


tsukamoto_sets = build_tsukamoto_sets(fuzzy_sets['atr'])


# ================================
# 2. FUNCIÓN INVERSA (GARANTIZADA)
# ================================

def inverse_tsukamoto(alpha, params):
    a, b, c, d = params
    
    if alpha <= 0:
        return a
    if alpha >= 1:
        return d
    
    # creciente
    if b > a:
        return a + alpha * (b - a)
    
    # decreciente
    if d > c:
        return d - alpha * (d - c)
    
    return (a + d) / 2


# ================================
# 3. VISUALIZACIÓN COMPARATIVA
# ================================

y = universes['atr']

plt.figure(figsize=(10, 5))

for label in fuzzy_sets['atr']:
    mamdani = [membership_trapezoid(v, *fuzzy_sets['atr'][label]) for v in y]
    tsuka = [membership_trapezoid(v, *tsukamoto_sets[label]) for v in y]
    
    plt.plot(y, mamdani, '--', alpha=0.4)
    plt.plot(y, tsuka, label=f"{label} (TSK)")

plt.title("Conversión Mamdani → Tsukamoto (Monótono)")
plt.xlabel("Atractivo")
plt.ylabel("Membresía")
plt.legend(fontsize='small')
plt.grid(alpha=0.2)
plt.show()


# ================================
# 4. FUZZIFICACIÓN (REUTILIZADA)
# ================================

input_map = {
    'vol': input_values['volatilidad'],
    'ret': input_values['retorno'],
    'sha': input_values['sharpe']
}

mu = {
    var: {
        label: membership_trapezoid(input_map[var], *params)
        for label, params in fuzzy_sets[var].items()
    }
    for var in input_map
}


# ================================
# 5. EVALUACIÓN TSUKAMOTO
# ================================

resultados = []

for regla in base_reglas:
    
    grados = [
        mu[var][etiqueta]
        for var, etiqueta in regla['antecedentes'].items()
    ]
    
    alpha = np.min(grados) if grados else 0
    
    params = tsukamoto_sets[regla['consecuente']]
    z = inverse_tsukamoto(alpha, params)
    
    resultados.append({
        'Regla': regla['id'],
        'Alpha': alpha,
        'Salida z': z,
        'Perfil': regla['perfil']
    })


# ================================
# 6. AGREGACIÓN
# ================================

df = pd.DataFrame(resultados)

num = np.sum(df['Alpha'] * df['Salida z'])
den = np.sum(df['Alpha'])

score = num / den if den > 0 else 0


# ================================
# 7. RESULTADOS
# ================================

print("\n--- Tsukamoto (Monótono derivado de Mamdani) ---\n")

display(df.style.format({
    "Alpha": "{:.3f}",
    "Salida z": "{:.2f}"
}))

print(f"\nSCORE FINAL: {score:.2f} / 100")
```

A diferencia de Mamdani, que produce una *agregación difusa* (un área resultante sobre el universo de salida), el sistema de **Tsukamoto** genera un valor numérico directo por regla y agrega esos valores mediante un promedio ponderado. Esto lo hace particularmente adecuado cuando se requieren indicadores numéricos consistentes y comparables.

1. **Funciones de salida monótonas:**  
   En Tsukamoto, cada conjunto de salida se define mediante una función de membresía **monótona** (creciente o decreciente). Esto permite invertir la función para obtener un valor crisp por regla:
$$\alpha_i = \mu_{C_i}(y_i) \;\Rightarrow\; y_i = \mu_{C_i}^{-1}(\alpha_i)$$
En este caso, las funciones de salida monótonas se obtienen a partir de los conjuntos definidos para Mamdani, preservando su semántica pero garantizando invertibilidad.

2. **Agregación directa (sin centroide):**  
   En lugar de agregar áreas y calcular un centroide, el resultado global se obtiene mediante un promedio ponderado:
$$y^* = \frac{\sum_i \alpha_i\, y_i}{\sum_i \alpha_i}$$
Esto reduce el costo computacional y simplifica la interpretación.

3. **Naturaleza del resultado:**  
   Mientras Mamdani preserva la coexistencia de interpretaciones lingüísticas (por ejemplo, “bajo” y “medio” activos simultáneamente), Tsukamoto produce una salida numérica única por activo. Esto es útil cuando el objetivo es comparar, ordenar o automatizar decisiones, por ejemplo, en un sistema de monitoreo o priorización.

4. **Relación con Mamdani:**  
   Aquí, Tsukamoto no se trata como un modelo alternativo completamente independiente, sino como una **proyección funcional** del mismo sistema semántico usado en Mamdani, bajo la restricción adicional de monotonía en la salida.

En la siguiente tabla resumimos las diferencias entre ambos modelos:
| Aspecto | Mamdani | Tsukamoto |
|--------|---------|-----------|
| **Tipo de salida** | Conjunto difuso agregado (distribución) | Valor numérico directo por regla |
| **Funciones de salida** | Arbitrarias (no requieren monotonicidad) | Monótonas (invertibles) |
| **Mecanismo de inferencia** | Recorte y agregación de conjuntos | Inversión funcional por regla |
| **Defuzzificación** | Centroide (integración del área) | Promedio ponderado de salidas |
| **Interpretabilidad lingüística** | Alta: permite coexistencia de etiquetas | Moderada: cada regla produce un valor puntual |
| **Carga computacional** | Mayor (agregación y centroide) | Menor (cálculo directo) |
| **Adecuación típica** | Análisis cualitativo, explicación del sistema | Indicadores numéricos, automatización y ranking |
| **Rol en este trabajo** | Referencia semántica y comparativa | Proyección funcional basada en la misma semántica |

Mamdani mantiene explícitamente la ambigüedad lingüística del sistema y resulta más cercano a la forma en que un analista razona con etiquetas cualitativas. Tsukamoto impone estructura funcional adicional, produciendo un indicador numérico estable y comparable. En este trabajo, ambos comparten la misma base de reglas y universos lingüísticos; la diferencia radica en cómo cada modelo procesa y condensa la información.
   

### C. Sistema de Takagi-Sugeno-Kang (TSK)

El modelo **TSK** (Takagi, Sugeno y Kang) representa la evolución de la lógica difusa hacia la **aproximación funcional**. A diferencia de los modelos basados en conjuntos de salida semánticos (Mamdani), TSK utiliza funciones matemáticas explícitas de las variables de entrada como consecuentes. Es, en esencia, un modelo de **regresión local supervisado por lógica difusa**.

#### 1. Estructura de las Reglas y Órdenes del Sistema

En un sistema TSK, el conocimiento experto decide **qué modelo matemático** aplicar según las condiciones del antecedente. La forma general de una regla es:

$$\text{SI } x_1 \text{ es } A_i \text{ Y } x_2 \text{ es } B_i \implies y_i = f_i(x_1, x_2, \dots, x_n)$$

Dependiendo de la complejidad de la función $f_i$, clasificamos el sistema por su orden:

*   **Orden Cero:** El consecuente es una constante ($y_i = c_i$). Es ideal para clasificaciones discretas o cuando se busca un comportamiento similar a un promedio ponderado de valores puntuales.
*   **Primer Orden:** El consecuente es una ecuación lineal (un hiperplano):
    $$y_i = p_i x_1 + q_i x_2 + r_i$$
*   **Segundo Orden:** El consecuente es una ecuación cuadrática que permite capturar curvaturas e interacciones entre las variables de entrada:
    $$y_i = a_i x_1^2 + b_i x_2^2 + d_i (x_1 x_2) + p_i x_1 + q_i x_2 + r_i$$

#### 2. Discusión Formal de los Parámetros y su Interpretación Financiera

En la gestión de activos, los coeficientes actúan como **sensibilidades marginales**:

*   **Coeficientes Lineales ($p_i, q_i$):** Representan las derivadas parciales de la regla $$\frac{\partial y_i}{\partial x_j}$$ Si $p_i < 0$, la regla penaliza activamente la volatilidad (aversión al riesgo).
*   **Términos Cuadráticos ($a_i, b_i$):** Capturan la **aceleración del impacto**. Modelan la "no-linealidad" del pánico: un aumento de volatilidad en niveles ya altos (ej. del 40% al 45%) se penaliza más severamente que en niveles bajos.
*   **Términos de Interacción ($d_i$):** Capturan el **efecto cruzado**. Por ejemplo, un alto retorno puede ser positivo, pero si su interacción con una alta volatilidad es negativa, el sistema "desconfía" de ese retorno por considerarlo ruidoso.

#### 3. Mecanismo de Inferencia: El Modelo de Conmutación Suave

La salida global ($y^*$) no requiere defuzzificación por áreas; es una suma ponderada de las respuestas de cada regla activa:

$$y^* = \frac{\sum_{i=1}^{n} \alpha_i \cdot f_i(\mathbf{x})}{\sum_{i=1}^{n} \alpha_i}$$

#### 4. Criterio de Selección: ¿Qué orden elegir?

La elección del orden del modelo TSK depende del equilibrio entre **interpretabilidad** y **precisión**:

1.  **Elegir Orden Cero:** Cuando el objetivo es puramente cualitativo (ej. "Comprar", "Vender", "Mantener") y solo necesitamos un puntaje base sin importar las variaciones marginales dentro de la regla.
2.  **Elegir Primer Orden:** Es el estándar para análisis financiero. Se utiliza cuando se necesita que el atractivo del activo responda linealmente a cambios en el Sharpe Ratio o el Retorno dentro de un mismo régimen de mercado.
3.  **Elegir Segundo Orden:** Recomendado cuando se dispone de datos para entrenamiento (ANFIS) o cuando se desea modelar fenómenos de **convexidad y efectos de red** (interacciones entre variables). Es el más preciso pero el más propenso al *overfitting* si los coeficientes no se calibran con rigor estadístico.

Los valores tienden a concentrarse en extremos relativos: activos con combinaciones favorables de retorno y eficiencia reciben puntuaciones altas, mientras que aquellos con perfiles menos atractivos son penalizados con claridad. Esto refleja su naturaleza como modelo de regresión local, donde la salida depende explícitamente de combinaciones lineales de las variables de entrada.

Finalmente, el modelo **Takagi-Sugeno de segundo orden (TSK2)** introduce efectos no lineales que modifican de forma importante la evaluación. En general, se observa una reducción de puntajes en activos con mayor volatilidad o relaciones inestables entre riesgo y retorno. Este comportamiento es consistente con la inclusión de términos cuadráticos e interacciones, que amplifican el impacto del riesgo en niveles elevados. Observemos como se aleja con fuerza de activos con bajo Sharpe como Intel o Bitcoin.

En conjunto, la gráfica ilustra cómo distintos enfoques difusos producen interpretaciones complementarias del mismo problema. Más que existir un modelo “correcto”, cada uno enfatiza un aspecto distinto del análisis: estabilidad, sensibilidad, linealidad o penalización no lineal del riesgo.

### Medición del Conflicto entre reglas difusos (Diagnóstico de consenso)

La gráfica sugiere que el sistema Mamdani actúa como un **integrador de consenso**: amplifica señales cuando múltiples reglas coinciden y atenúa cuando existe conflicto. Para formalizar esta intuición, podemos definir una métrica explícita de **conflicto difuso** de la sigueine manera:

Sea el conjunto de reglas activas $i = 1, \dots, n$, con:

- $\alpha_i$: nivel de activación de la regla  
- $y_i$: salida de la regla 

Definimos el **conflicto como la dispersión ponderada**:

$$\text{Conflicto} = \sqrt{\frac{\sum_i \alpha_i (y_i - \mu_y)^2}{\sum_i \alpha_i}}$$

donde 

$$\mu_y = \frac{\sum_i \alpha_i y_i}{\sum_i \alpha_i}$$

Es, decir, el podemos calcular el conflicto como una **desviación ponderada difusa**.

En el caso del modelo de Mandani, no contamos con el valor de $y_i$ para cada conjunto (la defuzzificación se realiza despues de integrar todas las reglas), entonces, calculamos una estimación de la salida de la regla como una aproximación representativa del conjunto de salida asociado a cada regla, utilizando el centro del núcleo (o centroide local) del conjunto difuso correspondiente:

$$y_i \approx \frac{b_i + c_i}{2}$$

donde $[a_i, b_i, c_i, d_i]$ son los parámetros del trapecio que define la etiqueta lingüística del consecuente de la regla. Calculando los valores de acuerdo a los datos actuales obtenemos:


```python
# ==============================================================================
# COMPARACIÓN DE MODELOS DIFUSOS + MÉTRICA DE CONFLICTO (DESV. ESTÁNDAR)
# ==============================================================================

# ================================
# FUNCIÓN AUXILIAR: CONFLICTO
# ================================

def conflicto_desviacion(alphas, outputs):
    alphas = np.array(alphas)
    outputs = np.array(outputs)
    
    if np.sum(alphas) == 0:
        return 0.0
    
    mu_y = np.sum(alphas * outputs) / np.sum(alphas)
    var = np.sum(alphas * (outputs - mu_y)**2) / np.sum(alphas)
    
    return np.sqrt(var)


# ================================
# 1. MAMDANI (aproximación por núcleo)
# ================================

alphas_mamdani = []
outputs_mamdani = []

for regla in base_reglas:
    grados = [
        mu[var][etiqueta]
        for var, etiqueta in regla['antecedentes'].items()
    ]
    
    alpha = np.min(grados) if grados else 0.0
    
    if alpha > 0:
        etiqueta = regla['consecuente']
        a, b, c, d = fuzzy_sets['atr'][etiqueta]
        
        # Centro del núcleo
        y_i = (b + c) / 2
        
        alphas_mamdani.append(alpha)
        outputs_mamdani.append(y_i)

conf_mamdani = conflicto_desviacion(alphas_mamdani, outputs_mamdani)


# ================================
# 2. TSUKAMOTO
# ================================

alphas_tsuka = []
outputs_tsuka = []

for regla in base_reglas:
    grados = [
        mu[var][etiqueta]
        for var, etiqueta in regla['antecedentes'].items()
    ]
    
    alpha = np.min(grados) if grados else 0.0
    
    if alpha > 0:
        etiqueta = regla['consecuente']
        a, b, c, d = fuzzy_sets['atr'][etiqueta]
        
        # Inversión aproximada (monotónica)
        if etiqueta in ['Muy Bajo', 'Bajo']:
            y_i = d - alpha * (d - a)
        else:
            y_i = a + alpha * (d - a)
        
        alphas_tsuka.append(alpha)
        outputs_tsuka.append(y_i)

conf_tsuka = conflicto_desviacion(alphas_tsuka, outputs_tsuka)


# ================================
# 3. TSK PRIMER ORDEN
# ================================

alphas_tsk1 = []
outputs_tsk1 = []

v = input_map['vol']
r = input_map['ret']
s = input_map['sha']

for regla in base_reglas:
    grados = [
        mu[var][etiqueta]
        for var, etiqueta in regla['antecedentes'].items()
    ]
    
    alpha = np.min(grados) if grados else 0.0
    
    if alpha > 0:
        m = modelos_consecuentes[regla['consecuente']]
        
        y_i = (
            m['p'] * v +
            m['q'] * r +
            m['r'] * s +
            m['c']
        )
        
        y_i = np.clip(y_i, 0, 100)
        
        alphas_tsk1.append(alpha)
        outputs_tsk1.append(y_i)

conf_tsk1 = conflicto_desviacion(alphas_tsk1, outputs_tsk1)


# ================================
# 4. TSK SEGUNDO ORDEN
# ================================

alphas_tsk2 = []
outputs_tsk2 = []

for regla in base_reglas:
    grados = [
        mu[var][etiqueta]
        for var, etiqueta in regla['antecedentes'].items()
    ]
    
    alpha = np.min(grados) if grados else 0.0
    
    if alpha > 0:
        m = modelos_segundo_orden[regla['consecuente']]
        
        y_i = (
            m['v2'] * v**2 +
            m['r2'] * r**2 +
            m['vr'] * v * r +
            m['v'] * v +
            m['r'] * r +
            m['s'] * s +
            m['c']
        )
        
        y_i = np.clip(y_i, 0, 100)
        
        alphas_tsk2.append(alpha)
        outputs_tsk2.append(y_i)

conf_tsk2 = conflicto_desviacion(alphas_tsk2, outputs_tsk2)


# ================================
# 5. RESULTADOS FINALES
# ================================

df_modelos = pd.DataFrame([
    {"Modelo": "Mamdani",   "Score": y_final,            "Conflicto": conf_mamdani},
    {"Modelo": "Tsukamoto", "Score": score,    "Conflicto": conf_tsuka},
    {"Modelo": "TSK-1",     "Score": score_tsk,          "Conflicto": conf_tsk1},
    {"Modelo": "TSK-2",     "Score": score_tsk2,   "Conflicto": conf_tsk2},
])

display(df_modelos.style.format({
    "Score": "{:.2f}",
    "Conflicto": "{:.2f}"
}))
```

## Ejercicio avanzado: Toma de decisiones colectiva bajo incertidumbre

En la gestión de activos y contextos organizacionales complejos, las decisiones críticas rara vez recaeran en un solo algoritmo o individuo. El desafío real surge al integrar un grupo de actores con niveles heterogéneos de conocimiento, intereses y percepciones. 

Este escenario introduce desafíos que la estadística tradicional a menudo simplifica en exceso:
*   **Imprecisión Semántica:** "¿Qué significa 'atractivo' para un economista frente a un ingeniero?" Las opiniones son cualitativas.
*   **Influencia y Sesgo:** Los participantes pueden verse presionados por la autoridad o la tendencia del grupo (efecto rebaño).
*   **Experticia Variable:** No todas las opiniones deben pesar lo mismo; es necesario discriminar la calidad del aporte según la **reputación**.

---

### El Antecedente Clásico: El Método Delphi
Como vimos con el modelo SIR en epidemiología, el punto de partida estándar en toma de decisiones es el **Método Delphi**. Es un proceso iterativo basado en el anonimato y la retroalimentación controlada.

<p align="center">
  <img src="images/Método_Delphi.png" width="600">
</p>

**Limitaciones del Delphi Tradicional:**
1.  **Ambigüedad del lenguaje**: Trata etiquetas como "Probable" como números rígidos, perdiendo matices.
2.  **Ponderación Estática**: Usualmente aplica una media aritmética simple ($\frac{1}{n}$), ignorando que algunos expertos son consistentemente más precisos que otros.

---

### La Evolución Difusa: Método MQDM (*moviQuest Decision Making*)

El método **MQDM** evoluciona el concepto de Delphi al integrarlo con **Lógica Difusa** y **Redes Sociales Estructuradas**. Podemos clasificarlo como un **Sistema de Inferencia tipo Mamdani Orientado a Consenso**. A diferencia de los modelos vistos anteriormente, aquí los "sensores" son humanos y las "reglas" se forman mediante la **agregación ponderada de reputaciones**.

#### A. Representación del "Sentimiento" Difuso
A diferencia de los triángulos que usamos para el portafolio, el MQDM utiliza **curvas logísticas** para modelar la opinión (ej. escala Likert: *Disagree, Neutral, Agree*). 
*   **Fundamento Psicológico:** El cambio de opinión humana no es lineal; tendemos a permanecer en "estados cómodos" o atractores de baja energía. Las curvas capturan esa transición suave entre el desacuerdo y el consenso.

#### B. Ponderación Dinámica de los Agentes ($W_j$)
La "inteligencia" del sistema reside en cómo calcula el peso de cada opinión. El peso $W_j$ de un agente no es un parámetro fijo, sino un **indicador de confianza multivariable**:

$$W_j^i = \kappa_j (1 - \bar{E}_j) \left( \frac{2}{1 + e^{-8\Omega_\gamma}} - 1 \right) \left( 1 - \omega (1 - G(X_j^{(i)}, X_j^{(i-1)})) \right)$$

Este bloque es un **Motor de Inferencia** en sí mismo que evalúa:
*   $\kappa_j$: **Autoconfianza** (¿Qué tan seguro dice estar el experto?).
*   $\bar{E}_j$: **Error Histórico** (¿Qué tan preciso ha sido en el pasado?).
*   $\Omega_\gamma$: **Maestría del Grupo** (¿Pertenece a un departamento técnico o administrativo?).
*   $\omega$ e $G(\cdot)$: **Factor de Inercia y Similitud**, que evitan que el sistema sea manipulado por cambios bruscos de opinión entre rondas.

#### C. Agregación y Estabilidad Iterativa
La opinión colectiva $O^{(i)}$ se obtiene mediante una **Agregación Ponderada** de las evaluaciones difusas. El proceso no busca un valor puntual inmediato, sino la **Convergencia**:

$$O^{(i)} = \frac{\sum_{j=1}^{n} W_j^i V_j^i}{\sum_{j=1}^{n} W_j^i}$$

El sistema se detiene solo cuando se alcanza la **estabilidad iterativa**, es decir, cuando las opiniones han convergido hacia un consenso robusto que unifica la sabiduría de los expertos con la precisión del historial de datos.


```python
# DATOS DE PRUEBA
# =====================

def generar_datos(opcion="yahoo", ticker="NVDA", n_points=1000): 
    """
    Genera y grafica df_crisp_data.
    Opciones: 'yahoo', 'lorenz', 'mackey-glass'
    """
    plt.figure(figsize=(12, 5))
    
    if opcion == "yahoo":
        raw = yf.download(ticker, period="2y", interval="1d", auto_adjust=True)
        if isinstance(raw.columns, pd.MultiIndex):
            raw.columns = raw.columns.get_level_values(0)
        raw['Returns'] = np.log(raw['Close'] / raw['Close'].shift(1))
        df_crisp_data = raw.dropna().copy()
        
        df_crisp_data['Close'].plot(color='royalblue')
        plt.title(f"Serie de Tiempo: Precios Históricos {ticker} (Yahoo Finance)")
        plt.ylabel("Precio")

    elif opcion == "lorenz":
        dt = 0.01
        xs, ys, zs = np.empty(n_points), np.empty(n_points), np.empty(n_points)
        xs[0], ys[0], zs[0] = (0., 1., 1.05)
        for i in range(n_points - 1):
            xs[i + 1] = xs[i] + (10 * (ys[i] - xs[i])) * dt
            ys[i + 1] = ys[i] + (xs[i] * (28 - zs[i]) - ys[i]) * dt
            zs[i + 1] = zs[i] + (xs[i] * ys[i] - (8/3) * zs[i]) * dt
        df_crisp_data = pd.DataFrame({'Value': xs, 'Returns': np.gradient(xs)})
        
        plt.plot(df_crisp_data['Value'], color='darkorange')
        plt.title("Serie de Tiempo: Atractor de Lorenz (Componente X)")
        plt.ylabel("Amplitud")

    elif opcion == "mackey-glass":
        tau, n, beta, gamma = 17, 10, 0.2, 0.1
        mg = np.ones(n_points + tau)
        for i in range(tau, n_points + tau - 1):
            mg[i+1] = mg[i] + beta * mg[i-tau] / (1 + mg[i-tau]**10) - gamma * mg[i]
        series = mg[tau:]
        df_crisp_data = pd.DataFrame({'Value': series, 'Returns': np.gradient(series)})
        
        plt.plot(df_crisp_data['Value'], color='seagreen')
        plt.title("Serie de Tiempo: Ecuación de Mackey-Glass")
        plt.ylabel("Amplitud")

    plt.grid(True, alpha=0.3)
    plt.show()
    return df_crisp_data

# --- SELECCIÓN DEL CONJUNTO DE DATOS ---
tipo_serie = "lorenz" # "yahoo", "lorenz", "mackey-glass"
ticker_fin = ""   # "NVDA", "AAPL", "BTC-USD", "GC=F" (Oro) o "" para Lorenz y MackeyGlass

df_crisp_data = generar_datos(opcion=tipo_serie, ticker=ticker_fin)
```

#### A. El Diccionario de Conceptos: Fuzzificación del Sentimiento
En MQDM, la **Base de Datos Lingüística** no utiliza triángulos simples, sino **funciones logísticas** (sigmoides) para mapear las opiniones (ej. escalas Likert). Esta elección no es arbitraria, sino que responde a un fundamento de **neurodinámica**:

*   **Estados de Confort (Atractores):** El modelo asume que las posiciones "En desacuerdo", "Neutral" y "De acuerdo" actúan como estados de baja energía o "cómodos" para el ser humano. 
*   **Transiciones No Lineales:** El cambio de una opinión a otra no es proporcional; los individuos tienden a resistirse al cambio hasta que la evidencia es abrumadora. Las curvas logísticas capturan este comportamiento "S", donde la transición en las fronteras de los conjuntos es más realista que una rampa lineal.
*   **Parámetros Críticos:** Cada conjunto se define por su valor modal ($\nu$) y su ancho de banda ($\lambda$), lo que permite que el sistema maneje la **ambigüedad del lenguaje** (ej. lo que para un experto es "Neutral", para otro puede tener un sesgo hacia el acuerdo).

#### B. El Motor de Inferencia: Reglas Difusas Dinámicas
A diferencia de los sistemas de control clásicos donde las reglas son estáticas (SI A ENTONCES B), en MQDM las reglas de inferencia son **dinámicas y adaptativas**. El sistema genera una "regla de autoridad" por cada participante en cada ronda:

*   **Ponderación de la Verdad:** El peso $W_j^i$ actúa como el **nivel de disparo ($\alpha$)** de la opinión de un agente. Si un experto tiene una alta reputación ($\bar{E}_j$) y pertenece a un grupo de alta maestría ($\Omega_\gamma$), su "regla" dominará la superficie de agregación.
*   **Manejo de Conflictos:** El motor utiliza la **Similitud de Gower ($G$)** para detectar inconsistencias. Si un agente cambia de opinión drásticamente entre rondas, el **Factor de Inercia ($\omega$)** recorta automáticamente su peso, protegiendo al sistema de participantes indecisos o influenciables.
*   **Fusión de Señales Heterogéneas:** El motor combina variables psicológicas (autoconfianza $\kappa_j$) con variables de rendimiento histórico (precisión $\bar{E}_j$), logrando una inferencia que pondera la **sabiduría colectiva** por encima del simple promedio.

#### C. Control por Estabilidad Iterativa
El proceso de inferencia en MQDM no es un evento único (*one-shot*), sino un ciclo de **convergencia**. El motor de inferencia es controlado por un lazo de retroalimentación basado en la estabilidad:

*   **Retroalimentación Controlada:** Tras cada ronda, el sistema presenta un "Candidate Solution" (el resultado agregado actual). Los agentes revisan esta tendencia y pueden reajustar su postura (Plead/Revote).
*   **Criterio de Parada ($\varepsilon$):** El sistema alcanza la "decisión estable" cuando la diferencia entre la opinión colectiva actual $O^{(i)}$ y la anterior $O^{(i-1)}$ es menor a un umbral de tolerancia. 
*   **Optimización del Consenso:** Esta iteración permite que el "ruido" de las opiniones extremas se disipe, permitiendo que la masa de la decisión se desplace hacia un equilibrio robusto, similar a cómo el **Centroide** de Mamdani encuentra el equilibrio de una forma geométrica compleja.

Hemos pasado de un sistema experto cerrado a un ecosistema de inteligencia colectiva donde la lógica difusa actúa como el árbitro de la verdad.


### Contexto del ejemplo

A lo largo de esta sección usaremos como hilo conductor la **calibración del modelo SIR**. El modelo SIR describe la dinámica de una epidemia mediante tres compartimentos:

$$\frac{dS}{dt} = -\beta \frac{SI}{N}, \qquad \frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I, \qquad \frac{dR}{dt} = \gamma I$$

donde $\beta$ es la tasa de contagio y $\gamma$ la tasa de recuperación. El número reproductivo básico $R_0 = \beta / \gamma$ determina la dinámica global:

- $R_0 > 1$: la epidemia crece.
- $R_0 = 1$: la epidemia se estabiliza en el pico.
- $R_0 < 1$: la epidemia decrece.

Nuestro sistema difuso tomará como entrada observable la **tasa de cambio de infectados** $\Delta I = I(t) - I(t-1)$ y el **nivel actual de infectados** $I(t)$, ambos normalizados respecto a la población $N$, y producirá como salida una estimación difusa de $\beta$ — el parámetro más sensible y difícil de medir directamente.

---

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

BG       = "#09090E"
CYAN     = "#00FFFF"
AMBER    = "#FFB300"
LAVENDER = "#B388FF"
GREEN    = "#69FF47"
PINK     = "#FF6EC7"
WHITE    = "#E8E8F0"
GRAY     = "#3A3A4A"

# ── Universos de discurso ─────────────────────────────────────────────────────
x_dI  = np.linspace(-0.05, 0.05, 500)   # ΔI/N  — tasa de cambio de infectados
x_I   = np.linspace(0, 1, 500)          # I/N   — nivel de infectados
x_b   = np.linspace(0, 1, 500)          # β     — tasa de contagio estimada

# ── Funciones auxiliares ──────────────────────────────────────────────────────
def trimf(x, a, b, c):
    return np.maximum(0, np.minimum((x-a)/(b-a+1e-10), (c-x)/(c-b+1e-10)))

def trapmf(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x-a)/(b-a+1e-10), 1), (d-x)/(d-c+1e-10)))

def S(x, a, b):
    m = (a+b)/2
    return np.where(x<=a, 0, np.where(x<=m, 2*((x-a)/(b-a))**2,
           np.where(x<=b, 1-2*((x-b)/(b-a))**2, 1))).astype(float)

def Z(x, a, b):
    return 1 - S(x, a, b)

# ── Membresías de entrada ─────────────────────────────────────────────────────
# ΔI/N
mf_dI = {
    "decreciendo": Z(x_dI, -0.02, 0.0),
    "estable":     trimf(x_dI, -0.01, 0.0, 0.01),
    "creciendo":   S(x_dI, 0.0, 0.02),
}
# I/N
mf_I = {
    "bajo":  Z(x_I, 0.1, 0.35),
    "medio": trimf(x_I, 0.2, 0.5, 0.8),
    "alto":  S(x_I, 0.55, 0.85),
}
# β (salida Mamdani)
mf_b = {
    "muy_bajo": Z(x_b, 0.0, 0.25),
    "bajo":     trimf(x_b, 0.1, 0.25, 0.45),
    "medio":    trimf(x_b, 0.35, 0.5, 0.65),
    "alto":     trimf(x_b, 0.55, 0.7, 0.85),
    "muy_alto": S(x_b, 0.7, 0.95),
}

# ── Punto de observación ──────────────────────────────────────────────────────
obs_dI = 0.018   # epidemia creciendo rápido
obs_I  = 0.62    # nivel de infectados alto

# ── Fuzzificación ─────────────────────────────────────────────────────────────
def fuzzify(val, x_univ, mfs):
    return {k: float(np.interp(val, x_univ, v)) for k, v in mfs.items()}

deg_dI = fuzzify(obs_dI, x_dI, mf_dI)
deg_I  = fuzzify(obs_I,  x_I,  mf_I)

# ── Base de reglas: (dI_label, I_label) → beta_label ─────────────────────────
rules = [
    ("creciendo",   "alto",  "muy_alto"),
    ("creciendo",   "medio", "alto"),
    ("estable",     "medio", "medio"),
    ("decreciendo", "medio", "bajo"),
    ("decreciendo", "bajo",  "muy_bajo"),
]

# ── Inferencia (min para T-norma, max para agregación) ───────────────────────
agg = np.zeros_like(x_b)
activations = []

for dI_lbl, I_lbl, b_lbl in rules:
    alpha = min(deg_dI[dI_lbl], deg_I[I_lbl])
    clipped = np.minimum(alpha, mf_b[b_lbl])
    agg = np.maximum(agg, clipped)
    activations.append((dI_lbl, I_lbl, b_lbl, alpha, clipped))

# ── Defuzzificación ───────────────────────────────────────────────────────────
# COG
cog = np.sum(x_b * agg) / (np.sum(agg) + 1e-10)

# MOM
max_val = np.max(agg)
mom_mask = agg >= max_val - 1e-4
mom = np.mean(x_b[mom_mask])

# ── Figura ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(18, 14), facecolor=BG)
gs  = gridspec.GridSpec(3, 3, figure=fig, hspace=0.6, wspace=0.35)

def styled(ax, title, xlabel="", ylabel="$\\mu$"):
    ax.set_facecolor(BG)
    ax.set_title(title, color=WHITE, fontsize=9, fontfamily="monospace", pad=5)
    ax.set_xlabel(xlabel, color=GRAY, fontsize=8)
    ax.set_ylabel(ylabel, color=GRAY, fontsize=8)
    ax.tick_params(colors=GRAY, labelsize=7.5)
    for sp in ax.spines.values(): sp.set_edgecolor(GRAY)
    ax.axhline(0, color=GRAY, lw=0.4, ls="--", alpha=0.3)
    ax.axhline(1, color=GRAY, lw=0.4, ls="--", alpha=0.3)
    return ax

colors_in = [CYAN, AMBER, LAVENDER]

# Fila 0 — membresías ΔI/N
ax0 = styled(fig.add_subplot(gs[0, 0:2]), "Membresías de entrada — $\\Delta I/N$", "$\\Delta I/N$")
for (lbl, y), col in zip(mf_dI.items(), colors_in):
    ax0.plot(x_dI, y, color=col, lw=2, label=lbl)
    ax0.fill_between(x_dI, y, alpha=0.07, color=col)
ax0.axvline(obs_dI, color=WHITE, lw=1.2, ls=":", alpha=0.7, label=f"obs={obs_dI}")
ax0.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 0 — membresías I/N
ax1 = styled(fig.add_subplot(gs[0, 2]), "Membresías de entrada — $I/N$", "$I/N$")
for (lbl, y), col in zip(mf_I.items(), colors_in):
    ax1.plot(x_I, y, color=col, lw=2, label=lbl)
    ax1.fill_between(x_I, y, alpha=0.07, color=col)
ax1.axvline(obs_I, color=WHITE, lw=1.2, ls=":", alpha=0.7, label=f"obs={obs_I}")
ax1.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 1 — reglas activadas
rule_colors = [CYAN, AMBER, LAVENDER, GREEN, PINK]
for idx, (dI_lbl, I_lbl, b_lbl, alpha, clipped) in enumerate(activations):
    ax = styled(fig.add_subplot(gs[1, idx % 3]),
                f"R{idx+1}: {dI_lbl} ∧ {b_lbl}\nα={alpha:.3f}", "$\\beta$")
    ax.plot(x_b, mf_b[b_lbl], color=rule_colors[idx], lw=1.5, ls="--", alpha=0.5)
    ax.fill_between(x_b, clipped, alpha=0.25, color=rule_colors[idx])
    ax.plot(x_b, clipped, color=rule_colors[idx], lw=2)
    ax.axhline(alpha, color=WHITE, lw=0.8, ls=":", alpha=0.5)

# Fila 2 — agregación + defuzzificación
ax_agg = styled(fig.add_subplot(gs[2, :]), "Agregación y defuzzificación — Mamdani", "$\\beta$")
ax_agg.fill_between(x_b, agg, alpha=0.15, color=LAVENDER)
ax_agg.plot(x_b, agg, color=LAVENDER, lw=2.2, label="Conjunto agregado")
ax_agg.axvline(cog, color=CYAN,  lw=2, ls="--", label=f"COG  β* = {cog:.3f}")
ax_agg.axvline(mom, color=AMBER, lw=2, ls="--", label=f"MOM  β* = {mom:.3f}")
ax_agg.set_xlim(0, 1)
ax_agg.legend(fontsize=9, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

fig.suptitle("Sistema Mamdani — Estimación de β (modelo SIR)",
             color=WHITE, fontsize=12, fontfamily="monospace", y=0.995)
plt.show()
```

```python
# ── TSK de orden 1 sobre el mismo problema SIR ────────────────────────────────
# Reutiliza x_dI, x_I, mf_dI, mf_I, deg_dI, deg_I, obs_dI, obs_I de la celda anterior

# Parámetros lineales por regla: β_i = p_i * ΔI/N + q_i * I/N + r_i
# Calibrados heurísticamente para el dominio [0,1]
tsk_rules = [
    ("creciendo",   "alto",  dict(p=3.5,  q=0.6,  r=0.10)),
    ("creciendo",   "medio", dict(p=2.8,  q=0.4,  r=0.08)),
    ("estable",     "medio", dict(p=0.5,  q=0.3,  r=0.35)),
    ("decreciendo", "medio", dict(p=-1.0, q=0.2,  r=0.30)),
    ("decreciendo", "bajo",  dict(p=-2.0, q=0.1,  r=0.15)),
]

alphas, y_tsk = [], []
for dI_lbl, I_lbl, coef in tsk_rules:
    alpha = min(deg_dI[dI_lbl], deg_I[I_lbl])
    yi    = coef["p"] * obs_dI + coef["q"] * obs_I + coef["r"]
    yi    = np.clip(yi, 0, 1)
    alphas.append(alpha)
    y_tsk.append(yi)

alphas  = np.array(alphas)
y_tsk   = np.array(y_tsk)
beta_tsk = np.sum(alphas * y_tsk) / (np.sum(alphas) + 1e-10)

# ── Figura ────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5), facecolor=BG)

# Panel izquierdo — contribución de cada regla
ax = axes[0]
ax.set_facecolor(BG)
rule_labels = [f"R{i+1}\n{r[0][:4]}∧{r[1][:3]}" for i, r in enumerate(tsk_rules)]
bars = ax.bar(rule_labels, alphas, color=[CYAN, AMBER, LAVENDER, GREEN, PINK],
              alpha=0.8, width=0.5)
ax2 = ax.twinx()
ax2.set_facecolor(BG)
ax2.scatter(rule_labels, y_tsk, color=WHITE, zorder=5, s=60, label="$y_i$ (salida regla)")
ax2.set_ylabel("$y_i = \\beta_i$ estimado", color=WHITE, fontsize=9)
ax2.tick_params(colors=GRAY, labelsize=8)
ax2.set_ylim(0, 1)
ax.set_ylabel("Grado de activación $\\alpha_i$", color=GRAY, fontsize=9)
ax.set_ylim(0, 1)
ax.set_title("TSK — Activación y salida por regla", color=WHITE,
             fontsize=10, fontfamily="monospace")
ax.tick_params(colors=GRAY, labelsize=8)
for sp in ax.spines.values(): sp.set_edgecolor(GRAY)
for sp in ax2.spines.values(): sp.set_edgecolor(GRAY)

# Panel derecho — promedio ponderado visual
ax = axes[1]
ax.set_facecolor(BG)
weighted = alphas * y_tsk
ax.barh(rule_labels, weighted, color=[CYAN, AMBER, LAVENDER, GREEN, PINK], alpha=0.75)
ax.axvline(beta_tsk * np.sum(alphas), color=WHITE, lw=1.5, ls="--", alpha=0.6)
ax.set_xlabel("$\\alpha_i \\cdot y_i$", color=GRAY, fontsize=9)
ax.set_title(f"TSK — Contribución ponderada\n$\\beta^* = {beta_tsk:.3f}$",
             color=WHITE, fontsize=10, fontfamily="monospace")
ax.tick_params(colors=GRAY, labelsize=8)
for sp in ax.spines.values(): sp.set_edgecolor(GRAY)

fig.suptitle("Sistema TSK (orden 1) — Estimación de β (modelo SIR)",
             color=WHITE, fontsize=12, fontfamily="monospace")
fig.patch.set_facecolor(BG)
plt.tight_layout()
plt.show()

print(f"TSK  →  β* estimado = {beta_tsk:.4f}")
```

```python
# ── Tsukamoto sobre el mismo problema SIR ────────────────────────────────────
# Conjuntos de salida MONOTÓNICOS (S o Z) — invertibles analíticamente

def S_inv(alpha, a, b):
    """Inversa de la función S: dado alpha, devuelve y tal que S(y;a,b)=alpha."""
    m = (a + b) / 2
    alpha = np.clip(alpha, 1e-6, 1 - 1e-6)
    if alpha <= 0.5:
        return a + (b - a) * np.sqrt(alpha / 2)
    else:
        return b - (b - a) * np.sqrt((1 - alpha) / 2)

def Z_inv(alpha, a, b):
    """Inversa de la función Z."""
    return S_inv(1 - alpha, a, b)

# Conjuntos de salida monotónicos para β
tsuka_output_sets = {
    "muy_bajo": ("Z", 0.05, 0.30),
    "bajo":     ("Z", 0.20, 0.50),
    "medio_S":  ("S", 0.35, 0.65),
    "alto":     ("S", 0.50, 0.80),
    "muy_alto": ("S", 0.65, 0.95),
}

tsuka_rules = [
    ("creciendo",   "alto",  "muy_alto"),
    ("creciendo",   "medio", "alto"),
    ("estable",     "medio", "medio_S"),
    ("decreciendo", "medio", "bajo"),
    ("decreciendo", "bajo",  "muy_bajo"),
]

alphas_t, y_tsuka = [], []
for dI_lbl, I_lbl, b_lbl in tsuka_rules:
    alpha = min(deg_dI[dI_lbl], deg_I[I_lbl])
    kind, a, b = tsuka_output_sets[b_lbl]
    yi = S_inv(alpha, a, b) if kind == "S" else Z_inv(alpha, a, b)
    yi = np.clip(yi, 0, 1)
    alphas_t.append(alpha)
    y_tsuka.append(yi)

alphas_t  = np.array(alphas_t)
y_tsuka   = np.array(y_tsuka)
beta_tsuka = np.sum(alphas_t * y_tsuka) / (np.sum(alphas_t) + 1e-10)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(16, 10), facecolor=BG)
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.55, wspace=0.35)

rule_colors = [CYAN, AMBER, LAVENDER, GREEN, PINK]

# Fila 0 — conjunto de salida + punto invertido por regla
for idx, (dI_lbl, I_lbl, b_lbl, alpha, yi) in enumerate(
        zip(*zip(*tsuka_rules), alphas_t, y_tsuka)):
    ax = fig.add_subplot(gs[0, idx % 3])
    ax.set_facecolor(BG)
    kind, a, b = tsuka_output_sets[b_lbl]
    y_mf = S(x_b, a, b) if kind == "S" else Z(x_b, a, b)
    ax.plot(x_b, y_mf, color=rule_colors[idx], lw=2, ls="--", alpha=0.6)
    ax.axhline(alpha, color=WHITE, lw=1, ls=":", alpha=0.5, label=f"α={alpha:.3f}")
    ax.axvline(yi,    color=rule_colors[idx], lw=2, label=f"y={yi:.3f}")
    ax.scatter([yi], [alpha], color=WHITE, zorder=5, s=50)
    ax.set_title(f"R{idx+1}: {b_lbl}", color=WHITE, fontsize=9, fontfamily="monospace")
    ax.set_xlim(0, 1); ax.set_ylim(-0.05, 1.15)
    ax.set_xlabel("$\\beta$", color=GRAY, fontsize=8)
    ax.set_ylabel("$\\mu$",   color=GRAY, fontsize=8)
    ax.tick_params(colors=GRAY, labelsize=7.5)
    for sp in ax.spines.values(): sp.set_edgecolor(GRAY)
    ax.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 1 — resumen ponderado
ax_sum = fig.add_subplot(gs[1, :])
ax_sum.set_facecolor(BG)
rule_labels = [f"R{i+1}" for i in range(len(tsuka_rules))]
x_pos = np.arange(len(tsuka_rules))
bars  = ax_sum.bar(x_pos, alphas_t, width=0.35, color=rule_colors, alpha=0.6, label="$\\alpha_i$")
ax_sum.scatter(x_pos, y_tsuka, color=WHITE, zorder=5, s=70, label="$y_i$ (invertido)")
ax_sum.axhline(beta_tsuka, color=CYAN, lw=2, ls="--",
               label=f"$\\beta^*$ Tsukamoto = {beta_tsuka:.3f}")
ax_sum.set_xticks(x_pos); ax_sum.set_xticklabels(rule_labels, color=WHITE)
ax_sum.set_ylim(0, 1.1)
ax_sum.set_ylabel("valor", color=GRAY, fontsize=9)
ax_sum.set_title("Resumen — promedio ponderado Tsukamoto", color=WHITE,
                 fontsize=10, fontfamily="monospace")
ax_sum.tick_params(colors=GRAY, labelsize=8)
for sp in ax_sum.spines.values(): sp.set_edgecolor(GRAY)
ax_sum.legend(fontsize=8.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

fig.suptitle("Sistema Tsukamoto — Estimación de β (modelo SIR)",
             color=WHITE, fontsize=12, fontfamily="monospace", y=0.995)
plt.show()

print(f"Tsukamoto  →  β* estimado = {beta_tsuka:.4f}")

# ── Comparativa final ─────────────────────────────────────────────────────────
print("\n── Comparativa de estimaciones de β ──")
print(f"  Mamdani  COG : {cog:.4f}")
print(f"  Mamdani  MOM : {mom:.4f}")
print(f"  TSK ord.1    : {beta_tsk:.4f}")
print(f"  Tsukamoto    : {beta_tsuka:.4f}")
```

# Capítulo 4: Lógica Difusa y Sistemas Híbridos

## Modelos Suaves para Fenómenos Inciertos y Adaptativos

Este cuaderno implementa los conceptos fundamentales de la Lógica Difusa, explorando cómo la **pertenencia parcial** permite modelar conceptos lingüísticos vagos (como "alto" o "cerca de") y cómo esta lógica se convierte en un mecanismo **adaptativo y metaheurístico** cuando sus parámetros son dinámicos.

La clave para la Ciencia de Datos Avanzada es la **intepretabilidad** que ofrece esta lógica en combinación con el poder de optimización de los algoritmos evolutivos.

## 1. Funciones de Pertenencia: Cuantificando el Límite Suave

Las funciones de pertenencia $(\mu(x))$ son la base de la Lógica Difusa, ya que definen el grado con el que un valor numérico pertenece a un concepto lingüístico (un conjunto difuso). La forma de esta función (triangular, trapezoidal, gaussiana) determina cómo de gradual es la transición de "no pertenencia" a "pertenencia total".

```python
# [CÓDIGO 1: Definición de Funciones de Pertenencia]

def triangular_fuzzy(x, a, b, c):
    """Función de pertenencia triangular. a <= b <= c."""
    if x <= a or x >= c:
        return 0.0
    elif a < x <= b:
        return (x - a) / (b - a)
    else:
        return (c - x) / (c - b)

def trapezoidal_fuzzy(x, a, b, c, d):
    """Función de pertenencia trapezoidal. a <= b <= c <= d."""
    if x <= a or x >= d:
        return 0.0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1.0
    else:
        return (d - x) / (d - c)

def gaussian_fuzzy(x, c, w):
    """
    Función de pertenencia Gaussiana (suave y continua).
    c: centro (valor modal), w: ancho (dispersión).
    """
    return np.exp(-((x - c)**2) / (w**2))

print("Funciones de pertenencia cargadas.")
```

```python
# [CÓDIGO 1.2: Visualización Comparativa]

xs = np.linspace(-10, 10, 500)

# Parámetros para las funciones
tri = np.array([triangular_fuzzy(x, -5, 0, 5) for x in xs])
trap = np.array([trapezoidal_fuzzy(x, -7, -3, 3, 7) for x in xs])
gauss = gaussian_fuzzy(xs, c=0, w=3)

plt.figure(figsize=(12, 6))
plt.plot(xs, tri, label="Triangular (Bordes Duros)")
plt.plot(xs, trap, label="Trapezoidal (Zona Plana)")
plt.plot(xs, gauss, label="Gaussiana (Suave y Diferenciable)")
plt.legend(title="Forma de la Función")
plt.title("Comparación de Funciones de Pertenencia Difusas: Cuantificando el Concepto 'Cerca de Cero'")
plt.xlabel("Valor de la Variable de Entrada (x)")
plt.ylabel("Grado de Pertenencia ($\mu$)")
plt.show()

# 
```

## 3. Conjuntos Difusos Dinámicos: El Núcleo Adaptativo

La limitación de los conjuntos difusos estáticos es su incapacidad para modelar sistemas no estacionarios (cuyas categorías relevantes cambian con el tiempo). El **Conjunto Difuso Dinámico** permite que sus parámetros internos (`c` y `w`) evolucionen mediante un mecanismo similar a una mutación, lo que lo convierte en una entidad adaptable, ideal para la optimización metaheurística.

```python
# [CÓDIGO 3: Conjunto Difuso Dinámico y Mutación]

class DynamicFuzzySet:
    """
    Representa un conjunto difuso (Gaussiano) cuyos parámetros (centro y ancho)
    pueden ser ajustados o mutados a lo largo del tiempo.
    """
    def __init__(self, c, w, name=""):
        self.c = float(c) # Centro (valor modal)
        self.w = float(w) # Ancho (dispersión)
        self.name = name

    def mu(self, x):
        """Función de pertenencia Gaussiana."""
        # Utilizamos la función gaussiana por ser suave y diferenciable
        return np.exp(-((x - self.c)**2) / (self.w**2))

    def mutate(self, sigma_c=0.1, sigma_w=0.05):
        """
        Aplica una perturbación aleatoria a los parámetros, simulando
        la adaptación o la exploración en un proceso evolutivo.
        """
        self.c += np.random.normal(0, sigma_c)
        # Aseguramos que el ancho (w) no sea negativo ni demasiado pequeño.
        self.w = max(0.01, self.w + np.random.normal(0, sigma_w))

    def get_params(self):
        return (self.c, self.w)

# Ejemplo de evolución de un conjunto dinámico
dfs = DynamicFuzzySet(c=0, w=2, name="Medio")

history_c = []
history_w = []

for t in range(50):
    # El conjunto se adapta suavemente en cada 'paso de tiempo' o 'generación'
    dfs.mutate(sigma_c=0.2, sigma_w=0.1)
    history_c.append(dfs.c)
    history_w.append(dfs.w)

plt.figure(figsize=(12, 5))
plt.plot(history_c, label="Centro $c$ (Posición adaptativa)")
plt.plot(history_w, label="Ancho $w$ (Sensibilidad/Dispersión adaptativa)")
plt.legend()
plt.title("Evolución de Parámetros en un Conjunto Difuso Dinámico")
plt.xlabel("Paso de Tiempo / Iteración Evolutiva")
plt.ylabel("Valor del Parámetro")
plt.show()

# 
```

## 4. Sistemas Takagi–Sugeno (TS): Modelado Suave y Local

El modelo Takagi–Sugeno (TS) utiliza reglas donde el consecuente es un modelo lineal. La salida global del sistema se calcula como el promedio ponderado de las salidas de todos los modelos locales, donde los pesos son los grados de pertenencia (activación) de las reglas. Esto garantiza una superficie de respuesta continua y no lineal.

```python
# [CÓDIGO 4: Sistema de Inferencia Takagi–Sugeno]

def ts_model(x, rules):
    """
    Calcula la salida de un sistema Takagi-Sugeno de primer orden (consecuente lineal).
    """
    num, den = 0, 0
    for fuzzy_set, linear_fun in rules:
        # 1. Fuzzificación: Obtener el grado de pertenencia (mu)
        mu = fuzzy_set.mu(x)
        
        # 2. Consecuente: Calcular la salida local (y_local = ax + b)
        y_local = linear_fun(x)
        
        # 3. Agregación (suma ponderada: mu * y_local)
        num += mu * y_local
        den += mu
        
    # 4. Normalización (Defuzzificación)
    return num / den if den > 1e-6 else 0


# Definición de dos conjuntos difusos para particionar el espacio
A1 = DynamicFuzzySet(c=-3, w=2, name="Izquierda")
A2 = DynamicFuzzySet(c=3, w=2, name="Derecha")

# Definición de los modelos locales (consecuentes)
f1 = lambda x: 0.5 * x + 1    # Modelo para la región Izquierda
f2 = lambda x: -0.3 * x + 2   # Modelo para la región Derecha

# Estructura de las reglas: SI x es A_i ENTONCES y = f_i(x)
rules = [(A1, f1), (A2, f2)]

# Simulación y visualización
xs = np.linspace(-10, 10, 400)
ys = np.array([ts_model(x, rules) for x in xs])

plt.figure(figsize=(12, 6))
plt.plot(xs, ys, label="Salida Global TS", color='blue', linewidth=3)

# Mostrar la contribución de los modelos locales
plt.plot(xs, f1(xs), '--', label="Modelo Local f1 (en región A1)", alpha=0.6)
plt.plot(xs, f2(xs), '--', label="Modelo Local f2 (en región A2)", alpha=0.6)

# Mostrar los conjuntos difusos (el 'peso' de cada regla)
plt.plot(xs, [A1.mu(x) for x in xs], ':', label="Conjunto A1 (Pertenencia)")
plt.plot(xs, [A2.mu(x) for x in xs], ':', label="Conjunto A2 (Pertenencia)")

plt.legend()
plt.title("Sistema Takagi–Sugeno (TS): Interpolación Suave entre Modelos Locales")
plt.xlabel("Entrada x")
plt.ylabel("Salida y / Pertenencia $\mu$")
plt.ylim(-4, 4)
plt.show()

# 
```

## 5. Metaheurística y Lógica Difusa: Optimización de Parámetros

El ajuste de los parámetros del sistema TS (los centros $c_i$ y anchos $w_i$ de los conjuntos difusos) es un problema de optimización no lineal. Aquí es donde los algoritmos evolutivos (AG, DE, PSO) entran en juego, utilizando el **Error Cuadrático Medio (MSE)** como la función de aptitud (fitness) a minimizar. Los parámetros se codifican en el 'cromosoma' de la metaheurística.

```python
# [CÓDIGO 5: Función de Aptitud para Optimización Evolutiva (Conceptual)]

# 5.1 Generación de Datos Sintéticos No Lineales (Target)
np.random.seed(42)
xs_data = np.linspace(-10, 10, 100)
# Función a modelar (Ejemplo: seno amortiguado con ruido)
target_y = (np.sin(xs_data / 2) + np.cos(xs_data / 10)) * 2 + np.random.normal(0, 0.2, len(xs_data))

# 5.2 Función de Aptitud (Fitness) para una Metaheurística
def ts_fitness(params, xs, target):
    """
    Calcula el Error Cuadrático Medio (MSE) para un conjunto de parámetros (c1, w1, c2, w2).
    Esta es la función objetivo a minimizar por la metaheurística.
    """
    c1, w1, c2, w2 = params
    
    # Temporalmente ajustamos los conjuntos A1 y A2 con los parámetros propuestos por el optimizador
    A1_temp = DynamicFuzzySet(c1, abs(w1))
    A2_temp = DynamicFuzzySet(c2, abs(w2))
    rules_temp = [(A1_temp, f1), (A2_temp, f2)] # Usamos los mismos modelos locales f1, f2
    
    # Generar predicciones
    preds = np.array([ts_model(x, rules_temp) for x in xs])
    
    # Error Cuadrático Medio (MSE)
    return np.mean((preds - target)**2)

# Ejemplo de cálculo de fitness (usando los parámetros iniciales de A1 y A2)
initial_params = (A1.c, A1.w, A2.c, A2.w)
initial_error = ts_fitness(initial_params, xs_data, target_y)

print(f"El error inicial del modelo TS con parámetros no optimizados (MSE): {initial_error:.4f}")
print("Un algoritmo evolutivo buscaría un vector de parámetros que minimice este valor.")

plt.figure(figsize=(10, 5))
plt.scatter(xs_data, target_y, label='Datos Objetivo (Ruidosos)', s=10)
plt.plot(xs_data, [ts_model(x, rules) for x in xs_data], label='Predicción Inicial TS', color='red')
plt.legend()
plt.title('Modelo TS Inicial vs. Datos Objetivo')
```

## 6. Agentes Difusos y Sistemas Multiagente

Al encapsular un Conjunto Difuso Dinámico y una regla TS en una clase `FuzzyAgent`, se crea una **sociedad de agentes especializados**. Cada agente domina y modela una región del espacio. La salida global es el resultado de la cooperación ponderada de todos los agentes. Este enfoque mejora la robustez y la capacidad de adaptación ante cambios de régimen.

```python
# [CÓDIGO 6: Agentes Difusos y Modelo de Sociedad]

class FuzzyAgent:
    """
    Un agente con una región perceptual (fset) y un modelo local (model).
    """
    def __init__(self, c, w, model, name=""):
        # El conjunto difuso dinámico es la región de especialización del agente
        self.fset = DynamicFuzzySet(c, w, name=name) 
        self.model = model

    def perceive(self, x):
        """Devuelve el grado de activación (pertenencia) del agente."""
        return self.fset.mu(x)

    def act(self, x):
        """Calcula la salida de su modelo local."""
        return self.model(x)

    def adapt(self, sigma=0.05):
        """Permite que la región de especialización del agente evolucione."""
        self.fset.mutate(sigma, sigma/2)
        
    def __repr__(self):
        return f"Agente '{self.fset.name}' (c={self.fset.c:.2f}, w={self.fset.w:.2f})"


def fuzzy_society_output(x, agents):
    """
    Combina las salidas de todos los agentes ponderadas por su grado de percepción.
    """
    weights = np.array([ag.perceive(x) for ag in agents])
    outputs = np.array([ag.act(x) for ag in agents])
    
    sum_weights = np.sum(weights)
    
    if sum_weights < 1e-6:
        return 0
        
    # Salida agregada (promedio ponderado)
    return np.dot(weights, outputs) / sum_weights


# Ejemplo con tres Agentes especializados
agents = [
    FuzzyAgent(-6, 2, lambda x: 0.3 * x + 1, name="Régimen Bajo"),
    FuzzyAgent( 0, 1.5, lambda x: 2,           name="Régimen Medio"),
    FuzzyAgent( 6, 2, lambda x: -0.2 * x + 4, name="Régimen Alto"),
]

print("Agentes especializados en distintas regiones:")
for agent in agents:
    print(agent)

xs_agent = np.linspace(-10, 10, 400)
ys_society = [fuzzy_society_output(x, agents) for x in xs_agent]

plt.figure(figsize=(12, 6))
plt.plot(xs_agent, ys_society, label="Modelo Colectivo (Sociedad Difusa)", color='purple', linewidth=3)

# Visualizar la especialización de los agentes
for agent in agents:
    mu = [agent.perceive(x) for x in xs_agent]
    plt.plot(xs_agent, mu, ':', alpha=0.7, label=f"Percepción {agent.fset.name}")

plt.legend(loc='lower left')
plt.title("Sistema Multiagente Difuso: Cooperación por Especialización")
plt.xlabel("Entrada x")
plt.ylabel("Salida Agregada / Pertenencia $\mu$")
plt.show()

# 
```

## 7. Sistemas Difusos Fractales: Interpretando Estructuras Complejas

En sistemas caóticos o complejos, los atractores presentan estructuras fractales. Un conjunto difuso puede utilizarse para asignar un **significado conceptual suave** a una región geométrica compleja (como un lóbulo o un anillo). La función `fuzzy_shell` ilustra cómo se puede definir un concepto como "cercano a una distancia específica de un centro".

```python
# [CÓDIGO 7: "Cáscara Difusa" alrededor de una Estructura]

def fuzzy_shell(x, center, radius, softness):
    """
    Función de pertenencia que define una "cáscara" o "anillo" alrededor de un centro,
    útil para envolver regiones de un atractor (ej. un lóbulo del atractor de Lorenz).
    """
    # Distancia euclidiana del punto al centro
    dist = np.linalg.norm(x - center)
    
    # Cuantifica la cercanía a la distancia ideal (radius) mediante una Gaussiana
    return np.exp(-((dist - radius)**2) / (softness**2))

# Generamos puntos simulando una sección anular ruidosa de un atractor 2D
np.random.seed(42)
num_points = 1000
angles = np.random.uniform(0, 2 * np.pi, num_points)
r_base = 2.0
radii = r_base + np.random.normal(0, 0.4, num_points) # Añadir ruido
points = np.column_stack([radii * np.cos(angles), radii * np.sin(angles)])

# Calculamos el grado de pertenencia a la 'Cáscara' de radio 2.0
shell_mu = np.array([fuzzy_shell(p, center=np.array([0.0, 0.0]),
                                 radius=2.0, softness=0.5) for p in points])

plt.figure(figsize=(8, 8))
# Coloreamos cada punto por su grado de pertenencia a la 'cáscara'
scatter = plt.scatter(points[:, 0], points[:, 1], c=shell_mu, cmap='plasma', alpha=0.7)
plt.colorbar(scatter, label="Grado de Pertenencia Difusa ($\mu$)")
plt.title("Representación Difusa ('Cáscara') de una Estructura Compleja 2D")
plt.xlabel("Dimensión X (Estado 1)")
plt.ylabel("Dimensión Y (Estado 2)")
plt.axis('equal')
plt.show()

# 
```

En la realidad, la nitidez es una rareza. Los límites que definen categorías relevantes para el modelado de sistemas —desde la "temperatura alta" de una planta industrial hasta la "alta volatilidad" de un mercado financiero, o el "nivel de excitación" de una señal cerebral— no son líneas divisorias estrictas. Son, por naturaleza, **gradientes, matices y zonas intermedias**. La mayor parte de los fenómenos naturales, biológicos y económicos se expresan a través de transiciones suaves.La lógica clásica o Booleana, con su rigor binario (verdadero/falso; 1/0), es la base formal de las matemáticas y la informática, y funciona perfectamente en entornos donde la verdad es absoluta. Sin embargo, cuando esta lógica rígida se enfrenta al **lenguaje humano, la imprecisión en las mediciones o la vaguedad inherente a los conceptos cualitativos**, se vuelve dramáticamente insuficiente.La **Lógica Difusa (Fuzzy Logic)**, introducida por Lotfi Zadeh en 1965, surge precisamente para resolver esta insuficiencia. Su idea fundamental es simple, pero profundamente revolucionaria: **los conceptos pueden tener grados de verdad**.En lugar de forzar una decisión binaria, la lógica difusa nos permite expresar la pertenencia de un elemento a un concepto lingüístico mediante un número continuo en el intervalo $[0, 1]$:- Un valor de 0 implica la no-pertenencia plena.- Un valor de 1 implica la pertenencia plena.- Un valor intermedio (ej. 0.7) indica un **grado de pertenencia parcial**.Esta graduación nos proporciona la herramienta matemática para codificar la intuición humana y los fenómenos continuos, permitiendo que un dato pertenezca simultáneamente, aunque con diferente intensidad, a múltiples categorías (ej. "moderada" y "alta").

Un **conjunto difuso ($A$)** no es más que la representación formal de un concepto impreciso. Mientras que un conjunto clásico solo permite la inclusión o exclusión total de un elemento, un conjunto difuso define, para cada elemento de su universo de discurso ($U$), un **grado de pertenencia** $\mu_A(x)$.Esta relación se define a través de una **función de pertenencia** ($\mu_A: U \to [0, 1]$). Es la curva, la forma de esta función, la que modela matemáticamente el concepto lingüístico (ej. "caliente" o "pequeño").Las formas más comunes que se emplean en ingeniería y ciencia de datos son:- **Triangulares y Trapezoidales:** Son simples, computacionalmente económicas y se usan a menudo en sistemas de control y prototipado rápido. Permiten zonas de pertenencia completa (valor 1).- **Gaussianas (y Campana Generalizada):** Son suaves y continuas. Su naturaleza diferenciable las hace ideales para ser ajustadas o **optimizadas mediante algoritmos de gradiente o técnicas metaheurísticas**.- **Sigmoidales:** Se emplean para modelar transiciones abiertas, como el concepto de "muy grande" o "muy pequeño", donde la pertenencia tiende gradualmente a 0 o 1 en los límites del universo.La capacidad de definir estas fronteras de manera **suave** (en contraste con los límites abruptos de los conjuntos clásicos) es lo que convierte a la lógica difusa en un lenguaje tan adecuado para los sistemas continuos.

La verdadera potencia de la lógica difusa reside en su capacidad para modelar el razonamiento humano a través de **reglas de inferencia** expresadas en lenguaje natural:$$\text{SI (Condición Difusa) ENTONCES (Consecuente Difuso o Numérico)}$$Ejemplo:$$\text{SI} \quad (\text{Velocidad} \ \text{es} \ \text{ALTA}) \quad \text{Y} \quad (\text{Distancia} \ \text{es} \ \text{CORTA})$$$$\text{ENTONCES} \quad (\text{Frenado} \ \text{es} \ \text{FUERTE})$$Al sustituir las variables lógicas por grados de pertenencia, la inferencia difusa se convierte en un mecanismo de interpolación suave entre reglas de conocimiento.Los dos modelos principales de inferencia son:- **Modelo Mamdani (Salida Difusa):** El consecuente es un conjunto difuso ("Frenado es FUERTE"). La salida final requiere un proceso de **defuzzificación** para obtener un valor numérico ejecutable (ej. qué tanta presión aplicar al freno).- **Modelo Takagi–Sugeno (TS) (Salida Numérica/Funcional):** El consecuente es una función lineal o constante de las variables de entrada (ej. $y = ax + b$). Su estructura de modelado local es excepcionalmente útil para sistemas dinámicos.

Es fundamental entender la lógica difusa bajo un doble enfoque, especialmente en un contexto de ciencia de datos avanzada.### 4.1. Como Herramienta de RepresentaciónEn su forma estática, la lógica difusa es un **lenguaje no clásico** para:- **Suavizar Fronteras:** Proporciona una alternativa continua a las fronteras discretas de los **Rough Sets**, que definen conceptos con límites aproximados ("seguramente en" / "posiblemente en") pero carecen de gradualidad.- **Codificar Conocimiento Experto:** Permite a un ingeniero o científico modelar un sistema basado en reglas intuitivas, algo inalcanzable para la Lógica Temporal o Modal, que son por naturaleza simbólicas y discretas.- **Tratamiento de Datos Continuos y Ruidosos:** La continuidad de sus funciones de pertenencia la hace ideal para procesar señales y series temporales con ruido.### 4.2. Como Mecanismo Adaptativo (Metaheurístico)Aquí es donde la Lógica Difusa se integra directamente con las **Metaheurísticas**. Un conjunto difuso está definido por parámetros (ej. centro, ancho o dispersión). Si permitimos que estos parámetros **cambien, evolucionen o se ajusten** en respuesta a la dinámica del entorno o a la optimización de un error de predicción, la lógica difusa deja de ser solo un lenguaje estático y se convierte en un **componente activo de la adaptación y la búsqueda**.Esto se logra cuando:- Los parámetros de las funciones de pertenencia se codifican como cromosomas y se ajustan mediante **Algoritmos Genéticos (AG), Optimización por Enjambre de Partículas (PSO), o Estrategias Evolutivas (DE)**.- La estructura misma de las reglas (la selección de variables de entrada, la topología) se optimiza mediante **Programación Genética (PG)**.En este contexto híbrido, la lógica difusa actúa como un **sensor inteligente** cuya propia percepción del mundo se afina iterativamente para minimizar el error, un proceso intrínsecamente metaheurístico.

La principal limitación de los conjuntos difusos tradicionales es su **naturaleza atemporal y fija**. Un conjunto difuso fijo (ej. "Precio Alto") puede ser una representación instantánea adecuada, pero no captura la **evolución temporal** de las categorías de un sistema dinámico o no estacionario. Por ejemplo, el concepto de "demanda normal de energía" varía drásticamente entre el verano y el invierno.El concepto de **Conjunto Difuso Dinámico ($A_D$)** resuelve esta limitación al incluir una dimensión temporal en la función de pertenencia. Para un valor dado $x$, su grado de pertenencia en $A$ cambia a lo largo del tiempo $T$ según una **función temporal de pertenencia** $\mu^x(t): T \to [0, 1]$.Los conjuntos difusos dinámicos pueden verse como superficies en el espacio $U \times T \times [0, 1]$. En la práctica, esto implica que los parámetros que definen la forma del conjunto difuso (principalmente el **valor modal** —la posición— y el **ancho de banda** —la dispersión—) cambian con el tiempo o en función de otras variables contextuales.- **Aplicación a Series de Tiempo:** Una variante de estos, los **conjuntos difusos dinámicos suaves** (introducidos en tu tesis), han demostrado ser superiores a los fijos para la **predicción de series de tiempo no estacionarias**, ya que su lógica se adapta al régimen cambiante del fenómeno subyacente (tendencias crecientes o decrecientes).De esta manera, cada conjunto difuso se convierte en una **entidad que aprende y se desplaza**, actuando como un sensor perceptivo adaptable a los cambios de régimen de un sistema complejo.

Los Sistemas de Inferencia Takagi–Sugeno (TS) ofrecen una de las arquitecturas híbridas más eficientes. Su fortaleza radica en el balance entre **interpretabilidad lingüística** (a través de los conjuntos difusos del antecedente) y **potencia de modelado numérico** (a través de las funciones del consecuente).En un modelo TS, las reglas no tienen un consecuente difuso como en Mamdani, sino un modelo local, típicamente lineal o de orden cero:$$\text{SI } (x_1 \ \text{es} \ A_i) \ \text{Y} \ (x_2 \ \text{es} \ B_i) \quad \text{ENTONCES} \quad y = a_i x_1 + b_i x_2 + c_i$$Cada regla actúa como un **modelo local** que es válido en una pequeña región del espacio de estados (la definida por los conjuntos difusos $A_i$ y $B_i$). El sistema agrega las salidas de todos los modelos locales mediante un **promedio ponderado suave**, donde los pesos son los grados de pertenencia de las reglas activadas.$$y_{\text{final}} = \frac{\sum_{i=1}^{R} \omega_i y_i}{\sum_{i=1}^{R} \omega_i}$$El resultado es un **modelo global continuo, suave y no lineal** construido a partir de piezas locales simples y cuya lógica subyacente es completamente interpretable. Esta capacidad de interpolación suave entre modelos lineales lo convierte en una herramienta excelente para:- **Control Robótico y de Procesos:** Asegurando transiciones estables.- **Predicción de Series de Tiempo No Lineales:** Modelando diferentes regímenes de comportamiento (ej. crecimiento, colapso, estancamiento) con modelos locales especializados.

Al estudiar sistemas caóticos como Lorenz o series financieras, se observan **patrones fractales** en el espacio de estados. Estos atractores complejos se caracterizan por tener múltiples escalas y densas estructuras auto-similares.La lógica difusa ofrece un mecanismo elegante para **suavizar la interpretación** de estas estructuras. En lugar de tener que describir cada punto del atractor, un sistema difuso (generalmente TS) permite:- Definir reglas que corresponden a **regiones conceptuales** del atractor (ej. "el lóbulo superior", "el punto de equilibrio").- Capturar las **transiciones** entre subcomportamientos mediante la naturaleza gradual de las funciones de pertenencia.El resultado es un **sistema híbrido** que utiliza un conjunto de reglas lingüísticas para capturar, de manera interpretable, la dinámica compleja de un atractor fractal.

Llevando el concepto de conjunto difuso dinámico (Sección 5) un paso más allá, un **agente difuso** es una entidad que no solo ajusta su percepción, sino que también **toma decisiones, compite y se especializa** en una región específica del espacio de estados.En un **Sistema Multiagente (SMA) Difuso**, cada agente puede poseer:- Una **Región Perceptual Difusa** (su área de especialización, definida por un conjunto difuso dinámico).- Un **Modelo Local** (su regla TS para esa región).- Una **Función de Aptitud** y **Mecanismos de Mutación/Adaptación** que le permiten evolucionar.La robustez de los SMA difusos se debe a la **diversidad** y a la **cooperación/competencia** de los agentes. Si un sistema dinámico cambia de régimen (ej. una crisis financiera), algunos agentes (los especializados en baja volatilidad) pueden volverse inactivos, mientras que los nuevos agentes (especializados en alta volatilidad) asumen el control de la predicción o la acción. Este enfoque distribuido y adaptable supera a menudo la robustez de un modelo monolítico como una red neuronal simple.

En la práctica moderna de la Ciencia de Datos, el enfoque más poderoso es el **híbrido**. La lógica difusa no busca reemplazar, sino potenciar, otras técnicas:| Técnica | Aporte Clave al Híbrido || :--- | :--- || **Lógica Difusa** | Proporciona **interpretación** del conocimiento, **suavidad** en las transiciones y un **lenguaje** adaptable. || **Algoritmos Evolutivos (AG, DE, PSO)** | Se encargan del **ajuste fino y global** de los parámetros de los conjuntos difusos para minimizar el error de predicción. || **Redes Neuronales** | Aportan un poder de **modelado numérico local** que puede usarse como consecuente de una regla (ej. un nodo neuronal en el consecuente de una regla TS). || **Programación Genética** | Permite **optimizar la estructura** (las reglas) del sistema, no solo sus parámetros. |

La Lógica Difusa es mucho más que una técnica algorítmica; es una **filosofía de modelado** que reconoce la realidad del mundo como un continuo de matices. Nos permite pasar de un mundo de límites absolutos a un mundo de gradientes, ambigüedades y transiciones.En este capítulo, hemos transitado desde los conjuntos difusos estáticos hasta los **agentes difusos dinámicos y autónomos**. Hemos visto cómo la integración con las metaheurísticas convierte estos modelos en **entidades vivas** capaces de:- Interpretar los datos en términos conceptuales.- Adaptar su propia percepción (sus conjuntos difusos) a los cambios de régimen.- Cooperar y competir para mantener un modelo colectivo robusto.El resultado final es una plataforma para construir **modelos inteligentes, suaves, adaptables y, fundamentalmente, interpretables**, que pueden enfrentarse a la complejidad de las series temporales no estacionarias y los sistemas caóticos que definen nuestro universo de datos.
