## 1. El papel de la lógica como lenguaje

El razonamiento humano es, por naturaleza, variable e inconsistente. Cada persona desarrolla una concepción del mundo y un marco de razonamiento que depende de las experiencias que va acumulando, el contexto en el que ocurren y la secuencia de esos eventos. Con esa herramienta imprecisa y altamente subjetiva, el ser humano pretende comprender y actuar sobre un mundo que es también incierto y cambiante. Para lograrlo —y para poder colaborar con otras personas— necesitábamos un mecanismo que permitiera determinar la validez de un juicio independientemente de quién lo emitiera.

Uno de los logros más importantes del pensamiento humano fue el desarrollo de la lógica clásica, que formalizó los mecanismos del razonamiento y los liberó de la subjetividad del hablante. Este esquema se apoya en tres principios fundamentales:

1. **Principio de Identidad:** Una cosa es lo que es ($A = A$).
2. **Principio de No Contradicción:** Algo no puede ser y no ser al mismo tiempo ($\neg(A \land \neg A)$).
3. **Principio del Tercero Excluido:** Una proposición es verdadera o es falsa; no existe una tercera opción ($A \lor \neg A$).

<p align="center">
  <img src="images/logica_clasica.png" width="800">
</p>


Estos tres principios son pilares del razonamiento matemático formal y, al mismo tiempo, bases epistemológicas del método científico.

El poder de la lógica clásica proviene de una simplificación radical: obliga a las *proposiciones sobre* la realidad a expresarse en términos bivaluados. La realidad es un flujo continuo de experiencias; la lógica binaria ofrece una regla de decisión que poda esa complejidad para permitirnos actuar. Decir "este hongo es venenoso (1) o no lo es (0)" es una simplificación vital que ignora grados de toxicidad en favor de la supervivencia.

Pero esa misma simplificación introduce una tensión profunda: la realidad es continua, ambigua y contextual, mientras que la lógica clásica es discreta, rígida y absoluta. Podemos entenderla como un instrumento perfectamente diseñado para la consistencia formal, pero que le queda demasiado ajustado a la naturaleza. Esa tensión es el punto de partida de la lógica difusa.


## 2. Los límites de la lógica clásica

Los límites de la lógica clásica han sido expuestas a través de las llamadas "*paradojas*". Estas narraciones breves no son curiosidades filosóficas marginales (aunque según se dice Séneca las consideraba una pérdida de tiempo): Las paradojas son señales estructurales que indican dónde el sistema de representación es insuficiente.

<p align="center">
  <img src="images/Escher_Hands.png" width="400">
</p>

Una paradoja aparece cuando la realidad no cabe dentro del sistema lógico que intentamos imponerle. Formalmente, muchas paradojas pueden expresarse como ecuaciones sin solución en lógica binaria:

$$x = \neg x$$

$$\text{No existe solución en } \{0, 1\}$$

Esto no significa que el problema esté mal planteado. Significa que el sistema lógico es insuficiente. 


### Paradojas clásicas

Las siguientes paradojas comparten una característica común: introducen autorreferencia, continuidad o vaguedad, y la lógica clásica responde con contradicción. La contradicción no está en la realidad; está en la representación.

**La Paradoja del Mentiroso**

Esta es la más antigua y directa de todas, atribuida a Eubúlides de Mileto (siglo IV a.C.). 

> Un cretense dice: *"Todos los cretenses mienten"*. Si la afirmación es verdadera, entonces él también miente, por lo que es falsa. Si es falsa, entonces él dice la verdad, por lo que es verdadera. La proposición colapsa en un bucle sin fondo.


**La Paradoja del Barbero (Russell, 1901)**

> En un pueblo hay un barbero que sigue una regla estricta: *afeita exactamente a quienes no se afeitan a sí mismos*.

La pregunta inevitable es: ¿entonces, quién afeita al barbero?

- Si se afeita a sí mismo, viola su regla.
- Si no se afeita, su regla lo obliga a hacerlo.

La proposición no puede ser 0 ni 1.


**La Paradoja de la Horca**

> Un marajá condena a un prisionero, pero le ofrece una última oportunidad: *"Di una sentencia. Si es verdadera, morirás en la horca. Si es falsa, morirás decapitado."* El prisionero, después de reflexionar, responde: *"Moriré decapitado."*

¿Cuál es la decisión correcta del marajá? 

- Si lo manda decapitar, la sentencia se comprueba como verdadera, y debió colgarlo (no decapitarlo).
- Si lo cuelga, la sentencia resulta falsa, pero entonces no debió ahorcarlo (debió decapitarlo).

El sistema binario es incapaz de procesar esta situación. No porque sea irracional, sino porque la lógica clásica no tiene categorías para la dependencia circular entre una afirmación y su consecuencia.

**La Paradoja de Sorites (el montón de arena)**

> Un millón de granos de arena forman un montón. Retirar un grano no destruye el montón. Aplicando ese razonamiento repetidamente debemos concluir que un solo grano es un montón.

El problema aquí es que la lógica clásica no tiene forma de trazar el límite porque el límite no existe de manera precisa. La paradoja de Sorites no es un bucle lógico: es una frontera borrosa, y eso es algo cualitativamente diferente.

> **Nota histórica:** También atribuida a Eubúlides, su motivación era epistemológica: demostrar que el lenguaje ordinario no puede capturar la realidad con precisión. Durante siglos fue tratada como un problema menor o un juego de palabras. Su verdadero peso solo se hizo evidente en el siglo XX, cuando Lotfi Zadeh reconoció en ella no un defecto del lenguaje, sino una característica del mundo. La paradoja de Sorites es, en cierto sentido, el punto de partida implícito de toda la lógica difusa.


### ¿Incertidumbre o vaguedad? El gato de Schrödinger como punto de partida

Las paradojas anteriores sugieren que hay al menos dos tipos distintos de problemas que la lógica clásica no puede manejar: la **autorreferencia** y la **vaguedad**. Otro problema que la lógica clásica no resuelve satisfactoriamente es la **incertidumbre**, un concepto que es importante distinguir del de *vaguedad*, antes de entrar en detalles de la Lógica difusa.

Podemos utilizar el experimento mental del gato de Schrödinger para ilustrar la diferencia con precisión:

>Un gato es encerrado en una caja con un dispositivo que incluye una fuente radiactiva, un detector y veneno. Si el átomo se desintegra (asumiendo una probabilidad de este evento del 50%), el veneno se libera y el gato muere; si no, vive.

- La probabilidad trata la situación del gato como incertidumbre: **el gato tiene un 50% de probabilidad de estar vivo**. La ignorancia es nuestra: al abrir la caja, el estado colapsará a 0 o 1. **La incertidumbre es epistemológica**.
  
- La lógica difusa, por otra parte, vería la situación del gato como "vaga": **mientras la caja está cerrada, el gato se encuentra simultáneamente en el conjunto de los vivos y el de los muertos, con un grado de pertenencia parcial en cada conjunto**. En este caso, no es que *no sepamos* cuál es su estado; es que su estado es intrínsecamente una combinación de ambos. **La vaguedad es ontológica**.

La vaguedad es una falta de definición, la incertidumbre es una falta de conocimiento.

</p>


Otra metáfora que se emplea para enfatizar esta diferencia es la siguiente: Imaginemos que estamos en el desierto y tenemos sed y nos ofrecen dos botellas de agua:

<ul>
    <li>La primera botella <em>tiene un 5% de probabilidad de que esté totalmente envenenada.</em> La presencia del veneno es binario: está o no está. En este caso tenemos incertidumbre acerca de si la botella pertenece al conjunto de botellas de agua potable o al conjunto de botellas de agua envenenada.</li><br>

<li>La segunda botella <em>está ligeramente envenenada</em> ($\mu=0.05$). En este caso, sabemos que la botella pertenece <em>un poco</em> al conjunto de botellas de agua envenenada. El veneno está diluido en toda el agua; sabemos exactamente qué hay adentro, pero la definición de <em>agua potable</em> se ha vuelto borrosa.

## 3. El camino hacia lo no-binario

Antes de la llegada de la lógica difusa en 1965, la humanidad ya sospechaba que el blanco o negro era insuficiente. Diferentes pensadores intentaron añadir matices al pensamiento formal para rescatar la complejidad de la realidad. El recorrido es más largo de lo que suele reconocerse, y empieza, con cierta ironía, en el mismo lugar donde comenzó la lógica clásica.

---

### A. Aristóteles y el futuro contingente: la primera grieta

Quien sistematizó los principios de la lógica clásica fue también el primero en encontrar sus límites. Aristóteles planteó el problema de las **proposiciones sobre el futuro**, los llamados *futuros contingentes*:

> *"Mañana habrá una batalla naval."*

- Si hoy decimos que es *verdadera*, la batalla es inevitable — caemos en el determinismo.
- Si decimos que es *falsa*, la batalla es imposible — negamos la contingencia.

Ninguna de las dos opciones es aceptable. La solución de Aristóteles fue sugerir que ciertas proposiciones aún no tienen un valor de verdad definido: su verdad depende del tiempo y del contexto. Con eso sembró una idea que tardaría siglos en formalizarse — que la verdad no tiene por qué ser una función estática de dos valores.

---

### B. Lógicas multivaluadas y modales: los primeros formalismos

En el siglo XX, el lógico polaco **Jan Łukasiewicz** retomó la intuición aristotélica y la formalizó. Su contribución fue doble:

1. **Lógica trivaluada:** Introdujo un tercer valor, $\frac{1}{2}$, para proposiciones indeterminadas, rompiendo por primera vez el monopolio de $\{0, 1\}$.
2. **Lógicas de $n$ valores:** Łukasiewicz desarrolló lógicas con cualquier número finito de valores de verdad, e incluso exploró versiones infinito-valuadas — un paso conceptual que anticipa directamente a Zadeh.

En paralelo, las **lógicas modales** ampliaron el vocabulario del razonamiento formal añadiendo operadores para la *necesidad* ($\Box$) y la *posibilidad* ($\Diamond$). No es lo mismo decir "llueve" que "es posible que llueva" — la distinción captura matices que la lógica clásica no puede expresar.

Sin embargo, todas estas lógicas seguían siendo discretas: añadían escalones al sistema, no un gradiente continuo.

---

### C. Lógica paraconsistente: convivir con la contradicción

Las lógicas multivaluadas atacaron el problema de la indeterminación. Pero las paradojas de autorreferencia — el Mentiroso, el Barbero, la Horca — plantean un problema distinto: no ausencia de valor de verdad, sino *contradicción activa*. En la lógica clásica, una contradicción es catastrófica: del principio de explosión (*ex contradictione sequitur quodlibet*) se sigue que de una premisa falsa puede derivarse cualquier conclusión, haciendo el sistema inútil.

La **lógica paraconsistente**, desarrollada formalmente por **Newton da Costa** en los años 60 y por **Graham Priest** con su sistema LP (*Logic of Paradox*), propone una solución radical: abandonar el principio de explosión. Un sistema paraconsistente puede contener contradicciones locales sin colapsar globalmente. La paradoja del Mentiroso deja de ser un error fatal y se convierte en una proposición que simplemente es *verdadera y falsa a la vez* — un estado que el sistema tolera sin propagarlo al resto del razonamiento.

Esto tiene consecuencias prácticas: bases de datos inconsistentes, sistemas de información con fuentes contradictorias, y razonamiento jurídico con normas en conflicto son dominios donde la lógica paraconsistente ofrece herramientas que la lógica clásica no puede proveer.

---

### D. Lógica difusa: el gradiente continuo

En 1965, **Lotfi A. Zadeh** dio el salto definitivo. En lugar de añadir uno o dos valores extra, propuso que el grado de pertenencia de un elemento a un conjunto puede ser **cualquier número real en el intervalo $[0, 1]$**:

$$\mu_A(x) \in [0, 1]$$

La pertenencia deja de ser una puerta — abierta o cerrada — y se convierte en una rampa. Un elemento puede pertenecer *parcialmente* a un conjunto con grado $\mu = 0.3$, sin que eso implique nada automático sobre su relación con el conjunto complementario. Esto permite que un sistema formal opere con conceptos lingüísticos como *"muy"*, *"poco"* o *"más o menos"* — los llamados **modificadores hedónicos** — que la lógica clásica simplemente no puede representar.

La paradoja de Sorites, que durante siglos pareció un juego de palabras sin solución, encuentra aquí su respuesta natural: no existe un límite preciso entre "montón" y "no montón" porque *montón* es un concepto difuso. La frontera no está ausente por falta de información — está ausente porque el concepto mismo no la tiene.

---

### E. Conjuntos ásperos (*Rough Sets*): la frontera de la ignorancia

Vale mencionar un desarrollo paralelo que complementa, sin reemplazar, a la lógica difusa. En 1982, **Zdzisław Pawlak** propuso los *Rough Sets* desde un ángulo distinto. Para entender la diferencia, conviene recuperar la distinción que establecimos con el gato de Schrödinger:

- La **lógica difusa** habita el lado *ontológico*: la vaguedad está en el concepto mismo. "Alto" no tiene una frontera precisa porque la altura es un continuo y el lenguaje no la corta en ningún punto privilegiado.
- Los **conjuntos ásperos** habitan el lado *epistemológico*: el concepto puede estar perfectamente bien definido, pero nuestra granularidad de datos no es suficiente para determinar si un objeto pertenece o no.

Cuando la información disponible no permite distinguir entre dos objetos, Pawlak propuso representar el conjunto mediante dos aproximaciones:

- **Aproximación inferior:** los elementos que *con certeza* pertenecen al conjunto.
- **Aproximación superior:** los elementos que *posiblemente* pertenecen.

La región entre ambas es la *zona de frontera* — la huella de nuestra ignorancia sobre los datos. En ciencia de datos, los conjuntos ásperos son especialmente útiles para reducción de atributos y manejo de datos inconsistentes en tablas de decisión.

---

### F. Una mirada de conjunto

Cada sistema que hemos visto ataca un tipo distinto de límite de la lógica clásica:

| Paradoja / Problema | Naturaleza | Respuesta |
| :--- | :--- | :--- |
| Futuro contingente | Indeterminación temporal | Łukasiewicz — introduce $\frac{1}{2}$ |
| Mentiroso / Horca / Barbero | Autorreferencia contradictoria | Lógica paraconsistente — tolera contradicciones locales |
| Sorites | Vaguedad ontológica del concepto | Lógica difusa — disuelve la frontera en un gradiente |
| Indiscernibilidad de datos | Imprecisión epistemológica | Conjuntos ásperos — aproximación inferior y superior |

La lógica clásica no está equivocada — opera con precisión dentro de sus supuestos. Lo que estas extensiones reconocen es que esos supuestos no cubren todo el territorio del razonamiento real. La lógica difusa, en particular, no corrige a Aristóteles: lo extiende hacia el dominio donde sus herramientas no alcanzan.

## 4. Conjuntos difusos: formalización

Un **conjunto difuso** $A$ sobre un universo de discurso $X$ se define mediante una **función de membresía**:

$$\mu_A : X \rightarrow [0, 1]$$

que asigna a cada elemento $x \in X$ un grado de pertenencia $\mu_A(x)$. Cuando $\mu_A(x) \in \{0, 1\}$ para todo $x$, recuperamos el conjunto clásico como caso especial.

### Conceptos fundamentales

- **Soporte:** $\text{supp}(A) = \{ x \in X \mid \mu_A(x) > 0 \}$
- **Núcleo:** $\text{core}(A) = \{ x \in X \mid \mu_A(x) = 1 \}$
- **Altura:** $h(A) = \sup_{x \in X} \mu_A(x)$
- **Corte-$\alpha$:** $A_\alpha = \{ x \in X \mid \mu_A(x) \geq \alpha \}$

Un conjunto difuso es **normal** si $h(A) = 1$, y **convexo** si para todo $x_1, x_2 \in X$ y $\lambda \in [0,1]$:

$$\mu_A(\lambda x_1 + (1-\lambda)x_2) \geq \min(\mu_A(x_1), \mu_A(x_2))$$

---

### Funciones de membresía

La elección de la función de membresía determina cómo se modela la transición entre pertenencia y no pertenencia. Las funciones más utilizadas son:

**1. Triangular** — transición lineal simétrica o asimétrica:

$$\mu(x; a, b, c) = \max\left(0,\ \min\left(\frac{x-a}{b-a},\ \frac{c-x}{c-b}\right)\right)$$

**2. Trapezoidal** — núcleo extendido con transiciones lineales:

$$\mu(x; a, b, c, d) = \max\left(0,\ \min\left(\frac{x-a}{b-a},\ 1,\ \frac{d-x}{d-c}\right)\right)$$

**3. Gaussiana** — transición suave, núcleo en $c$, dispersión $\sigma$:

$$\mu(x; c, \sigma) = e^{-\frac{(x-c)^2}{2\sigma^2}}$$

**4. Función S** — transición suave monotónica creciente:

$$\mu(x; a, b) = \begin{cases} 0 & x \leq a \\ 2\left(\dfrac{x-a}{b-a}\right)^2 & a < x \leq \dfrac{a+b}{2} \\ 1 - 2\left(\dfrac{x-b}{b-a}\right)^2 & \dfrac{a+b}{2} < x \leq b \\ 1 & x > b \end{cases}$$

**5. Función Z** — complemento de la función S, monotónica decreciente:

$$\mu(x; a, b) = 1 - S(x; a, b)$$

**6. Función Pi** — combinación de S y Z, forma de campana asimétrica:

$$\mu(x; a, b, c, d) = \begin{cases} S(x; a, b) & x \leq \dfrac{b+c}{2} \\ Z(x; c, d) & x > \dfrac{b+c}{2} \end{cases}$$

**7. Sigmoides como aproximación gaussiana** — la función sigmoide y su versión de diferencia (*bell sigmoide*) permiten aproximar formas gaussianas con control explícito sobre la pendiente:

$$\mu(x; a, c) = \frac{1}{1 + e^{-a(x-c)}}$$

Una campana simétrica se construye como diferencia de dos sigmoides:

$$\mu(x; a, c_1, c_2) = \sigma(x; a, c_1) - \sigma(x; a, c_2), \quad c_1 < c_2$$

A diferencia de la gaussiana, la sigmoide permite controlar la *asimetría* de la transición ajustando independientemente cada flanco.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ── Paleta del curso ────────────────────────────────────────────────────────
BG      = "#09090E"
CYAN    = "#00FFFF"
AMBER   = "#FFB300"
LAVENDER= "#B388FF"
GREEN   = "#69FF47"
PINK    = "#FF6EC7"
WHITE   = "#E8E8F0"
GRAY    = "#3A3A4A"

x = np.linspace(0, 10, 500)

# ── Funciones de membresía ───────────────────────────────────────────────────

def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

def trapezoidal(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x - a) / (b - a), 1), (d - x) / (d - c)))

def gaussiana(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))

def funcion_S(x, a, b):
    m = (a + b) / 2
    y = np.where(x <= a, 0,
        np.where(x <= m, 2 * ((x - a) / (b - a)) ** 2,
        np.where(x <= b, 1 - 2 * ((x - b) / (b - a)) ** 2, 1)))
    return y.astype(float)

def funcion_Z(x, a, b):
    return 1 - funcion_S(x, a, b)

def funcion_Pi(x, a, b, c, d):
    mid = (b + c) / 2
    return np.where(x <= mid, funcion_S(x, a, b), funcion_Z(x, c, d))

def sigmoide(x, a, c):
    return 1 / (1 + np.exp(-a * (x - c)))

def bell_sigmoide(x, a, c1, c2):
    return np.clip(sigmoide(x, a, c1) - sigmoide(x, a, c2), 0, 1)

# ── Figura ───────────────────────────────────────────────────────────────────

fig = plt.figure(figsize=(16, 14), facecolor=BG)
gs  = gridspec.GridSpec(3, 3, figure=fig, hspace=0.55, wspace=0.35)

plots = [
    ("Triangular",        triangular(x, 2, 5, 8),          CYAN),
    ("Trapezoidal",       trapezoidal(x, 1, 3, 7, 9),       AMBER),
    ("Gaussiana",         gaussiana(x, 5, 1),               LAVENDER),
    ("Función S",         funcion_S(x, 2, 7),               GREEN),
    ("Función Z",         funcion_Z(x, 3, 8),               PINK),
    ("Función Pi",        funcion_Pi(x, 1, 3, 7, 9),        CYAN),
    ("Sigmoide",          sigmoide(x, 2, 5),                AMBER),
    ("Bell sigmoide\n(aprox. gaussiana)", bell_sigmoide(x, 2, 3, 7), LAVENDER),
]

axes = [fig.add_subplot(gs[i // 3, i % 3]) for i in range(len(plots))]

for ax, (title, y, color) in zip(axes, plots):
    ax.set_facecolor(BG)
    ax.plot(x, y, color=color, linewidth=2.2)
    ax.fill_between(x, y, alpha=0.08, color=color)
    ax.set_title(title, color=WHITE, fontsize=10, fontfamily="monospace", pad=6)
    ax.set_xlim(0, 10)
    ax.set_ylim(-0.05, 1.15)
    ax.set_xlabel("$x$", color=GRAY, fontsize=9)
    ax.set_ylabel("$\\mu(x)$", color=GRAY, fontsize=9)
    ax.tick_params(colors=GRAY, labelsize=8)
    for spine in ax.spines.values():
        spine.set_edgecolor(GRAY)
    ax.axhline(1.0, color=GRAY, linewidth=0.5, linestyle="--", alpha=0.4)
    ax.axhline(0.0, color=GRAY, linewidth=0.5, linestyle="--", alpha=0.4)

# Celda vacía (3x3 con 8 plots)
fig.delaxes(fig.add_subplot(gs[2, 2]))

fig.suptitle("Funciones de membresía — Conjuntos Difusos",
             color=WHITE, fontsize=13, fontfamily="monospace", y=0.98)

plt.savefig("membresia.png", dpi=150, bbox_inches="tight", facecolor=BG)
plt.show()
```

### Elección de la función de membresía

La función de membresía no es un detalle técnico menor: es la hipótesis central del modelo difuso. Elegirla bien significa entender qué tipo de transición describe el fenómeno que queremos representar, con qué precisión conocemos esa transición, y qué costo computacional podemos asumir.

---

### Triangular

**Forma:** rampa de subida lineal, pico en $b$, rampa de bajada lineal.

**Ventajas:**
- Interpretable: los tres parámetros $(a, b, c)$ tienen significado directo — inicio, centro y fin de la zona de transición.
- Computacionalmente trivial: solo operaciones lineales.
- Fácil de ajustar manualmente por un experto de dominio.

**Desventajas:**
- El pico es una esquina — la derivada es discontinua en $b$, lo que puede generar problemas en sistemas que requieren gradientes suaves.
- No modela bien fenómenos con transiciones asimétricas complejas.

**Usos típicos:** sistemas de control difuso clásicos, prototipado rápido, representación de variables lingüísticas simples (*"temperatura media"*, *"velocidad moderada"*).

---

### Trapezoidal

**Forma:** rampa de subida, meseta central entre $b$ y $c$, rampa de bajada.

**Ventajas:**
- Permite modelar un **núcleo extendido**: una zona donde el elemento pertenece completamente al conjunto, no solo un punto.
- Más robusta que la triangular ante pequeñas variaciones del valor de entrada dentro de la meseta.

**Desventajas:**
- Cuatro parámetros hacen el ajuste manual algo más laborioso.
- Las esquinas en $b$ y $c$ mantienen el problema de discontinuidad en la derivada.

**Usos típicos:** representación de rangos aceptables en control de calidad, variables con zona de confort bien definida (*"temperatura confortable entre 20°C y 24°C"*).

---

### Gaussiana

**Forma:** campana simétrica suave centrada en $c$, dispersión controlada por $\sigma$.

**Ventajas:**
- Infinitamente diferenciable — ideal para sistemas que requieren gradientes.
- Conexión natural con distribuciones de probabilidad y modelos estadísticos.
- Transición suave y continua en todo su dominio.

**Desventajas:**
- Soporte técnicamente infinito: nunca llega exactamente a cero, lo que puede generar activaciones residuales no deseadas.
- Siempre simétrica — no puede modelar transiciones asimétricas sin modificaciones.
- Más costosa computacionalmente que las funciones lineales por tramos.

**Usos típicos:** sistemas neuro-difusos (ANFIS), modelado de variables continuas con comportamiento unimodal, cualquier aplicación donde la suavidad del gradiente es prioritaria.

---

### Función S

**Forma:** transición suave monotónica creciente de 0 a 1.

**Ventajas:**
- Modela conceptos abiertos por la derecha: *"muy alto"*, *"muy rápido"*, *"suficientemente caliente"*.
- Transición suave sin las esquinas de la trapezoidal.
- Solo dos parámetros: punto de inicio y punto de saturación.

**Desventajas:**
- No tiene núcleo acotado — una vez que alcanza 1, permanece ahí indefinidamente.
- No sirve para modelar conceptos con zona central de pertenencia máxima y decaimiento posterior.

**Usos típicos:** umbrales de activación, representación de extremos en escalas lingüísticas (*"joven"* en el extremo inferior, *"anciano"* en el superior), preprocesamiento de señales.

---

### Función Z

**Forma:** complemento exacto de la función S — transición suave monotónica decreciente.

**Ventajas y desventajas:** simétricas a las de la función S, con el comportamiento invertido.

**Usos típicos:** mismos contextos que la función S, pero para el extremo opuesto de la escala lingüística. En la práctica, S y Z suelen aparecer juntas como los flancos de una partición difusa.

---

### Función Pi

**Forma:** campana construida como concatenación de una función S y una función Z.

**Ventajas:**
- Núcleo extendido con transiciones suaves — combina lo mejor de la trapezoidal y la gaussiana.
- Control independiente sobre los flancos de subida y bajada mediante cuatro parámetros.
- Puede ser asimétrica, a diferencia de la gaussiana.

**Desventajas:**
- Cuatro parámetros requieren más cuidado en el ajuste.
- Menos común en la literatura, lo que reduce la disponibilidad de implementaciones de referencia.

**Usos típicos:** modelado de variables lingüísticas centrales con transiciones suaves (*"temperatura media"*, *"presión normal"*), aplicaciones donde la simetría de la gaussiana es demasiado restrictiva.

---

### Sigmoide

**Forma:** curva en S suave con pendiente controlada por $a$ y punto de inflexión en $c$.

**Ventajas:**
- Diferenciable en todo su dominio — compatible con optimización por gradiente.
- Ampliamente disponible en frameworks de aprendizaje automático.
- El parámetro $a$ controla directamente la *sharpness* de la transición: valores grandes aproximan una función escalón, valores pequeños producen transiciones muy suaves.

**Desventajas:**
- Monotónica — no modela conceptos con pertenencia máxima en una zona interior.
- El soporte es técnicamente infinito en ambas direcciones.

**Usos típicos:** sistemas neuro-difusos donde los parámetros se aprenden por retropropagación, activaciones en redes difusas, modelado de umbrales graduales.

---

### Bell sigmoide (aproximación gaussiana)

**Forma:** diferencia de dos sigmoides — campana asimétrica controlable.

**Ventajas:**
- Hereda la diferenciabilidad de la sigmoide — apta para gradiente.
- Permite **asimetría** en los flancos ajustando independientemente $c_1$ y $c_2$.
- Puede aproximar una gaussiana cuando $c_1$ y $c_2$ son simétricamente equidistantes del centro.
- Útil cuando se quiere una forma gaussiana pero con transiciones más abruptas en uno de los flancos.

**Desventajas:**
- Menos interpretable que la gaussiana — los parámetros $c_1$ y $c_2$ no tienen el mismo significado intuitivo que $c$ y $\sigma$.
- La normalización a $[0,1]$ requiere cuidado cuando $a$ es pequeño.

**Usos típicos:** sistemas neuro-difusos con variables asimétricas, modelado de respuestas sensoriales o perceptuales donde el umbral de entrada y el de saturación no son equidistantes del centro.

---

### Guía de selección rápida

| Criterio | Función recomendada |
| :--- | :--- |
| Máxima interpretabilidad | Triangular, Trapezoidal |
| Núcleo puntual, gradiente suave | Gaussiana |
| Núcleo extendido, gradiente suave | Pi |
| Concepto abierto (extremo de escala) | S / Z |
| Compatible con aprendizaje por gradiente | Gaussiana, Sigmoide, Bell sigmoide |
| Asimetría en los flancos | Pi, Bell sigmoide |
| Prototipado rápido | Triangular |
| Recursos computacionales limitados | Triangular, Trapezoidal, S / Z |

## 6. Operaciones sobre conjuntos difusos

Las operaciones sobre conjuntos difusos generalizan las operaciones clásicas de la teoría de conjuntos. El principio rector es el mismo que guió a Zadeh en 1965: cuando los grados de membresía se restringen a $\{0,1\}$, todas las operaciones difusas deben colapsar exactamente a sus equivalentes clásicos.

---

### A. Operaciones estándar de Zadeh

Sean $A$ y $B$ dos conjuntos difusos sobre $X$. Las operaciones estándar se definen puntualmente:

**Complemento:**
$$\mu_{\neg A}(x) = 1 - \mu_A(x)$$

**Unión:**
$$\mu_{A \cup B}(x) = \max(\mu_A(x),\ \mu_B(x))$$

**Intersección:**
$$\mu_{A \cap B}(x) = \min(\mu_A(x),\ \mu_B(x))$$

Estas tres operaciones satisfacen la mayoría de las propiedades de la lógica clásica — conmutatividad, asociatividad, distributividad, leyes de De Morgan — con una excepción notable:

$$\mu_{A \cup \neg A}(x) \neq 1 \qquad \mu_{A \cap \neg A}(x) \neq 0$$

El **principio del tercero excluido** y el **principio de no contradicción** no se sostienen en general para conjuntos difusos. Esto no es un defecto — es precisamente lo que permite modelar la vaguedad.

---

### B. T-normas y T-conormas: generalizando la intersección y la unión

Las operaciones de Zadeh son solo una elección posible. Una **T-norma** $T: [0,1]^2 \rightarrow [0,1]$ es cualquier función que generaliza la intersección satisfaciendo:

1. **Conmutatividad:** $T(a, b) = T(b, a)$
2. **Asociatividad:** $T(a, T(b, c)) = T(T(a, b), c)$
3. **Monotonía:** $b \leq c \Rightarrow T(a, b) \leq T(a, c)$
4. **Condición de frontera:** $T(a, 1) = a$

Las **T-conormas** (o S-normas) generalizan la unión de forma análoga, con condición de frontera $S(a, 0) = a$.

Las T-normas más utilizadas son:

| Nombre | $T(a, b)$ | Interpretación |
| :--- | :--- | :--- |
| Mínimo (Zadeh) | $\min(a, b)$ | Conservadora — solo lo que ambos comparten |
| Producto algebraico | $a \cdot b$ | Penaliza fuertemente membresías bajas |
| Łukasiewicz | $\max(0,\ a + b - 1)$ | La más restrictiva — exige alta membresía en ambos |
| Drástica | $T_D(a,b)$ | Caso límite — solo activa si uno de los dos vale 1 |

Y las T-conormas correspondientes:

| Nombre | $S(a, b)$ | Interpretación |
| :--- | :--- | :--- |
| Máximo (Zadeh) | $\max(a, b)$ | El más permisivo |
| Suma probabilística | $a + b - a \cdot b$ | Análoga a la probabilidad de unión |
| Łukasiewicz | $\min(1,\ a + b)$ | Saturación en 1 |
| Drástica | $S_D(a,b)$ | Solo activa si uno de los dos vale 0 |

La elección de T-norma y T-conorma define el **comportamiento lógico** del sistema difuso. En aplicaciones de control, el producto algebraico suaviza las transiciones; en clasificación, el mínimo de Zadeh es más interpretable.

---

### C. Modificadores hedónicos

Los **modificadores hedónicos** (o *hedges*) son operadores que transforman un conjunto difuso para representar matices lingüísticos. Se aplican directamente sobre la función de membresía:

| Modificador | Operación | Efecto |
| :--- | :--- | :--- |
| *very* (muy) | $\mu^2(x)$ | Concentra — reduce membresías intermedias |
| *slightly* (ligeramente) | $\mu^{0.5}(x)$ | Dilata — amplía la zona de transición |
| *more or less* (más o menos) | $\mu^{1/3}(x)$ | Dilata más agresivamente |
| *extremely* (extremadamente) | $\mu^3(x)$ | Concentra más agresivamente |
| *not* (no) | $1 - \mu(x)$ | Complemento estándar |

En general, un modificador hedónico de potencia $p$ actúa como:

$$\mu_{\text{hedge}}(x) = [\mu_A(x)]^p$$

- $p > 1$ **concentra** el conjunto — exige mayor grado de membresía para pertenecer.
- $p < 1$ **dilata** el conjunto — acepta membresías más bajas.

Esto permite construir automáticamente la representación de *"muy caliente"* a partir de *"caliente"*, sin necesidad de definir una nueva función de membresía desde cero.

---

### D. Principio de extensión

El **principio de extensión de Zadeh** permite generalizar cualquier función clásica $f: X \rightarrow Y$ para que opere sobre conjuntos difusos en lugar de elementos puntuales.

Dada $f: X \rightarrow Y$ y un conjunto difuso $A$ sobre $X$, el conjunto difuso imagen $f(A)$ sobre $Y$ se define como:

$$\mu_{f(A)}(y) = \sup_{x \in X,\ f(x) = y} \mu_A(x)$$

Es decir: el grado de membresía de $y$ en la imagen es el máximo grado de membresía entre todos los $x$ que mapean a $y$.

**¿Por qué importa?** Porque permite propagar incertidumbre difusa a través de modelos matemáticos. Si la entrada de un sistema es un conjunto difuso — por ejemplo, *"temperatura aproximadamente 30°C"* — el principio de extensión define cómo calcular la salida difusa correspondiente. Es la base teórica de la **aritmética difusa** y de los sistemas de inferencia difusa que veremos en la siguiente sección.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ── Paleta del curso ─────────────────────────────────────────────────────────
BG       = "#09090E"
CYAN     = "#00FFFF"
AMBER    = "#FFB300"
LAVENDER = "#B388FF"
GREEN    = "#69FF47"
PINK     = "#FF6EC7"
WHITE    = "#E8E8F0"
GRAY     = "#3A3A4A"

x = np.linspace(0, 10, 500)

# ── Conjuntos base ────────────────────────────────────────────────────────────
def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

A = triangular(x, 1, 4, 7)   # "temperatura media"
B = triangular(x, 4, 7, 10)  # "temperatura alta"

# ── Operaciones estándar ──────────────────────────────────────────────────────
complement_A   = 1 - A
union_AB       = np.maximum(A, B)
intersect_AB   = np.minimum(A, B)

# ── T-normas ──────────────────────────────────────────────────────────────────
tnorm_min      = np.minimum(A, B)
tnorm_prod     = A * B
tnorm_luka     = np.maximum(0, A + B - 1)

# ── T-conormas ────────────────────────────────────────────────────────────────
tconorm_max    = np.maximum(A, B)
tconorm_prob   = A + B - A * B
tconorm_luka   = np.minimum(1, A + B)

# ── Modificadores hedónicos ───────────────────────────────────────────────────
very_A         = A ** 2
slightly_A     = A ** 0.5
extremely_A    = A ** 3
more_or_less_A = A ** (1/3)

# ── Principio de extensión: f(x) = x² aplicada a conjunto difuso ──────────────
y_range = np.linspace(0, 100, 1000)

def extension_principle(x, mu_A, f, y_range):
    """Para cada y en y_range, busca el sup de mu_A(x) donde f(x) ≈ y."""
    mu_fA = np.zeros_like(y_range)
    f_vals = f(x)
    for i, y in enumerate(y_range):
        mask = np.abs(f_vals - y) < 0.5
        mu_fA[i] = np.max(mu_A[mask]) if np.any(mask) else 0.0
    return mu_fA

mu_ext = extension_principle(x, A, lambda x: x**2, y_range)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(18, 20), facecolor=BG)
gs  = gridspec.GridSpec(4, 3, figure=fig, hspace=0.55, wspace=0.35)

def styled_ax(ax, title):
    ax.set_facecolor(BG)
    ax.set_title(title, color=WHITE, fontsize=9.5, fontfamily="monospace", pad=6)
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(-0.05, 1.15)
    ax.set_xlabel("$x$", color=GRAY, fontsize=8)
    ax.set_ylabel("$\\mu(x)$", color=GRAY, fontsize=8)
    ax.tick_params(colors=GRAY, labelsize=7.5)
    for spine in ax.spines.values():
        spine.set_edgecolor(GRAY)
    ax.axhline(1.0, color=GRAY, linewidth=0.4, linestyle="--", alpha=0.3)
    ax.axhline(0.0, color=GRAY, linewidth=0.4, linestyle="--", alpha=0.3)
    return ax

def plot_base(ax):
    ax.plot(x, A, color=CYAN,  linewidth=1.4, linestyle="--", alpha=0.5, label="$A$")
    ax.plot(x, B, color=AMBER, linewidth=1.4, linestyle="--", alpha=0.5, label="$B$")

# Fila 0 — operaciones estándar
panels_std = [
    (gs[0,0], "Complemento  $\\neg A$",   [(x, complement_A, LAVENDER, "$1-\\mu_A$")]),
    (gs[0,1], "Unión  $A \\cup B$",        [(x, union_AB,     GREEN,    "$\\max(A,B)$")]),
    (gs[0,2], "Intersección  $A \\cap B$", [(x, intersect_AB, PINK,     "$\\min(A,B)$")]),
]
for spec, title, curves in panels_std:
    ax = styled_ax(fig.add_subplot(spec), title)
    plot_base(ax)
    for xd, yd, col, lbl in curves:
        ax.plot(xd, yd, color=col, linewidth=2.2, label=lbl)
        ax.fill_between(xd, yd, alpha=0.08, color=col)
    ax.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 1 — T-normas
panels_tnorm = [
    (gs[1,0], "T-norma: Mínimo",          tnorm_min,  CYAN),
    (gs[1,1], "T-norma: Producto",         tnorm_prod, AMBER),
    (gs[1,2], "T-norma: Łukasiewicz",      tnorm_luka, LAVENDER),
]
for spec, title, y_op, col in panels_tnorm:
    ax = styled_ax(fig.add_subplot(spec), title)
    plot_base(ax)
    ax.plot(x, y_op, color=col, linewidth=2.2)
    ax.fill_between(x, y_op, alpha=0.08, color=col)

# Fila 2 — T-conormas
panels_tconorm = [
    (gs[2,0], "T-conorma: Máximo",         tconorm_max,  CYAN),
    (gs[2,1], "T-conorma: Suma prob.",      tconorm_prob, AMBER),
    (gs[2,2], "T-conorma: Łukasiewicz",     tconorm_luka, LAVENDER),
]
for spec, title, y_op, col in panels_tconorm:
    ax = styled_ax(fig.add_subplot(spec), title)
    plot_base(ax)
    ax.plot(x, y_op, color=col, linewidth=2.2)
    ax.fill_between(x, y_op, alpha=0.08, color=col)

# Fila 3 — Hedges + Principio de extensión
ax_hedge = styled_ax(fig.add_subplot(gs[3, 0:2]), "Modificadores hedónicos sobre $A$")
ax_hedge.plot(x, A,             color=WHITE,    linewidth=1.6, linestyle="--", label="$A$ (base)",            alpha=0.5)
ax_hedge.plot(x, very_A,        color=CYAN,     linewidth=2.0, label="$very\\ A\\ (p=2)$")
ax_hedge.plot(x, slightly_A,    color=AMBER,    linewidth=2.0, label="$slightly\\ A\\ (p=0.5)$")
ax_hedge.plot(x, extremely_A,   color=PINK,     linewidth=2.0, label="$extremely\\ A\\ (p=3)$")
ax_hedge.plot(x, more_or_less_A,color=GREEN,    linewidth=2.0, label="$more\\ or\\ less\\ A\\ (p=1/3)$")
ax_hedge.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3, loc="upper right")

ax_ext = fig.add_subplot(gs[3, 2])
ax_ext.set_facecolor(BG)
ax_ext.set_title("Principio de extensión\n$f(x)=x^2$", color=WHITE, fontsize=9.5,
                  fontfamily="monospace", pad=6)
ax_ext.plot(y_range, mu_ext, color=LAVENDER, linewidth=2.2)
ax_ext.fill_between(y_range, mu_ext, alpha=0.08, color=LAVENDER)
ax_ext.set_xlim(0, 100)
ax_ext.set_ylim(-0.05, 1.15)
ax_ext.set_xlabel("$y = x^2$", color=GRAY, fontsize=8)
ax_ext.set_ylabel("$\\mu_{f(A)}(y)$", color=GRAY, fontsize=8)
ax_ext.tick_params(colors=GRAY, labelsize=7.5)
for spine in ax_ext.spines.values():
    spine.set_edgecolor(GRAY)

fig.suptitle("Operaciones sobre conjuntos difusos",
             color=WHITE, fontsize=13, fontfamily="monospace", y=0.995)

plt.savefig("operaciones_difusas.png", dpi=150, bbox_inches="tight", facecolor=BG)
plt.show()
```

## 7. Sistemas de inferencia difusa

Un **sistema de inferencia difusa** (FIS, *Fuzzy Inference System*) es un marco computacional que traduce conocimiento expreso en lenguaje natural — *"si la epidemia está en fase de crecimiento rápido, entonces la tasa de contagio estimada es alta"* — en una función matemática bien definida que mapea entradas numéricas a salidas numéricas.

El pipeline general tiene cuatro etapas:

$$\text{Entrada numérica} \xrightarrow{\text{Fuzzificación}} \text{Grados de membresía} \xrightarrow{\text{Inferencia}} \text{Conjuntos difusos de salida} \xrightarrow{\text{Defuzzificación}} \text{Salida numérica}$$

Los tres sistemas clásicos — Mamdani, Takagi-Sugeno y Tsukamoto — comparten este pipeline pero difieren en cómo representan la salida de cada regla.

---

### Contexto del ejemplo

A lo largo de esta sección usaremos como hilo conductor la **calibración del modelo SIR**. El modelo SIR describe la dinámica de una epidemia mediante tres compartimentos:

$$\frac{dS}{dt} = -\beta \frac{SI}{N}, \qquad \frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I, \qquad \frac{dR}{dt} = \gamma I$$

donde $\beta$ es la tasa de contagio y $\gamma$ la tasa de recuperación. El número reproductivo básico $R_0 = \beta / \gamma$ determina la dinámica global:

- $R_0 > 1$: la epidemia crece.
- $R_0 = 1$: la epidemia se estabiliza en el pico.
- $R_0 < 1$: la epidemia decrece.

Nuestro sistema difuso tomará como entrada observable la **tasa de cambio de infectados** $\Delta I = I(t) - I(t-1)$ y el **nivel actual de infectados** $I(t)$, ambos normalizados respecto a la población $N$, y producirá como salida una estimación difusa de $\beta$ — el parámetro más sensible y difícil de medir directamente.

---

### A. Sistema de Mamdani

El sistema de Mamdani (1975) es el más interpretable: tanto las entradas como las salidas son conjuntos difusos, y las reglas tienen la forma natural:

$$\text{SI } x_1 \text{ es } A_i \text{ Y } x_2 \text{ es } B_i \text{ ENTONCES } y \text{ es } C_i$$

**Etapa 1 — Fuzzificación**

Cada entrada numérica se convierte en grados de membresía respecto a los conjuntos difusos definidos para esa variable. Para $\Delta I / N$ definimos tres conjuntos: *decreciendo*, *estable*, *creciendo*. Para $I/N$: *bajo*, *medio*, *alto*.

**Etapa 2 — Base de reglas**

El experto codifica su conocimiento en reglas linguísticas. Ejemplo:

| # | Si $\Delta I/N$ es... | Y $I/N$ es... | Entonces $\beta$ es... |
|:--|:--|:--|:--|
| R1 | creciendo | alto | muy\_alto |
| R2 | creciendo | medio | alto |
| R3 | estable | medio | medio |
| R4 | decreciendo | medio | bajo |
| R5 | decreciendo | bajo | muy\_bajo |

**Etapa 3 — Motor de inferencia**

Para cada regla $i$, el grado de activación se calcula con una T-norma (usualmente mínimo):

$$\alpha_i = T(\mu_{A_i}(\Delta I/N),\ \mu_{B_i}(I/N))$$

La salida de cada regla es el conjunto difuso $C_i$ *recortado* al nivel $\alpha_i$:

$$\mu_{C_i'}(y) = \min(\alpha_i,\ \mu_{C_i}(y))$$

El conjunto difuso de salida agregado es la unión de todas las salidas recortadas:

$$\mu_{\text{out}}(y) = \max_i \mu_{C_i'}(y)$$

**Etapa 4 — Defuzzificación**

El conjunto difuso agregado debe convertirse en un valor numérico concreto. Los dos métodos más comunes son:

*Centroide (COG — Center of Gravity):*
$$y^* = \frac{\int y \cdot \mu_{\text{out}}(y)\, dy}{\int \mu_{\text{out}}(y)\, dy}$$

*Media de máximos (MOM — Mean of Maxima):*
$$y^* = \frac{1}{|\mathcal{M}|} \int_{\mathcal{M}} y\, dy, \qquad \mathcal{M} = \{y \mid \mu_{\text{out}}(y) = \max \mu_{\text{out}}\}$$

El centroide pondera toda la distribución de salida y produce resultados más suaves. La media de máximos ignora la forma del conjunto y solo atiende a la zona de mayor activación — más rápida pero menos sensible.

---

### B. Sistema de Takagi-Sugeno (TSK)

El sistema TSK (Takagi y Sugeno, 1985) reemplaza los conjuntos difusos de salida por **funciones de las entradas**. Las reglas tienen la forma:

$$\text{SI } x_1 \text{ es } A_i \text{ Y } x_2 \text{ es } B_i \text{ ENTONCES } y_i = f_i(x_1, x_2)$$

En la versión de **orden cero**, $f_i$ es una constante. En la versión de **primer orden**:

$$y_i = p_i x_1 + q_i x_2 + r_i$$

La salida final es un promedio ponderado por los grados de activación:

$$y^* = \frac{\sum_i \alpha_i \cdot y_i}{\sum_i \alpha_i}$$

**Ventajas sobre Mamdani:**
- No requiere defuzzificación — la salida es siempre numérica.
- Computacionalmente más eficiente.
- Compatible con optimización por gradiente — los parámetros $p_i, q_i, r_i$ pueden aprenderse (ANFIS).

**Limitación:** las reglas son menos interpretables — el consecuente es una ecuación, no un término lingüístico.

En el contexto SIR, cada regla produce una estimación lineal de $\beta$ como función de $\Delta I/N$ e $I/N$. La salida final es la combinación ponderada de esas estimaciones locales.

---

### C. Sistema de Tsukamoto

El sistema de Tsukamoto es una variante de Mamdani donde los conjuntos difusos de salida son **monotónicos** — funciones S o Z exclusivamente. Esto permite invertir la función de membresía y obtener directamente un valor numérico $y_i$ para cada regla:

$$\alpha_i = \mu_{C_i}(y_i) \implies y_i = \mu_{C_i}^{-1}(\alpha_i)$$

La salida final es el promedio ponderado:

$$y^* = \frac{\sum_i \alpha_i \cdot y_i}{\sum_i \alpha_i}$$

**Posición en el espacio de sistemas difusos:**

| Sistema | Consecuente | Defuzzificación | Interpretabilidad | Eficiencia |
|:--|:--|:--|:--|:--|
| Mamdani | Conjunto difuso | COG / MOM | Alta | Baja |
| Tsukamoto | Conjunto monotónico invertible | Promedio ponderado | Media | Media |
| TSK orden 0 | Constante | Promedio ponderado | Media | Alta |
| TSK orden 1 | Función lineal | Promedio ponderado | Baja | Alta |

Tsukamoto ocupa un lugar intermedio: mantiene conjuntos difusos en el consecuente — lo que preserva parte de la interpretabilidad — pero evita la defuzzificación costosa de Mamdani al requerir monotonía.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

BG       = "#09090E"
CYAN     = "#00FFFF"
AMBER    = "#FFB300"
LAVENDER = "#B388FF"
GREEN    = "#69FF47"
PINK     = "#FF6EC7"
WHITE    = "#E8E8F0"
GRAY     = "#3A3A4A"

# ── Universos de discurso ─────────────────────────────────────────────────────
x_dI  = np.linspace(-0.05, 0.05, 500)   # ΔI/N  — tasa de cambio de infectados
x_I   = np.linspace(0, 1, 500)          # I/N   — nivel de infectados
x_b   = np.linspace(0, 1, 500)          # β     — tasa de contagio estimada

# ── Funciones auxiliares ──────────────────────────────────────────────────────
def trimf(x, a, b, c):
    return np.maximum(0, np.minimum((x-a)/(b-a+1e-10), (c-x)/(c-b+1e-10)))

def trapmf(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x-a)/(b-a+1e-10), 1), (d-x)/(d-c+1e-10)))

def S(x, a, b):
    m = (a+b)/2
    return np.where(x<=a, 0, np.where(x<=m, 2*((x-a)/(b-a))**2,
           np.where(x<=b, 1-2*((x-b)/(b-a))**2, 1))).astype(float)

def Z(x, a, b):
    return 1 - S(x, a, b)

# ── Membresías de entrada ─────────────────────────────────────────────────────
# ΔI/N
mf_dI = {
    "decreciendo": Z(x_dI, -0.02, 0.0),
    "estable":     trimf(x_dI, -0.01, 0.0, 0.01),
    "creciendo":   S(x_dI, 0.0, 0.02),
}
# I/N
mf_I = {
    "bajo":  Z(x_I, 0.1, 0.35),
    "medio": trimf(x_I, 0.2, 0.5, 0.8),
    "alto":  S(x_I, 0.55, 0.85),
}
# β (salida Mamdani)
mf_b = {
    "muy_bajo": Z(x_b, 0.0, 0.25),
    "bajo":     trimf(x_b, 0.1, 0.25, 0.45),
    "medio":    trimf(x_b, 0.35, 0.5, 0.65),
    "alto":     trimf(x_b, 0.55, 0.7, 0.85),
    "muy_alto": S(x_b, 0.7, 0.95),
}

# ── Punto de observación ──────────────────────────────────────────────────────
obs_dI = 0.018   # epidemia creciendo rápido
obs_I  = 0.62    # nivel de infectados alto

# ── Fuzzificación ─────────────────────────────────────────────────────────────
def fuzzify(val, x_univ, mfs):
    return {k: float(np.interp(val, x_univ, v)) for k, v in mfs.items()}

deg_dI = fuzzify(obs_dI, x_dI, mf_dI)
deg_I  = fuzzify(obs_I,  x_I,  mf_I)

# ── Base de reglas: (dI_label, I_label) → beta_label ─────────────────────────
rules = [
    ("creciendo",   "alto",  "muy_alto"),
    ("creciendo",   "medio", "alto"),
    ("estable",     "medio", "medio"),
    ("decreciendo", "medio", "bajo"),
    ("decreciendo", "bajo",  "muy_bajo"),
]

# ── Inferencia (min para T-norma, max para agregación) ───────────────────────
agg = np.zeros_like(x_b)
activations = []

for dI_lbl, I_lbl, b_lbl in rules:
    alpha = min(deg_dI[dI_lbl], deg_I[I_lbl])
    clipped = np.minimum(alpha, mf_b[b_lbl])
    agg = np.maximum(agg, clipped)
    activations.append((dI_lbl, I_lbl, b_lbl, alpha, clipped))

# ── Defuzzificación ───────────────────────────────────────────────────────────
# COG
cog = np.sum(x_b * agg) / (np.sum(agg) + 1e-10)

# MOM
max_val = np.max(agg)
mom_mask = agg >= max_val - 1e-4
mom = np.mean(x_b[mom_mask])

# ── Figura ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(18, 14), facecolor=BG)
gs  = gridspec.GridSpec(3, 3, figure=fig, hspace=0.6, wspace=0.35)

def styled(ax, title, xlabel="", ylabel="$\\mu$"):
    ax.set_facecolor(BG)
    ax.set_title(title, color=WHITE, fontsize=9, fontfamily="monospace", pad=5)
    ax.set_xlabel(xlabel, color=GRAY, fontsize=8)
    ax.set_ylabel(ylabel, color=GRAY, fontsize=8)
    ax.tick_params(colors=GRAY, labelsize=7.5)
    for sp in ax.spines.values(): sp.set_edgecolor(GRAY)
    ax.axhline(0, color=GRAY, lw=0.4, ls="--", alpha=0.3)
    ax.axhline(1, color=GRAY, lw=0.4, ls="--", alpha=0.3)
    return ax

colors_in = [CYAN, AMBER, LAVENDER]

# Fila 0 — membresías ΔI/N
ax0 = styled(fig.add_subplot(gs[0, 0:2]), "Membresías de entrada — $\\Delta I/N$", "$\\Delta I/N$")
for (lbl, y), col in zip(mf_dI.items(), colors_in):
    ax0.plot(x_dI, y, color=col, lw=2, label=lbl)
    ax0.fill_between(x_dI, y, alpha=0.07, color=col)
ax0.axvline(obs_dI, color=WHITE, lw=1.2, ls=":", alpha=0.7, label=f"obs={obs_dI}")
ax0.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 0 — membresías I/N
ax1 = styled(fig.add_subplot(gs[0, 2]), "Membresías de entrada — $I/N$", "$I/N$")
for (lbl, y), col in zip(mf_I.items(), colors_in):
    ax1.plot(x_I, y, color=col, lw=2, label=lbl)
    ax1.fill_between(x_I, y, alpha=0.07, color=col)
ax1.axvline(obs_I, color=WHITE, lw=1.2, ls=":", alpha=0.7, label=f"obs={obs_I}")
ax1.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 1 — reglas activadas
rule_colors = [CYAN, AMBER, LAVENDER, GREEN, PINK]
for idx, (dI_lbl, I_lbl, b_lbl, alpha, clipped) in enumerate(activations):
    ax = styled(fig.add_subplot(gs[1, idx % 3]),
                f"R{idx+1}: {dI_lbl} ∧ {b_lbl}\nα={alpha:.3f}", "$\\beta$")
    ax.plot(x_b, mf_b[b_lbl], color=rule_colors[idx], lw=1.5, ls="--", alpha=0.5)
    ax.fill_between(x_b, clipped, alpha=0.25, color=rule_colors[idx])
    ax.plot(x_b, clipped, color=rule_colors[idx], lw=2)
    ax.axhline(alpha, color=WHITE, lw=0.8, ls=":", alpha=0.5)

# Fila 2 — agregación + defuzzificación
ax_agg = styled(fig.add_subplot(gs[2, :]), "Agregación y defuzzificación — Mamdani", "$\\beta$")
ax_agg.fill_between(x_b, agg, alpha=0.15, color=LAVENDER)
ax_agg.plot(x_b, agg, color=LAVENDER, lw=2.2, label="Conjunto agregado")
ax_agg.axvline(cog, color=CYAN,  lw=2, ls="--", label=f"COG  β* = {cog:.3f}")
ax_agg.axvline(mom, color=AMBER, lw=2, ls="--", label=f"MOM  β* = {mom:.3f}")
ax_agg.set_xlim(0, 1)
ax_agg.legend(fontsize=9, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

fig.suptitle("Sistema Mamdani — Estimación de β (modelo SIR)",
             color=WHITE, fontsize=12, fontfamily="monospace", y=0.995)
plt.savefig("mamdani_SIR.png", dpi=150, bbox_inches="tight", facecolor=BG)
plt.show()
```

```python
# ── TSK de orden 1 sobre el mismo problema SIR ────────────────────────────────
# Reutiliza x_dI, x_I, mf_dI, mf_I, deg_dI, deg_I, obs_dI, obs_I de la celda anterior

# Parámetros lineales por regla: β_i = p_i * ΔI/N + q_i * I/N + r_i
# Calibrados heurísticamente para el dominio [0,1]
tsk_rules = [
    ("creciendo",   "alto",  dict(p=3.5,  q=0.6,  r=0.10)),
    ("creciendo",   "medio", dict(p=2.8,  q=0.4,  r=0.08)),
    ("estable",     "medio", dict(p=0.5,  q=0.3,  r=0.35)),
    ("decreciendo", "medio", dict(p=-1.0, q=0.2,  r=0.30)),
    ("decreciendo", "bajo",  dict(p=-2.0, q=0.1,  r=0.15)),
]

alphas, y_tsk = [], []
for dI_lbl, I_lbl, coef in tsk_rules:
    alpha = min(deg_dI[dI_lbl], deg_I[I_lbl])
    yi    = coef["p"] * obs_dI + coef["q"] * obs_I + coef["r"]
    yi    = np.clip(yi, 0, 1)
    alphas.append(alpha)
    y_tsk.append(yi)

alphas  = np.array(alphas)
y_tsk   = np.array(y_tsk)
beta_tsk = np.sum(alphas * y_tsk) / (np.sum(alphas) + 1e-10)

# ── Figura ────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5), facecolor=BG)

# Panel izquierdo — contribución de cada regla
ax = axes[0]
ax.set_facecolor(BG)
rule_labels = [f"R{i+1}\n{r[0][:4]}∧{r[1][:3]}" for i, r in enumerate(tsk_rules)]
bars = ax.bar(rule_labels, alphas, color=[CYAN, AMBER, LAVENDER, GREEN, PINK],
              alpha=0.8, width=0.5)
ax2 = ax.twinx()
ax2.set_facecolor(BG)
ax2.scatter(rule_labels, y_tsk, color=WHITE, zorder=5, s=60, label="$y_i$ (salida regla)")
ax2.set_ylabel("$y_i = \\beta_i$ estimado", color=WHITE, fontsize=9)
ax2.tick_params(colors=GRAY, labelsize=8)
ax2.set_ylim(0, 1)
ax.set_ylabel("Grado de activación $\\alpha_i$", color=GRAY, fontsize=9)
ax.set_ylim(0, 1)
ax.set_title("TSK — Activación y salida por regla", color=WHITE,
             fontsize=10, fontfamily="monospace")
ax.tick_params(colors=GRAY, labelsize=8)
for sp in ax.spines.values(): sp.set_edgecolor(GRAY)
for sp in ax2.spines.values(): sp.set_edgecolor(GRAY)

# Panel derecho — promedio ponderado visual
ax = axes[1]
ax.set_facecolor(BG)
weighted = alphas * y_tsk
ax.barh(rule_labels, weighted, color=[CYAN, AMBER, LAVENDER, GREEN, PINK], alpha=0.75)
ax.axvline(beta_tsk * np.sum(alphas), color=WHITE, lw=1.5, ls="--", alpha=0.6)
ax.set_xlabel("$\\alpha_i \\cdot y_i$", color=GRAY, fontsize=9)
ax.set_title(f"TSK — Contribución ponderada\n$\\beta^* = {beta_tsk:.3f}$",
             color=WHITE, fontsize=10, fontfamily="monospace")
ax.tick_params(colors=GRAY, labelsize=8)
for sp in ax.spines.values(): sp.set_edgecolor(GRAY)

fig.suptitle("Sistema TSK (orden 1) — Estimación de β (modelo SIR)",
             color=WHITE, fontsize=12, fontfamily="monospace")
fig.patch.set_facecolor(BG)
plt.tight_layout()
plt.savefig("tsk_SIR.png", dpi=150, bbox_inches="tight", facecolor=BG)
plt.show()

print(f"TSK  →  β* estimado = {beta_tsk:.4f}")
```

```python
# ── Tsukamoto sobre el mismo problema SIR ────────────────────────────────────
# Conjuntos de salida MONOTÓNICOS (S o Z) — invertibles analíticamente

def S_inv(alpha, a, b):
    """Inversa de la función S: dado alpha, devuelve y tal que S(y;a,b)=alpha."""
    m = (a + b) / 2
    alpha = np.clip(alpha, 1e-6, 1 - 1e-6)
    if alpha <= 0.5:
        return a + (b - a) * np.sqrt(alpha / 2)
    else:
        return b - (b - a) * np.sqrt((1 - alpha) / 2)

def Z_inv(alpha, a, b):
    """Inversa de la función Z."""
    return S_inv(1 - alpha, a, b)

# Conjuntos de salida monotónicos para β
tsuka_output_sets = {
    "muy_bajo": ("Z", 0.05, 0.30),
    "bajo":     ("Z", 0.20, 0.50),
    "medio_S":  ("S", 0.35, 0.65),
    "alto":     ("S", 0.50, 0.80),
    "muy_alto": ("S", 0.65, 0.95),
}

tsuka_rules = [
    ("creciendo",   "alto",  "muy_alto"),
    ("creciendo",   "medio", "alto"),
    ("estable",     "medio", "medio_S"),
    ("decreciendo", "medio", "bajo"),
    ("decreciendo", "bajo",  "muy_bajo"),
]

alphas_t, y_tsuka = [], []
for dI_lbl, I_lbl, b_lbl in tsuka_rules:
    alpha = min(deg_dI[dI_lbl], deg_I[I_lbl])
    kind, a, b = tsuka_output_sets[b_lbl]
    yi = S_inv(alpha, a, b) if kind == "S" else Z_inv(alpha, a, b)
    yi = np.clip(yi, 0, 1)
    alphas_t.append(alpha)
    y_tsuka.append(yi)

alphas_t  = np.array(alphas_t)
y_tsuka   = np.array(y_tsuka)
beta_tsuka = np.sum(alphas_t * y_tsuka) / (np.sum(alphas_t) + 1e-10)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(16, 10), facecolor=BG)
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.55, wspace=0.35)

rule_colors = [CYAN, AMBER, LAVENDER, GREEN, PINK]

# Fila 0 — conjunto de salida + punto invertido por regla
for idx, (dI_lbl, I_lbl, b_lbl, alpha, yi) in enumerate(
        zip(*zip(*tsuka_rules), alphas_t, y_tsuka)):
    ax = fig.add_subplot(gs[0, idx % 3])
    ax.set_facecolor(BG)
    kind, a, b = tsuka_output_sets[b_lbl]
    y_mf = S(x_b, a, b) if kind == "S" else Z(x_b, a, b)
    ax.plot(x_b, y_mf, color=rule_colors[idx], lw=2, ls="--", alpha=0.6)
    ax.axhline(alpha, color=WHITE, lw=1, ls=":", alpha=0.5, label=f"α={alpha:.3f}")
    ax.axvline(yi,    color=rule_colors[idx], lw=2, label=f"y={yi:.3f}")
    ax.scatter([yi], [alpha], color=WHITE, zorder=5, s=50)
    ax.set_title(f"R{idx+1}: {b_lbl}", color=WHITE, fontsize=9, fontfamily="monospace")
    ax.set_xlim(0, 1); ax.set_ylim(-0.05, 1.15)
    ax.set_xlabel("$\\beta$", color=GRAY, fontsize=8)
    ax.set_ylabel("$\\mu$",   color=GRAY, fontsize=8)
    ax.tick_params(colors=GRAY, labelsize=7.5)
    for sp in ax.spines.values(): sp.set_edgecolor(GRAY)
    ax.legend(fontsize=7.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

# Fila 1 — resumen ponderado
ax_sum = fig.add_subplot(gs[1, :])
ax_sum.set_facecolor(BG)
rule_labels = [f"R{i+1}" for i in range(len(tsuka_rules))]
x_pos = np.arange(len(tsuka_rules))
bars  = ax_sum.bar(x_pos, alphas_t, width=0.35, color=rule_colors, alpha=0.6, label="$\\alpha_i$")
ax_sum.scatter(x_pos, y_tsuka, color=WHITE, zorder=5, s=70, label="$y_i$ (invertido)")
ax_sum.axhline(beta_tsuka, color=CYAN, lw=2, ls="--",
               label=f"$\\beta^*$ Tsukamoto = {beta_tsuka:.3f}")
ax_sum.set_xticks(x_pos); ax_sum.set_xticklabels(rule_labels, color=WHITE)
ax_sum.set_ylim(0, 1.1)
ax_sum.set_ylabel("valor", color=GRAY, fontsize=9)
ax_sum.set_title("Resumen — promedio ponderado Tsukamoto", color=WHITE,
                 fontsize=10, fontfamily="monospace")
ax_sum.tick_params(colors=GRAY, labelsize=8)
for sp in ax_sum.spines.values(): sp.set_edgecolor(GRAY)
ax_sum.legend(fontsize=8.5, facecolor=BG, labelcolor=WHITE, framealpha=0.3)

fig.suptitle("Sistema Tsukamoto — Estimación de β (modelo SIR)",
             color=WHITE, fontsize=12, fontfamily="monospace", y=0.995)
plt.savefig("tsukamoto_SIR.png", dpi=150, bbox_inches="tight", facecolor=BG)
plt.show()

print(f"Tsukamoto  →  β* estimado = {beta_tsuka:.4f}")

# ── Comparativa final ─────────────────────────────────────────────────────────
print("\n── Comparativa de estimaciones de β ──")
print(f"  Mamdani  COG : {cog:.4f}")
print(f"  Mamdani  MOM : {mom:.4f}")
print(f"  TSK ord.1    : {beta_tsk:.4f}")
print(f"  Tsukamoto    : {beta_tsuka:.4f}")
```

# Capítulo 4: Lógica Difusa y Sistemas Híbridos

## Modelos Suaves para Fenómenos Inciertos y Adaptativos

Este cuaderno implementa los conceptos fundamentales de la Lógica Difusa, explorando cómo la **pertenencia parcial** permite modelar conceptos lingüísticos vagos (como "alto" o "cerca de") y cómo esta lógica se convierte en un mecanismo **adaptativo y metaheurístico** cuando sus parámetros son dinámicos.

La clave para la Ciencia de Datos Avanzada es la **intepretabilidad** que ofrece esta lógica en combinación con el poder de optimización de los algoritmos evolutivos.

## 1. Funciones de Pertenencia: Cuantificando el Límite Suave

Las funciones de pertenencia $(\mu(x))$ son la base de la Lógica Difusa, ya que definen el grado con el que un valor numérico pertenece a un concepto lingüístico (un conjunto difuso). La forma de esta función (triangular, trapezoidal, gaussiana) determina cómo de gradual es la transición de "no pertenencia" a "pertenencia total".

```python
# [CÓDIGO 1: Definición de Funciones de Pertenencia]

def triangular_fuzzy(x, a, b, c):
    """Función de pertenencia triangular. a <= b <= c."""
    if x <= a or x >= c:
        return 0.0
    elif a < x <= b:
        return (x - a) / (b - a)
    else:
        return (c - x) / (c - b)

def trapezoidal_fuzzy(x, a, b, c, d):
    """Función de pertenencia trapezoidal. a <= b <= c <= d."""
    if x <= a or x >= d:
        return 0.0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1.0
    else:
        return (d - x) / (d - c)

def gaussian_fuzzy(x, c, w):
    """
    Función de pertenencia Gaussiana (suave y continua).
    c: centro (valor modal), w: ancho (dispersión).
    """
    return np.exp(-((x - c)**2) / (w**2))

print("Funciones de pertenencia cargadas.")
```

```python
# [CÓDIGO 1.2: Visualización Comparativa]

xs = np.linspace(-10, 10, 500)

# Parámetros para las funciones
tri = np.array([triangular_fuzzy(x, -5, 0, 5) for x in xs])
trap = np.array([trapezoidal_fuzzy(x, -7, -3, 3, 7) for x in xs])
gauss = gaussian_fuzzy(xs, c=0, w=3)

plt.figure(figsize=(12, 6))
plt.plot(xs, tri, label="Triangular (Bordes Duros)")
plt.plot(xs, trap, label="Trapezoidal (Zona Plana)")
plt.plot(xs, gauss, label="Gaussiana (Suave y Diferenciable)")
plt.legend(title="Forma de la Función")
plt.title("Comparación de Funciones de Pertenencia Difusas: Cuantificando el Concepto 'Cerca de Cero'")
plt.xlabel("Valor de la Variable de Entrada (x)")
plt.ylabel("Grado de Pertenencia ($\mu$)")
plt.show()

# 
```

## 2. Operadores Difusos (T-Normas y S-Normas)

Los operadores de Zadeh (MIN para AND, MAX para OR) se utilizan para combinar los grados de pertenencia de múltiples condiciones en una regla. Por ejemplo, en el operador `AND`, la verdad de la conjunción está limitada por la condición **más débil**.

```python
# [CÓDIGO 2: Operadores Difusos de Zadeh]

def fuzzy_and(a, b):
    """Conjunción (T-Norma) de Zadeh: Mínimo."""
    return min(a, b)

def fuzzy_or(a, b):
    """Disyunción (S-Norma) de Zadeh: Máximo."""
    return max(a, b)

def fuzzy_not(a):
    """Negación de Zadeh: Complemento."""
    return 1 - a

# Ejemplo de uso
a_alto = 0.7  # Pertenencia a 'Temperatura Alta'
b_baja = 0.2  # Pertenencia a 'Humedad Baja'

print(f"Temperatura Alta (a): {a_alto}")
print(f"Humedad Baja (b): {b_baja}")
print(f"AND (min): {fuzzy_and(a_alto, b_baja):.2f} (El más débil domina la conjunción)")
print(f"OR (max): {fuzzy_or(a_alto, b_baja):.2f} (El más fuerte domina la disyunción)")
print(f"NOT (complemento): {fuzzy_not(a_alto):.2f}")
```

## 3. Conjuntos Difusos Dinámicos: El Núcleo Adaptativo

La limitación de los conjuntos difusos estáticos es su incapacidad para modelar sistemas no estacionarios (cuyas categorías relevantes cambian con el tiempo). El **Conjunto Difuso Dinámico** permite que sus parámetros internos (`c` y `w`) evolucionen mediante un mecanismo similar a una mutación, lo que lo convierte en una entidad adaptable, ideal para la optimización metaheurística.

```python
# [CÓDIGO 3: Conjunto Difuso Dinámico y Mutación]

class DynamicFuzzySet:
    """
    Representa un conjunto difuso (Gaussiano) cuyos parámetros (centro y ancho)
    pueden ser ajustados o mutados a lo largo del tiempo.
    """
    def __init__(self, c, w, name=""):
        self.c = float(c) # Centro (valor modal)
        self.w = float(w) # Ancho (dispersión)
        self.name = name

    def mu(self, x):
        """Función de pertenencia Gaussiana."""
        # Utilizamos la función gaussiana por ser suave y diferenciable
        return np.exp(-((x - self.c)**2) / (self.w**2))

    def mutate(self, sigma_c=0.1, sigma_w=0.05):
        """
        Aplica una perturbación aleatoria a los parámetros, simulando
        la adaptación o la exploración en un proceso evolutivo.
        """
        self.c += np.random.normal(0, sigma_c)
        # Aseguramos que el ancho (w) no sea negativo ni demasiado pequeño.
        self.w = max(0.01, self.w + np.random.normal(0, sigma_w))

    def get_params(self):
        return (self.c, self.w)

# Ejemplo de evolución de un conjunto dinámico
dfs = DynamicFuzzySet(c=0, w=2, name="Medio")

history_c = []
history_w = []

for t in range(50):
    # El conjunto se adapta suavemente en cada 'paso de tiempo' o 'generación'
    dfs.mutate(sigma_c=0.2, sigma_w=0.1)
    history_c.append(dfs.c)
    history_w.append(dfs.w)

plt.figure(figsize=(12, 5))
plt.plot(history_c, label="Centro $c$ (Posición adaptativa)")
plt.plot(history_w, label="Ancho $w$ (Sensibilidad/Dispersión adaptativa)")
plt.legend()
plt.title("Evolución de Parámetros en un Conjunto Difuso Dinámico")
plt.xlabel("Paso de Tiempo / Iteración Evolutiva")
plt.ylabel("Valor del Parámetro")
plt.show()

# 
```

## 4. Sistemas Takagi–Sugeno (TS): Modelado Suave y Local

El modelo Takagi–Sugeno (TS) utiliza reglas donde el consecuente es un modelo lineal. La salida global del sistema se calcula como el promedio ponderado de las salidas de todos los modelos locales, donde los pesos son los grados de pertenencia (activación) de las reglas. Esto garantiza una superficie de respuesta continua y no lineal.

```python
# [CÓDIGO 4: Sistema de Inferencia Takagi–Sugeno]

def ts_model(x, rules):
    """
    Calcula la salida de un sistema Takagi-Sugeno de primer orden (consecuente lineal).
    """
    num, den = 0, 0
    for fuzzy_set, linear_fun in rules:
        # 1. Fuzzificación: Obtener el grado de pertenencia (mu)
        mu = fuzzy_set.mu(x)
        
        # 2. Consecuente: Calcular la salida local (y_local = ax + b)
        y_local = linear_fun(x)
        
        # 3. Agregación (suma ponderada: mu * y_local)
        num += mu * y_local
        den += mu
        
    # 4. Normalización (Defuzzificación)
    return num / den if den > 1e-6 else 0


# Definición de dos conjuntos difusos para particionar el espacio
A1 = DynamicFuzzySet(c=-3, w=2, name="Izquierda")
A2 = DynamicFuzzySet(c=3, w=2, name="Derecha")

# Definición de los modelos locales (consecuentes)
f1 = lambda x: 0.5 * x + 1    # Modelo para la región Izquierda
f2 = lambda x: -0.3 * x + 2   # Modelo para la región Derecha

# Estructura de las reglas: SI x es A_i ENTONCES y = f_i(x)
rules = [(A1, f1), (A2, f2)]

# Simulación y visualización
xs = np.linspace(-10, 10, 400)
ys = np.array([ts_model(x, rules) for x in xs])

plt.figure(figsize=(12, 6))
plt.plot(xs, ys, label="Salida Global TS", color='blue', linewidth=3)

# Mostrar la contribución de los modelos locales
plt.plot(xs, f1(xs), '--', label="Modelo Local f1 (en región A1)", alpha=0.6)
plt.plot(xs, f2(xs), '--', label="Modelo Local f2 (en región A2)", alpha=0.6)

# Mostrar los conjuntos difusos (el 'peso' de cada regla)
plt.plot(xs, [A1.mu(x) for x in xs], ':', label="Conjunto A1 (Pertenencia)")
plt.plot(xs, [A2.mu(x) for x in xs], ':', label="Conjunto A2 (Pertenencia)")

plt.legend()
plt.title("Sistema Takagi–Sugeno (TS): Interpolación Suave entre Modelos Locales")
plt.xlabel("Entrada x")
plt.ylabel("Salida y / Pertenencia $\mu$")
plt.ylim(-4, 4)
plt.show()

# 
```

## 5. Metaheurística y Lógica Difusa: Optimización de Parámetros

El ajuste de los parámetros del sistema TS (los centros $c_i$ y anchos $w_i$ de los conjuntos difusos) es un problema de optimización no lineal. Aquí es donde los algoritmos evolutivos (AG, DE, PSO) entran en juego, utilizando el **Error Cuadrático Medio (MSE)** como la función de aptitud (fitness) a minimizar. Los parámetros se codifican en el 'cromosoma' de la metaheurística.

```python
# [CÓDIGO 5: Función de Aptitud para Optimización Evolutiva (Conceptual)]

# 5.1 Generación de Datos Sintéticos No Lineales (Target)
np.random.seed(42)
xs_data = np.linspace(-10, 10, 100)
# Función a modelar (Ejemplo: seno amortiguado con ruido)
target_y = (np.sin(xs_data / 2) + np.cos(xs_data / 10)) * 2 + np.random.normal(0, 0.2, len(xs_data))

# 5.2 Función de Aptitud (Fitness) para una Metaheurística
def ts_fitness(params, xs, target):
    """
    Calcula el Error Cuadrático Medio (MSE) para un conjunto de parámetros (c1, w1, c2, w2).
    Esta es la función objetivo a minimizar por la metaheurística.
    """
    c1, w1, c2, w2 = params
    
    # Temporalmente ajustamos los conjuntos A1 y A2 con los parámetros propuestos por el optimizador
    A1_temp = DynamicFuzzySet(c1, abs(w1))
    A2_temp = DynamicFuzzySet(c2, abs(w2))
    rules_temp = [(A1_temp, f1), (A2_temp, f2)] # Usamos los mismos modelos locales f1, f2
    
    # Generar predicciones
    preds = np.array([ts_model(x, rules_temp) for x in xs])
    
    # Error Cuadrático Medio (MSE)
    return np.mean((preds - target)**2)

# Ejemplo de cálculo de fitness (usando los parámetros iniciales de A1 y A2)
initial_params = (A1.c, A1.w, A2.c, A2.w)
initial_error = ts_fitness(initial_params, xs_data, target_y)

print(f"El error inicial del modelo TS con parámetros no optimizados (MSE): {initial_error:.4f}")
print("Un algoritmo evolutivo buscaría un vector de parámetros que minimice este valor.")

plt.figure(figsize=(10, 5))
plt.scatter(xs_data, target_y, label='Datos Objetivo (Ruidosos)', s=10)
plt.plot(xs_data, [ts_model(x, rules) for x in xs_data], label='Predicción Inicial TS', color='red')
plt.legend()
plt.title('Modelo TS Inicial vs. Datos Objetivo')
```

## 6. Agentes Difusos y Sistemas Multiagente

Al encapsular un Conjunto Difuso Dinámico y una regla TS en una clase `FuzzyAgent`, se crea una **sociedad de agentes especializados**. Cada agente domina y modela una región del espacio. La salida global es el resultado de la cooperación ponderada de todos los agentes. Este enfoque mejora la robustez y la capacidad de adaptación ante cambios de régimen.

```python
# [CÓDIGO 6: Agentes Difusos y Modelo de Sociedad]

class FuzzyAgent:
    """
    Un agente con una región perceptual (fset) y un modelo local (model).
    """
    def __init__(self, c, w, model, name=""):
        # El conjunto difuso dinámico es la región de especialización del agente
        self.fset = DynamicFuzzySet(c, w, name=name) 
        self.model = model

    def perceive(self, x):
        """Devuelve el grado de activación (pertenencia) del agente."""
        return self.fset.mu(x)

    def act(self, x):
        """Calcula la salida de su modelo local."""
        return self.model(x)

    def adapt(self, sigma=0.05):
        """Permite que la región de especialización del agente evolucione."""
        self.fset.mutate(sigma, sigma/2)
        
    def __repr__(self):
        return f"Agente '{self.fset.name}' (c={self.fset.c:.2f}, w={self.fset.w:.2f})"


def fuzzy_society_output(x, agents):
    """
    Combina las salidas de todos los agentes ponderadas por su grado de percepción.
    """
    weights = np.array([ag.perceive(x) for ag in agents])
    outputs = np.array([ag.act(x) for ag in agents])
    
    sum_weights = np.sum(weights)
    
    if sum_weights < 1e-6:
        return 0
        
    # Salida agregada (promedio ponderado)
    return np.dot(weights, outputs) / sum_weights


# Ejemplo con tres Agentes especializados
agents = [
    FuzzyAgent(-6, 2, lambda x: 0.3 * x + 1, name="Régimen Bajo"),
    FuzzyAgent( 0, 1.5, lambda x: 2,           name="Régimen Medio"),
    FuzzyAgent( 6, 2, lambda x: -0.2 * x + 4, name="Régimen Alto"),
]

print("Agentes especializados en distintas regiones:")
for agent in agents:
    print(agent)

xs_agent = np.linspace(-10, 10, 400)
ys_society = [fuzzy_society_output(x, agents) for x in xs_agent]

plt.figure(figsize=(12, 6))
plt.plot(xs_agent, ys_society, label="Modelo Colectivo (Sociedad Difusa)", color='purple', linewidth=3)

# Visualizar la especialización de los agentes
for agent in agents:
    mu = [agent.perceive(x) for x in xs_agent]
    plt.plot(xs_agent, mu, ':', alpha=0.7, label=f"Percepción {agent.fset.name}")

plt.legend(loc='lower left')
plt.title("Sistema Multiagente Difuso: Cooperación por Especialización")
plt.xlabel("Entrada x")
plt.ylabel("Salida Agregada / Pertenencia $\mu$")
plt.show()

# 
```

## 7. Sistemas Difusos Fractales: Interpretando Estructuras Complejas

En sistemas caóticos o complejos, los atractores presentan estructuras fractales. Un conjunto difuso puede utilizarse para asignar un **significado conceptual suave** a una región geométrica compleja (como un lóbulo o un anillo). La función `fuzzy_shell` ilustra cómo se puede definir un concepto como "cercano a una distancia específica de un centro".

```python
# [CÓDIGO 7: "Cáscara Difusa" alrededor de una Estructura]

def fuzzy_shell(x, center, radius, softness):
    """
    Función de pertenencia que define una "cáscara" o "anillo" alrededor de un centro,
    útil para envolver regiones de un atractor (ej. un lóbulo del atractor de Lorenz).
    """
    # Distancia euclidiana del punto al centro
    dist = np.linalg.norm(x - center)
    
    # Cuantifica la cercanía a la distancia ideal (radius) mediante una Gaussiana
    return np.exp(-((dist - radius)**2) / (softness**2))

# Generamos puntos simulando una sección anular ruidosa de un atractor 2D
np.random.seed(42)
num_points = 1000
angles = np.random.uniform(0, 2 * np.pi, num_points)
r_base = 2.0
radii = r_base + np.random.normal(0, 0.4, num_points) # Añadir ruido
points = np.column_stack([radii * np.cos(angles), radii * np.sin(angles)])

# Calculamos el grado de pertenencia a la 'Cáscara' de radio 2.0
shell_mu = np.array([fuzzy_shell(p, center=np.array([0.0, 0.0]),
                                 radius=2.0, softness=0.5) for p in points])

plt.figure(figsize=(8, 8))
# Coloreamos cada punto por su grado de pertenencia a la 'cáscara'
scatter = plt.scatter(points[:, 0], points[:, 1], c=shell_mu, cmap='plasma', alpha=0.7)
plt.colorbar(scatter, label="Grado de Pertenencia Difusa ($\mu$)")
plt.title("Representación Difusa ('Cáscara') de una Estructura Compleja 2D")
plt.xlabel("Dimensión X (Estado 1)")
plt.ylabel("Dimensión Y (Estado 2)")
plt.axis('equal')
plt.show()

# 
```

En la realidad, la nitidez es una rareza. Los límites que definen categorías relevantes para el modelado de sistemas —desde la "temperatura alta" de una planta industrial hasta la "alta volatilidad" de un mercado financiero, o el "nivel de excitación" de una señal cerebral— no son líneas divisorias estrictas. Son, por naturaleza, **gradientes, matices y zonas intermedias**. La mayor parte de los fenómenos naturales, biológicos y económicos se expresan a través de transiciones suaves.La lógica clásica o Booleana, con su rigor binario (verdadero/falso; 1/0), es la base formal de las matemáticas y la informática, y funciona perfectamente en entornos donde la verdad es absoluta. Sin embargo, cuando esta lógica rígida se enfrenta al **lenguaje humano, la imprecisión en las mediciones o la vaguedad inherente a los conceptos cualitativos**, se vuelve dramáticamente insuficiente.La **Lógica Difusa (Fuzzy Logic)**, introducida por Lotfi Zadeh en 1965, surge precisamente para resolver esta insuficiencia. Su idea fundamental es simple, pero profundamente revolucionaria: **los conceptos pueden tener grados de verdad**.En lugar de forzar una decisión binaria, la lógica difusa nos permite expresar la pertenencia de un elemento a un concepto lingüístico mediante un número continuo en el intervalo $[0, 1]$:- Un valor de 0 implica la no-pertenencia plena.- Un valor de 1 implica la pertenencia plena.- Un valor intermedio (ej. 0.7) indica un **grado de pertenencia parcial**.Esta graduación nos proporciona la herramienta matemática para codificar la intuición humana y los fenómenos continuos, permitiendo que un dato pertenezca simultáneamente, aunque con diferente intensidad, a múltiples categorías (ej. "moderada" y "alta").

Un **conjunto difuso ($A$)** no es más que la representación formal de un concepto impreciso. Mientras que un conjunto clásico solo permite la inclusión o exclusión total de un elemento, un conjunto difuso define, para cada elemento de su universo de discurso ($U$), un **grado de pertenencia** $\mu_A(x)$.Esta relación se define a través de una **función de pertenencia** ($\mu_A: U \to [0, 1]$). Es la curva, la forma de esta función, la que modela matemáticamente el concepto lingüístico (ej. "caliente" o "pequeño").Las formas más comunes que se emplean en ingeniería y ciencia de datos son:- **Triangulares y Trapezoidales:** Son simples, computacionalmente económicas y se usan a menudo en sistemas de control y prototipado rápido. Permiten zonas de pertenencia completa (valor 1).- **Gaussianas (y Campana Generalizada):** Son suaves y continuas. Su naturaleza diferenciable las hace ideales para ser ajustadas o **optimizadas mediante algoritmos de gradiente o técnicas metaheurísticas**.- **Sigmoidales:** Se emplean para modelar transiciones abiertas, como el concepto de "muy grande" o "muy pequeño", donde la pertenencia tiende gradualmente a 0 o 1 en los límites del universo.La capacidad de definir estas fronteras de manera **suave** (en contraste con los límites abruptos de los conjuntos clásicos) es lo que convierte a la lógica difusa en un lenguaje tan adecuado para los sistemas continuos.

La verdadera potencia de la lógica difusa reside en su capacidad para modelar el razonamiento humano a través de **reglas de inferencia** expresadas en lenguaje natural:$$\text{SI (Condición Difusa) ENTONCES (Consecuente Difuso o Numérico)}$$Ejemplo:$$\text{SI} \quad (\text{Velocidad} \ \text{es} \ \text{ALTA}) \quad \text{Y} \quad (\text{Distancia} \ \text{es} \ \text{CORTA})$$$$\text{ENTONCES} \quad (\text{Frenado} \ \text{es} \ \text{FUERTE})$$Al sustituir las variables lógicas por grados de pertenencia, la inferencia difusa se convierte en un mecanismo de interpolación suave entre reglas de conocimiento.Los dos modelos principales de inferencia son:- **Modelo Mamdani (Salida Difusa):** El consecuente es un conjunto difuso ("Frenado es FUERTE"). La salida final requiere un proceso de **defuzzificación** para obtener un valor numérico ejecutable (ej. qué tanta presión aplicar al freno).- **Modelo Takagi–Sugeno (TS) (Salida Numérica/Funcional):** El consecuente es una función lineal o constante de las variables de entrada (ej. $y = ax + b$). Su estructura de modelado local es excepcionalmente útil para sistemas dinámicos.

Es fundamental entender la lógica difusa bajo un doble enfoque, especialmente en un contexto de ciencia de datos avanzada.### 4.1. Como Herramienta de RepresentaciónEn su forma estática, la lógica difusa es un **lenguaje no clásico** para:- **Suavizar Fronteras:** Proporciona una alternativa continua a las fronteras discretas de los **Rough Sets**, que definen conceptos con límites aproximados ("seguramente en" / "posiblemente en") pero carecen de gradualidad.- **Codificar Conocimiento Experto:** Permite a un ingeniero o científico modelar un sistema basado en reglas intuitivas, algo inalcanzable para la Lógica Temporal o Modal, que son por naturaleza simbólicas y discretas.- **Tratamiento de Datos Continuos y Ruidosos:** La continuidad de sus funciones de pertenencia la hace ideal para procesar señales y series temporales con ruido.### 4.2. Como Mecanismo Adaptativo (Metaheurístico)Aquí es donde la Lógica Difusa se integra directamente con las **Metaheurísticas**. Un conjunto difuso está definido por parámetros (ej. centro, ancho o dispersión). Si permitimos que estos parámetros **cambien, evolucionen o se ajusten** en respuesta a la dinámica del entorno o a la optimización de un error de predicción, la lógica difusa deja de ser solo un lenguaje estático y se convierte en un **componente activo de la adaptación y la búsqueda**.Esto se logra cuando:- Los parámetros de las funciones de pertenencia se codifican como cromosomas y se ajustan mediante **Algoritmos Genéticos (AG), Optimización por Enjambre de Partículas (PSO), o Estrategias Evolutivas (DE)**.- La estructura misma de las reglas (la selección de variables de entrada, la topología) se optimiza mediante **Programación Genética (PG)**.En este contexto híbrido, la lógica difusa actúa como un **sensor inteligente** cuya propia percepción del mundo se afina iterativamente para minimizar el error, un proceso intrínsecamente metaheurístico.

La principal limitación de los conjuntos difusos tradicionales es su **naturaleza atemporal y fija**. Un conjunto difuso fijo (ej. "Precio Alto") puede ser una representación instantánea adecuada, pero no captura la **evolución temporal** de las categorías de un sistema dinámico o no estacionario. Por ejemplo, el concepto de "demanda normal de energía" varía drásticamente entre el verano y el invierno.El concepto de **Conjunto Difuso Dinámico ($A_D$)** resuelve esta limitación al incluir una dimensión temporal en la función de pertenencia. Para un valor dado $x$, su grado de pertenencia en $A$ cambia a lo largo del tiempo $T$ según una **función temporal de pertenencia** $\mu^x(t): T \to [0, 1]$.Los conjuntos difusos dinámicos pueden verse como superficies en el espacio $U \times T \times [0, 1]$. En la práctica, esto implica que los parámetros que definen la forma del conjunto difuso (principalmente el **valor modal** —la posición— y el **ancho de banda** —la dispersión—) cambian con el tiempo o en función de otras variables contextuales.- **Aplicación a Series de Tiempo:** Una variante de estos, los **conjuntos difusos dinámicos suaves** (introducidos en tu tesis), han demostrado ser superiores a los fijos para la **predicción de series de tiempo no estacionarias**, ya que su lógica se adapta al régimen cambiante del fenómeno subyacente (tendencias crecientes o decrecientes).De esta manera, cada conjunto difuso se convierte en una **entidad que aprende y se desplaza**, actuando como un sensor perceptivo adaptable a los cambios de régimen de un sistema complejo.

Los Sistemas de Inferencia Takagi–Sugeno (TS) ofrecen una de las arquitecturas híbridas más eficientes. Su fortaleza radica en el balance entre **interpretabilidad lingüística** (a través de los conjuntos difusos del antecedente) y **potencia de modelado numérico** (a través de las funciones del consecuente).En un modelo TS, las reglas no tienen un consecuente difuso como en Mamdani, sino un modelo local, típicamente lineal o de orden cero:$$\text{SI } (x_1 \ \text{es} \ A_i) \ \text{Y} \ (x_2 \ \text{es} \ B_i) \quad \text{ENTONCES} \quad y = a_i x_1 + b_i x_2 + c_i$$Cada regla actúa como un **modelo local** que es válido en una pequeña región del espacio de estados (la definida por los conjuntos difusos $A_i$ y $B_i$). El sistema agrega las salidas de todos los modelos locales mediante un **promedio ponderado suave**, donde los pesos son los grados de pertenencia de las reglas activadas.$$y_{\text{final}} = \frac{\sum_{i=1}^{R} \omega_i y_i}{\sum_{i=1}^{R} \omega_i}$$El resultado es un **modelo global continuo, suave y no lineal** construido a partir de piezas locales simples y cuya lógica subyacente es completamente interpretable. Esta capacidad de interpolación suave entre modelos lineales lo convierte en una herramienta excelente para:- **Control Robótico y de Procesos:** Asegurando transiciones estables.- **Predicción de Series de Tiempo No Lineales:** Modelando diferentes regímenes de comportamiento (ej. crecimiento, colapso, estancamiento) con modelos locales especializados.

Al estudiar sistemas caóticos como Lorenz o series financieras, se observan **patrones fractales** en el espacio de estados. Estos atractores complejos se caracterizan por tener múltiples escalas y densas estructuras auto-similares.La lógica difusa ofrece un mecanismo elegante para **suavizar la interpretación** de estas estructuras. En lugar de tener que describir cada punto del atractor, un sistema difuso (generalmente TS) permite:- Definir reglas que corresponden a **regiones conceptuales** del atractor (ej. "el lóbulo superior", "el punto de equilibrio").- Capturar las **transiciones** entre subcomportamientos mediante la naturaleza gradual de las funciones de pertenencia.El resultado es un **sistema híbrido** que utiliza un conjunto de reglas lingüísticas para capturar, de manera interpretable, la dinámica compleja de un atractor fractal.

Llevando el concepto de conjunto difuso dinámico (Sección 5) un paso más allá, un **agente difuso** es una entidad que no solo ajusta su percepción, sino que también **toma decisiones, compite y se especializa** en una región específica del espacio de estados.En un **Sistema Multiagente (SMA) Difuso**, cada agente puede poseer:- Una **Región Perceptual Difusa** (su área de especialización, definida por un conjunto difuso dinámico).- Un **Modelo Local** (su regla TS para esa región).- Una **Función de Aptitud** y **Mecanismos de Mutación/Adaptación** que le permiten evolucionar.La robustez de los SMA difusos se debe a la **diversidad** y a la **cooperación/competencia** de los agentes. Si un sistema dinámico cambia de régimen (ej. una crisis financiera), algunos agentes (los especializados en baja volatilidad) pueden volverse inactivos, mientras que los nuevos agentes (especializados en alta volatilidad) asumen el control de la predicción o la acción. Este enfoque distribuido y adaptable supera a menudo la robustez de un modelo monolítico como una red neuronal simple.

En la práctica moderna de la Ciencia de Datos, el enfoque más poderoso es el **híbrido**. La lógica difusa no busca reemplazar, sino potenciar, otras técnicas:| Técnica | Aporte Clave al Híbrido || :--- | :--- || **Lógica Difusa** | Proporciona **interpretación** del conocimiento, **suavidad** en las transiciones y un **lenguaje** adaptable. || **Algoritmos Evolutivos (AG, DE, PSO)** | Se encargan del **ajuste fino y global** de los parámetros de los conjuntos difusos para minimizar el error de predicción. || **Redes Neuronales** | Aportan un poder de **modelado numérico local** que puede usarse como consecuente de una regla (ej. un nodo neuronal en el consecuente de una regla TS). || **Programación Genética** | Permite **optimizar la estructura** (las reglas) del sistema, no solo sus parámetros. |

La Lógica Difusa es mucho más que una técnica algorítmica; es una **filosofía de modelado** que reconoce la realidad del mundo como un continuo de matices. Nos permite pasar de un mundo de límites absolutos a un mundo de gradientes, ambigüedades y transiciones.En este capítulo, hemos transitado desde los conjuntos difusos estáticos hasta los **agentes difusos dinámicos y autónomos**. Hemos visto cómo la integración con las metaheurísticas convierte estos modelos en **entidades vivas** capaces de:- Interpretar los datos en términos conceptuales.- Adaptar su propia percepción (sus conjuntos difusos) a los cambios de régimen.- Cooperar y competir para mantener un modelo colectivo robusto.El resultado final es una plataforma para construir **modelos inteligentes, suaves, adaptables y, fundamentalmente, interpretables**, que pueden enfrentarse a la complejidad de las series temporales no estacionarias y los sistemas caóticos que definen nuestro universo de datos.
