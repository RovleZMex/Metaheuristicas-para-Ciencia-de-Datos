## 1. Una solución avanzada al *No Free Lunch Theorem*

El **No Free Lunch Theorem** implica que ningún algoritmo es universalmente óptimo. Los SMA ofrecen una solución elegante a este problema: simplemente desplegamos diversos agentes simultáneamente, cada uno de ellos provisto de herramientas que no suelen integrarse en un solo proceso. Esta estrategia de resolver un problema de forma colectiva aplica a cualquier modelo de aprendizaje, y tiene una consecuencia que la industria tardó décadas en tomar en serio:

> **Escalar un modelo único no resuelve el NFL. Lo amplifica.**

La intuición es sencilla. Un modelo monolítico más grande tiene más capacidad representacional, pero sigue siendo *un solo modelo* con *un solo conjunto de parámetros* que intenta ser bueno en todo. A medida que escala, el costo computacional crece cuadráticamente, pero la cobertura del espacio de problemas no mejora de forma proporcional. El modelo se vuelve más caro y marginalmente mejor en promedio —pero sigue perdiendo en los paisajes donde no fue entrenado con suficiente densidad.

### El costo de la generalización

Para ver el problema con precisión, consideremos un LLM que intenta ser competente en código, razonamiento matemático, traducción, medicina 
y derecho simultáneamente. Cada dominio tiene su propia distribución, su propio  vocabulario especializado y sus propias regularidades.

Entrenar un modelo denso de $N$ parámetros sobre todos estos dominios tiene tres costos que se acumulan:

| Costo | Descripción | Consecuencia |
|:---|:---|:---|
| **Computacional** | Cada token activa *todos* los parámetros | Inferencia cara aunque el dominio sea estrecho |
| **De interferencia** | Los dominios compiten por la misma representación | Mejoras en uno degradan el desempeño en otro |
| **De especialización** | El modelo aprende el promedio de los dominios | No es experto en ninguno; es mediocre en todos |

El tercer costo es el más peligroso, porque suele no identificarse. Un modelo que intenta ser bueno en todo aprende representaciones que son un *compromiso* entre dominios. En términos geométricos: el modelo converge a un punto del espacio de parámetros que minimiza el error promedio sobre todos los dominios, pero ese punto no es óptimo para ninguno de ellos individualmente.

Esto no es un fallo de entrenamiento. Es una consecuencia estructural de intentar resolver con un solo conjunto de parámetros lo que requiere múltiples soluciones especializadas.

---

### La alternativa: especialización + coordinación

La respuesta que suele adoptándose al final —en ingeniería, en biología, en economía— es siempre la misma: **división del trabajo con coordinación**.

Un hospital no tiene un médico que sea experto en todo. Tiene cardiólogos, neurólogos, oncólogos —cada uno con conocimiento profundo en su dominio— y un sistema de derivaciones que dirige al paciente al especialista correcto. El sistema como conjunto resuelve lo que ningún médico individual podría.

La pregunta computacional es: ¿cómo implementamos ese principio en un modelo de aprendizaje automático?

La respuesta tiene dos ingredientes:

1. **Expertos especializados:** múltiples submodelos, cada uno optimizado para  un subconjunto del espacio de entrada.

2. **Un mecanismo de enrutamiento:** un componente que decide, para cada  entrada específica, qué experto o expertos activar.

Esta arquitectura es lo que se conoce como **Mixture of Experts**.

La ganancia no es solo en calidad de la predicción. Es en **eficiencia**: si  tenemos 8 expertos y activamos 2 por entrada, el costo de inferencia es el de  2 expertos, pero la *capacidad representacional* es la de 8. El modelo puede  ser enorme en parámetros totales y económico en parámetros activos.

Esa es la idea central de **Mixtral 8x7B** (*Mistral AI*) y de la arquitectura que, según reportes no confirmados oficialmente, subyace a *GPT-4*: modelos con decenas de expertos donde solo una fracción se activa por inferencia.

### La conexión con lo que ya construimos

Antes de formalizar, vale la pena detenerse un momento. El patrón que acabamos de describir no es nuevo en este curso: Es lo que ya vimos  con los Agentes predictores + árbitro (Blackboard). La diferencia entre lo que hemos visto y MoE formal es de *nivel de 
abstracción*: en los SMA que construimos, el enrutamiento era implícito (el coordinador, el quórum, el votante), más basado en heurísticas. En MoE, el enrutamiento es un componente explícito y *diferenciable* que se aprende junto con los expertos.

Esa diferenciabilidad es lo que hace posible entrenar el sistema completo con gradientes. Y es el puente que cruza desde las metaheurísticas hacia los grandes modelos de lenguaje modernos. MoE no es una arquitectura nueva. Es el nombre que la comunidad de deep learning le puso al principio que hemos aplicado a lo largo del curso:

## 2. Mixture of Experts — el mecanismo

### Definición formal

Un **Mixture of Experts** es una arquitectura compuesta por tres elementos:

$$\text{MoE} = \langle \{E_1, E_2, \ldots, E_n\},\; G,\; \text{Agregador} \rangle$$

donde:

- $\{E_1, \ldots, E_n\}$: conjunto de **expertos** —submodelos, cada uno con sus propios parámetros, especializados en una región del espacio de entrada.
- $G$: **gating network** —un componente que, dada una entrada $x$, produce  un vector de pesos $g(x) \in \mathbb{R}^n$ indicando qué expertos activar y con qué intensidad.
- **Agregador**: combina las salidas de los expertos activos ponderadas por $g(x)$.

<p align="center" style="margin-left:30px; margin-bottom: -15px;">
  <img src="images/MoE.png" width="800">
</p>
<br>

La salida del sistema para una entrada $x$ es:

$$y = \sum_{i=1}^{n} g_i(x) \cdot E_i(x)$$

En el caso **denso** todos los expertos contribuyen ($g_i(x) > 0\; \forall i$).  

En el caso sparse —el más usado en la práctica— el mecanismo de enrutamiento actúa de forma estrictamente selectiva: sólo los $k$ agentes con mayor $g(x)$ son físicamente activados en memoria, mientras que el resto de los expertos permanecen inactivos, omitiendo por completo el cálculo de sus salidas:

$$y = \sum_{i \in \text{top-k}(g(x))} g_i(x) \cdot E_i(x)$$

La esparsidad es la clave de la eficiencia: con $n=8$ expertos y $k=2$, activamos el 25% de los parámetros por inferencia manteniendo la capacidad representacional del sistema completo pero con una fracción de su costo computacional.

---

### La red de enrutamiento (o *gating network*)

La red de enrutamiento es el árbitro del sistema. Su función es aprender, durante el entrenamiento, qué experto es más competente para cada tipo de entrada.

La implementación más común es un **softmax sobre una proyección lineal**:

$$g(x) = \text{softmax}(W_g \cdot x)$$

donde $W_g \in \mathbb{R}^{n \times d}$ es una matriz de parámetros aprendibles y $d$ es la dimensión de la entrada. Para *top-k sparse*, aplicamos la selección antes del softmax:

$$g_i(x) = \begin{cases} \text{softmax}(W_g \cdot x)_i & \text{si } i \in \text{top-k}(W_g \cdot x) \\ 0 & \text{en otro caso} \end{cases}$$

La diferencia con el árbitro del ejercicio de Blackboard es fundamental:

| Árbitro del Blackboard | Gating network |
|:---|:---|
| Heurística fija (mediana) | Parámetros aprendidos durante entrenamiento |
| Regla definida por el diseñador | Regla descubierta de los datos |
| No diferenciable | Diferenciable → entrenable con gradientes |
| Agencia blanda | Agencia blanda, pero adaptativa |

---

### El problema del experto colapsado

Hay un problema fundamental que aparece al entrenar MoE con naive softmax: el **colapso del enrutamiento** (*router collapse*).

Si por azar el experto $E_1$ recibe una entrada ligeramente favorable al inicio del entrenamiento y produce una salida marginalmente mejor, la red de enrutamiento aprende a enviarle más tráfico. Con más tráfico, $E_1$ se entrena más y mejora. Con más mejora, recibe aún más tráfico. El ciclo se retroalimenta hasta que $E_1$ recibe casi toda la carga y el resto de los expertos degeneran por falta de gradientes.

La solución estándar es añadir un **término de balance** a la función de pérdida que penaliza distribuciones de carga muy asimétricas:

$$\mathcal{L}_{\text{total}} = \mathcal{L}_{\text{tarea}} + \alpha \cdot \mathcal{L}_{\text{balance}}$$

donde $\mathcal{L}_{\text{balance}}$ penaliza cuando un experto recibe significativamente más tráfico que los demás. En la práctica, Mixtral y modelos similares usan variantes de **noisy top-k gating**: se añade ruido gaussiano a los logits antes de seleccionar top-k, forzando exploración de expertos menos usados durante el entrenamiento:

$$\tilde{g}(x) = W_g \cdot x + \epsilon, \quad \epsilon \sim \mathcal{N}(0, \sigma^2)$$

La conexión con las metaheurísticas es directa: el colapso del enrutamiento es exactamente el mismo problema que la convergencia prematura en GA. La solución —ruido para forzar exploración— es también la misma.

### Implementación mínima

Para anclar la intuición, implementemos un MoE diferenciable en NumPy puro —sin frameworks de deep learning— sobre un problema de regresión simple.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# ──────────────────────────────────────────────────────────────────────────────
# DATOS — dos regímenes bien diferenciados
# El MoE debería aprender a especializar un experto por régimen,
# sin que le digamos dónde está el límite.
# ──────────────────────────────────────────────────────────────────────────────
def generar_datos(n=300):
    """
    Mismo espíritu que generar_serie() en el ejercicio SMA:
    dos comportamientos distintos que ningún modelo único captura bien.
      - Régimen 1 (x < 0): sinusoide amortiguada  ← AgenteARIMA hubiera ganado aquí
      - Régimen 2 (x ≥ 0): cuadrática con ruido   ← AgenteMLP hubiera ganado aquí
    """
    X = np.linspace(-3 * np.pi, 3 * np.pi, n).reshape(-1, 1)
    y = np.where(
        X.ravel() < 0,
        np.exp(0.15 * X.ravel()) * np.sin(X.ravel()),
        0.1 * X.ravel()**2 - 0.5 + np.random.normal(0, 0.1, n)
    )

    plt.figure(figsize=(6, 3))
    plt.plot(X, y, linewidth=2)
    plt.axvline(
        x=0,
        linestyle="--",
        alpha=0.7,
        label="Cambio de régimen"
    )
    plt.title("Función objetivo con dos regímenes")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)    
    plt.show()

    return X, y

X, y = generar_datos()

# ──────────────────────────────────────────────────────────────────────────────
# EXPERTOS — análogos a AgenteARIMA, AgenteKNN, AgenteMLP
# Misma estructura: clase abstracta implícita con predecir() que las subclases
# implementan. La diferencia: aquí son diferenciables (tienen gradientes).
# ──────────────────────────────────────────────────────────────────────────────
class Experto:
    """
    Clase abstracta implícita — igual que AgenteConsultor en el SMA.
    Cada experto tiene sus propios parámetros w, análogos al modelo
    interno de cada agente (ARIMA, KNN, MLP).
    """
    def __init__(self, n_features):
        self.w = np.random.randn(n_features) * 0.01  # parámetros propios

    def predecir(self, phi):
        """Equivalente a predecir_paso() en AgenteConsultor."""
        raise NotImplementedError

    def gradiente(self, phi, error):
        raise NotImplementedError


class ExpertoLineal(Experto):
    """
    Análogo a AgenteARIMA: captura tendencia global mediante
    una proyección lineal sobre features polinomiales.
    """
    def predecir(self, phi):
        return phi @ self.w

    def gradiente(self, phi, error):
        return phi.T @ error / len(error)


def expandir_features(X, grado=4):
    """
    Base polinomial normalizada.
    Sin normalización, x⁴ con x ∈ [-3π, 3π] produce valores ~10³
    que explotan el gradiente en las primeras épocas.
    """
    phi = np.hstack([X**i for i in range(grado + 1)])
    # Normalizar cada columna: media 0, std 1
    media = phi.mean(axis=0)
    std   = phi.std(axis=0) + 1e-8    # evitar división por cero en columna constante
    return (phi - media) / std

# ──────────────────────────────────────────────────────────────────────────────
# GATING NETWORK — análogo a AgenteCoordinador, pero aprendible
#
# AgenteCoordinador usaba una heurística fija: mediana con quórum.
# El GatingNetwork aprende a enrutar desde los datos mediante gradientes.
#
# Blackboard.pizarra[paso] = {agente: valor}   →   g(x) = [g₁, g₂, ..., gₙ]
# AgenteCoordinador.quorum = 2                 →   top-k = 2 (sparse gating)
# mediana de propuestas disponibles            →   promedio ponderado por g(x)
# ──────────────────────────────────────────────────────────────────────────────
class GatingNetwork:
    """
    Aprende qué experto activar para cada entrada x.
    En el SMA, el coordinador esperaba propuestas y aplicaba la mediana.
    Aquí, el gating decide *antes* de que los expertos trabajen — y
    esa decisión mejora con cada iteración de entrenamiento.
    """
    def __init__(self, n_features, n_expertos):
        self.W = np.random.randn(n_features, n_expertos) * 0.01

    def forward(self, phi, top_k=None):
        """
        Calcula pesos de activación para cada experto.
        top_k=None → gating denso (todos contribuyen)
        top_k=1    → gating sparse (solo el experto dominante)
        """
        logits = phi @ self.W
        logits -= logits.max(axis=1, keepdims=True)   # estabilidad numérica
        g = np.exp(logits)

        if top_k is not None:
            # Sparse gating: cero a todos excepto top-k
            # Análogo a quorum, pero selecciona los MÁS relevantes,
            # no los primeros en llegar.
            mascara = np.zeros_like(g)
            idx_top = np.argsort(g, axis=1)[:, -top_k:]
            for i, idx in enumerate(idx_top):
                mascara[i, idx] = g[i, idx]
            g = mascara

        return g / (g.sum(axis=1, keepdims=True) + 1e-8)  # normalización

    def gradiente(self, phi, g, grad_g):
        d_logits = g * (grad_g - (grad_g * g).sum(axis=1, keepdims=True))
        return phi.T @ d_logits / len(phi)

# ──────────────────────────────────────────────────────────────────────────────
# PIZARRA MoE — análoga a Blackboard, pero vectorizada
#
# En el SMA, la Blackboard era un objeto compartido donde los agentes
# depositaban propuestas asíncronamente. Aquí la "pizarra" es la matriz E:
#   E[n, i] = predicción del experto i para la muestra n
# El gating decide cuánto peso darle a cada columna.
# ──────────────────────────────────────────────────────────────────────────────
def construir_pizarra(phi, expertos):
    """
    Construye la 'pizarra' vectoriz
    ada: todos los expertos predicen
    para todas las muestras simultáneamente.
    En el SMA esto ocurría de forma asíncrona y secuencial (paso a paso).
    Aquí ocurre en un forward pass — la vectorización reemplaza los hilos.
    """
    return np.column_stack([e.predecir(phi) for e in expertos])  # (N, n_expertos)

# ──────────────────────────────────────────────────────────────────────────────
# ENTRENAMIENTO
# ──────────────────────────────────────────────────────────────────────────────
def entrenar_moe(X, y, n_expertos=2, grado=4, lr=0.005,   # lr reducido: 0.05 → 0.005
                 epocas=300, top_k=None):
    phi  = expandir_features(X, grado)
    y    = y.reshape(-1, 1)

    # Normalizar y también — misma razón
    y_media = y.mean()
    y_std   = y.std() + 1e-8
    y_norm  = (y - y_media) / y_std

    expertos = [ExpertoLineal(phi.shape[1]) for _ in range(n_expertos)]
    gating   = GatingNetwork(phi.shape[1], n_expertos)

    historia = {"loss": [], "pesos_gating": []}

    print(f"  Entrenando MoE: {n_expertos} expertos, "
          f"{'denso' if top_k is None else f'sparse top-{top_k}'}, "
          f"{epocas} épocas\n")
    print(f"  {'Época':>6} │ {'Loss':>8} │ Pesos gating (promedio por experto)")
    print(f"  {'─'*70}")

    for epoca in range(epocas):

        # ── Forward ──────────────────────────────────────────────────────────
        E     = construir_pizarra(phi, expertos)
        g     = gating.forward(phi, top_k=top_k)
        y_hat = (g * E).sum(axis=1, keepdims=True)

        # ── Loss (sobre y normalizada) ────────────────────────────────────────
        error = y_hat - y_norm
        loss  = float((error**2).mean())

        # Detección de divergencia — aborta antes de overflow
        if not np.isfinite(loss):
            print(f"  ⚠️  Divergencia detectada en época {epoca}. "
                  f"Reducir lr o revisar datos.")
            break

        # ── Backward: expertos ────────────────────────────────────────────────
        error_flat = error.ravel()
        for i, exp in enumerate(expertos):
            grad_w  = phi.T @ (2 * error_flat * g[:, i]) / len(phi)
            exp.w  -= lr * grad_w

        # ── Backward: gating ──────────────────────────────────────────────────
        grad_g = 2 * error_flat[:, None] * E / len(phi)
        dW     = gating.gradiente(phi, g, grad_g)

        # Clip de gradientes — previene overflow cuando el gating
        # produce logits extremos en las primeras épocas
        dW = np.clip(dW, -1.0, 1.0)
        gating.W -= lr * dW

        historia["loss"].append(loss)
        historia["pesos_gating"].append(g.mean(axis=0).copy())

        if epoca % 60 == 0:
            pesos_str = "  ".join([f"E{i+1}={g.mean(axis=0)[i]:.3f}"
                                   for i in range(n_expertos)])
            print(f"  {epoca:>6} │ {loss:>8.4f} │ {pesos_str}")

    print(f"  {'─'*70}\n")

    # Guardar parámetros de normalización para usarlos en visualización
    return expertos, gating, historia, (y_media, y_std)

print("═"*75)
print("  MoE — Implementación mínima")
print("═"*75 + "\n")

top_k = None
expertos, gating, historia, (y_media, y_std) = entrenar_moe(
    X, y, n_expertos=5, top_k=top_k
)
````

El comportamiento observado es interesante porque es evidente que el modelo **realmente aprende** a reducir la pérdida de forma estable: $\text{Loss}: 0.9774 \rightarrow 0.1817$ 

Esto indica que:

1. Los expertos están aprendiendo parámetros útiles.
2. El sistema completo converge numéricamente.
3. El MoE logra aproximar ambos regímenes de la serie.

Sin embargo, aparece un fenómeno importante: Todos los agentes obtienen un peso promedio idéntico $E_i = 0.200$

Esto significa que el **gating network no está especializándose**. ¿Por qué ocurre esto? Aunque conceptualmente el problema contiene dos regímenes distintos:

- $x < 0$ → comportamiento sinusoidal amortiguado
- $x \ge 0$ → comportamiento cuadrático

el modelo presenta una fuerte **simetría inicial**:

- ambos expertos comienzan casi iguales,
- el gating inicia cercano a una distribución uniforme,
- los gradientes que reciben son prácticamente equivalentes.

Como consecuencia:

$$
\text{experto}_1 \approx \text{experto}_2
$$

y el gating no encuentra incentivos para asignar regiones distintas del espacio de entrada. El sistema converge entonces hacia una solución degenerada:

> “dos expertos haciendo casi lo mismo”

en lugar de:

> “cada experto especializado en un régimen distinto”.


Este resultado muestra una diferencia importante entre:

- la **capacidad de especialización**, y
- la **especialización efectiva**.

Un modelo MoE no garantiza automáticamente que emerjan expertos especializados.

La especialización sólo aparece si:

1. existe diversidad entre expertos,
2. el gating recibe señales suficientemente distintas,
3. la optimización rompe la simetría inicial.

En implementaciones reales de MoE, esto suele lograrse mediante:

- **sparse gating** ($top\text{-}k$),
- ruido en los *logits*,
- regularización de carga (*load balancing*),
- inicializaciones diferentes,
- expertos más expresivos.

Observemos que en el sistema multiagente (SMA) de la lección previa, cada agente ya era distinto por diseño: la diversidad estaba impuesta arquitectónicamente. En el MoE:

- todos los expertos tienen inicialmente la misma arquitectura,
- la diversidad debe emerger durante el entrenamiento.

````python
# ──────────────────────────────────────────────────────────────────────────────
# VISUALIZACIÓN DEL MoE
#
# aquí inspeccionamos:
#
#   1. qué aprendió cada experto,
#   2. cómo enruta el gating,
#   3. cómo evoluciona el entrenamiento.
#
# La idea central es observar si emerge especialización espontánea.
# ──────────────────────────────────────────────────────────────────────────────

def visualizar_moe(
    X,
    y,
    expertos,
    gating,
    historia,
    y_media,
    y_std,
    grado=4,
    top_k=None,
    titulo=""
):

    # ──────────────────────────────────────────────────────────────────────
    # Forward completo
    # ──────────────────────────────────────────────────────────────────────
    phi = expandir_features(X, grado=grado)

    g = gating.forward(phi, top_k=top_k)
    E = construir_pizarra(phi, expertos)

    # Predicción agregada (desnormalizada)
    y_hat = (g * E).sum(axis=1) * y_std + y_media

    # Predicciones individuales
    E_des = E * y_std + y_media

    # ──────────────────────────────────────────────────────────────────────
    # Figura principal
    # ──────────────────────────────────────────────────────────────────────
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    colores = plt.cm.tab10(np.linspace(0, 1, len(expertos)))

    # =====================================================================
    # PANEL 1 — Predicciones
    # =====================================================================

    axes[0].scatter(
        X,
        y,
        s=10,
        alpha=0.25,
        color="gray",
        label="Datos"
    )

    for i in range(len(expertos)):
        axes[0].plot(
            X,
            E_des[:, i],
            linestyle="--",
            linewidth=1.5,
            color=colores[i],
            alpha=0.8,
            label=f"Experto {i+1}"
        )

    axes[0].plot(
        X,
        y_hat,
        color="black",
        linewidth=2.8,
        label="MoE agregado",
        zorder=10
    )

    axes[0].axvline(
        0,
        color="gray",
        linestyle=":",
        linewidth=1,
        alpha=0.7
    )

    axes[0].set_title(
        "Predicciones individuales y agregación",
        fontweight="bold"
    )

    axes[0].set_xlabel("x")
    axes[0].set_ylabel("y")
    axes[0].legend(fontsize=8)
    axes[0].grid(alpha=0.3)

    # =====================================================================
    # PANEL 2 — Gating
    # =====================================================================

    for i in range(g.shape[1]):

        axes[1].plot(
            X,
            g[:, i],
            linewidth=2,
            color=colores[i],
            label=f"g(x) → E{i+1}"
        )

        axes[1].fill_between(
            X.ravel(),
            0,
            g[:, i],
            alpha=0.08,
            color=colores[i]
        )

    axes[1].axvline(
        0,
        color="gray",
        linestyle=":",
        linewidth=1,
        alpha=0.7
    )

    axes[1].set_title(
        "Pesos aprendidos por el gating",
        fontweight="bold"
    )

    axes[1].set_xlabel("x")
    axes[1].set_ylabel("Peso de activación")

    axes[1].set_ylim(0, 1.05)

    axes[1].legend(fontsize=8)
    axes[1].grid(alpha=0.3)

    # =====================================================================
    # PANEL 3 — Curva de pérdida
    # =====================================================================

    axes[2].plot(
        historia["loss"],
        linewidth=2,
        color="black"
    )

    axes[2].set_title(
        "Convergencia del entrenamiento",
        fontweight="bold"
    )

    axes[2].set_xlabel("Época")
    axes[2].set_ylabel("MSE normalizado")

    axes[2].grid(alpha=0.3)

    # =====================================================================
    # Título general
    # =====================================================================

    if titulo:
        plt.suptitle(
            titulo,
            fontsize=13,
            fontweight="bold"
        )

    plt.tight_layout()
    plt.show()

# ──────────────────────────────────────────────────────────────────────────────
# Visualización final
# ──────────────────────────────────────────────────────────────────────────────

visualizar_moe(
    X,
    y,
    expertos,
    gating,
    historia,
    y_media,
    y_std,
    grado=4,
    top_k=top_k,
    titulo=f"MoE mínimo — especialización emergente vs coordinación explícita (top_k = {top_k})"
)
````

#### Comparativa entre coordinación heurística (SMA) y gating aprendido (MoE)
Veamos como se comportarían ambas estrategias, haciendo una aclaración: **Ni el SMA, ni MoA** están en su mejor aptitud en este ejemplo: 

<ol>
    <li>
        El SMA debería contar con un número mayor de agentes y, sobre todo, con una mayor heterogeneidad funcional entre ellos. En un escenario más representativo, cada agente utilizaría métodos predictivos distintos (por ejemplo, modelos estadísticos, vecinos cercanos, redes neuronales o modelos especializados por régimen), además de procesos de entrenamiento diferenciados y acceso potencialmente distinto a la información. De igual forma, el agente coordinador podría implementar mecanismos de consenso más sofisticados que una mediana fija, incluyendo estrategias adaptativas, votación ponderada, reputación dinámica o negociación multiagente. Por tanto, el SMA evaluado aquí representa una versión deliberadamente simplificada, donde la diversidad estructural y la coordinación avanzada aún no están plenamente explotadas.
    </li>
    <li>
        El MoE utilizado también se encuentra fuera de su contexto óptimo. Las arquitecturas Mixture of Experts suelen mostrar sus mayores ventajas en problemas de alta complejidad, alta dimensionalidad o fuerte heterogeneidad funcional, donde distintos expertos pueden especializarse en subespacios claramente diferenciados. En este experimento, sin embargo, los expertos son homogéneos, lineales y comparten la misma representación polinomial de entrada, lo que reduce considerablemente las posibilidades de especialización emergente. Además, la tarea utilizada es relativamente simple para una arquitectura de coordinación aprendida, por lo que el sistema termina parcialmente sobreparametrizado: el gating debe aprender rutas y mecanismos de asignación sin disponer de suficiente diversidad funcional entre expertos para aprovechar plenamente dicha capacidad adaptativa.
    </li>
</ol>

En consecuencia, la comparación realizada no debe interpretarse como una evaluación definitiva de superioridad entre SMA y MoE, sino como un análisis controlado sobre cómo distintos mecanismos de coordinación se comportan cuando la diversidad entre componentes es limitada. Precisamente por esta simplificación, el experimento permite aislar con mayor claridad el efecto del mecanismo de agregación y mostrar que la especialización efectiva depende críticamente de la heterogeneidad funcional de los expertos/agentes.


````python
# ──────────────────────────────────────────────────────────────────────────────
# COMPARATIVA EXPLÍCITA: coordinador heurístico vs gating aprendido
#
# La diferencia central entre el SMA del ejercicio y este MoE:
# ──────────────────────────────────────────────────────────────────────────────
# Comparativa con desescalado correcto
phi        = expandir_features(X, grado=4)
g          = gating.forward(phi)
E          = construir_pizarra(phi, expertos)

dictamen_sma = (np.median(E, axis=1) * y_std + y_media)
dictamen_moe = ((g * E).sum(axis=1) * y_std + y_media)

mse_sma = float(((dictamen_sma - y.ravel())**2).mean())
mse_moe = float(((dictamen_moe - y.ravel())**2).mean())

print("═"*65)
print("  COMPARATIVA: coordinador heurístico vs gating aprendido")
print("═"*65)
print(f"\n  Coordinador SMA (mediana fija):    MSE = {mse_sma:.4f}")
print(f"  Gating MoE     (softmax aprendido): MSE = {mse_moe:.4f}")
print(f"\n  Mejora relativa: {(mse_sma - mse_moe) / mse_sma * 100:.1f}%")
print("═"*65)
````

Todo ello utilizando expertos muy similares entre sí.

## 3. De MoE a MoA — cuando los expertos son modelos completos

### El salto de nivel

MoE opera al nivel de **capas y parámetros** en una red profunda: los expertos son funciones  diferenciables dentro de un modelo más grande, entrenadas end-to-end con gradientes. El gating es una red neuronal que aprende a enrutar.

**Mixture of Agents** (MoA) opera al nivel de **modelos completos**: cada agente es un LLM independiente con su propia arquitectura, sus propios pesos y, crucialmente, su propio *prompt de sistema* que define su perspectiva. El gating deja de ser una red neuronal y se convierte en un **protocolo de coordinación**.

La pérdida de diferenciabilidad tiene un costo: no podemos entrenar el sistema de coordinación con gradientes. Pero gana algo más valioso en ciertos contextos: **cada agente puede ser heterogéneo** —diferentes arquitecturas, diferentes tamaños, diferentes capacidades— y el sistema puede componerse en tiempo de ejecución sin reentrenamiento.

---

### La arquitectura MoA

Un sistema MoA tiene tres capas:

<p align="center" style="margin-left:30px; margin-bottom: -15px;">
  <img src="images/MoA-arquitectura.png" width="300">
</p>
<br>

**Capa 1 — Agentes proponentes:** reciben la misma entrada y producen respuestas independientes desde sus perspectivas especializadas. La 
diversidad aquí es una propiedad de diseño, no un accidente.

**Capa 2 — Agente agregador:** recibe las respuestas de todos los proponentes y produce una síntesis. No promedia —razona sobre las propuestas, detecta contradicciones, pondera por confiabilidad y produce una respuesta que ningún proponente podría haber generado solo.

**Capa 3 (opcional) — Capas adicionales:** el proceso puede iterarse. La salida del agregador se convierte en entrada de una nueva ronda de propuestas, refinando progresivamente la respuesta.

---

### La diversidad como propiedad de diseño

El beneficio principal de MoA no es tener más cómputo —es tener **perspectivas genuinamente distintas**. Para que eso ocurra, los agentes 
proponentes deben ser diseñados para disentir. Un MoA donde todos los agentes producen la misma respuesta es computacionalmente más caro que un modelo único y no ofrece ninguna ventaja.

Hay tres mecanismos para inducir diversidad, con distintos niveles de potencia y costo:

#### Mecanismo 1 — Temperatura

La temperatura $T$ controla la distribución de probabilidad sobre el vocabulario en cada paso de generación. Dado un vector de logits 
$z \in \mathbb{R}^{|V|}$, la probabilidad de cada token es:

$$P(w_i \mid \text{contexto}) = \frac{\exp(z_i / T)}{\sum_j \exp(z_j / T)}$$

El efecto de $T$ sobre la distribución es geométricamente claro:

| Temperatura | Efecto sobre la distribución | Comportamiento del agente |
|:---|:---|:---|
| $T \to 0$ | Colapsa en el token más probable (argmax) | Determinista, repetible, conservador |
| $T = 1.0$ | Distribución original del modelo | Balance entre exploración y explotación |
| $T > 1.0$ | Aplana la distribución (más uniforme) | Creativo, impredecible, propenso a errores |
| $T \to \infty$ | Distribución uniforme sobre vocabulario | Aleatorio puro, incoherente |


````python
import numpy as np
import matplotlib.pyplot as plt

# Ilustración del efecto de temperatura sobre logits de ejemplo
logits = np.array([3.5, 2.1, 1.8, 0.9, 0.3, -0.5, -1.2])
tokens = ["el", "un", "este", "su", "mi", "aquel", "dicho"]
temperaturas = [0.3, 0.7, 1.0, 1.5, 2.5]
colores = ["#1a3a6b", "#2d6a9f", "#4C72B0", "#8ab4d4", "#c8dff0"]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Panel izquierdo: distribuciones por temperatura
for T, color in zip(temperaturas, colores):
    logits_T = logits / T
    logits_T -= logits_T.max()
    probs = np.exp(logits_T) / np.exp(logits_T).sum()
    axes[0].plot(tokens, probs, marker='o', lw=1.8, color=color,
                 label=f"T={T}", markersize=5)

axes[0].set_title("Efecto de la temperatura\nsobre la distribución de tokens")
axes[0].set_ylabel("Probabilidad")
axes[0].legend(fontsize=9)
axes[0].grid(alpha=0.3)
axes[0].tick_params(axis='x', rotation=15)

# Panel derecho: entropía en función de T
ts = np.linspace(0.1, 3.0, 200)
entropias = []
for T in ts:
    logits_T = logits / T
    logits_T -= logits_T.max()
    probs = np.exp(logits_T) / np.exp(logits_T).sum()
    H = -np.sum(probs * np.log(probs + 1e-12))
    entropias.append(H)

axes[1].plot(ts, entropias, color="#4C72B0", lw=2)
axes[1].axvline(1.0, color="gray", ls="--", lw=1, label="T=1 (base)")
for T_mark, label in [(0.3, "conservador"), (1.5, "creativo")]:
    idx = np.argmin(np.abs(ts - T_mark))
    axes[1].axvline(T_mark, color="#DD8452", ls=":", lw=1.2, alpha=0.8)
    axes[1].annotate(f"T={T_mark}\n({label})",
                     xy=(T_mark, entropias[idx]),
                     xytext=(T_mark + 0.15, entropias[idx] - 0.15),
                     fontsize=8, color="#DD8452")

axes[1].set_title("Entropía de la distribución\nen función de T")
axes[1].set_xlabel("Temperatura T")
axes[1].set_ylabel("Entropía H (bits)")
axes[1].legend(fontsize=9)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
````

- El **agregador** usa $T$ baja (0.1–0.3): necesita ser preciso y  reproducible al sintetizar. Un agregador creativo introduce ruido 
  en la etapa crítica.
- Los **proponentes** usan $T$ media-alta (0.5–1.2): necesitan  explorar el espacio de respuestas posibles. Si todos usan $T=0.1$, 
  convergen a la misma respuesta y el MoA degenera en un modelo único más caro.

#### Mecanismo 2 — Top-p (nucleus sampling)

La temperatura controla *cuánta* probabilidad se redistribuye. **Top-p** controla *sobre cuántos tokens* se redistribuye. Dado un 
umbral $p \in (0, 1]$, solo se consideran los tokens cuya probabilidad acumulada llega a $p$:

$$V_p = \arg\min_{V' \subseteq V} \sum_{w \in V'} P(w) \geq p$$

$$P'(w_i) = \begin{cases} P(w_i) / Z & \text{si } w_i \in V_p \\ 0 & \text{en otro caso} \end{cases}$$

donde $Z = \sum_{w \in V_p} P(w)$ renormaliza.

| Top-p | Comportamiento | Cuándo usarlo |
|:---|:---|:---|
| `top_p=0.1` | Solo el token más probable | Tareas de clasificación o extracción |
| `top_p=0.7` | Vocabulario restringido pero diverso | Balance: análisis técnico |
| `top_p=0.95` | Vocabulario amplio | Generación creativa, exploración |
| `top_p=1.0` | Sin filtrado (nucleus completo) | Máxima diversidad |

La diferencia clave con temperatura: top-p **adapta el vocabulario activo** a la confianza del modelo. En posiciones donde el modelo 
es muy seguro (un token domina con 90% de probabilidad), top-p=0.9 solo permite ese token. En posiciones ambiguas (10 tokens con ~10% 
cada uno), top-p=0.9 permite todos. La temperatura, en cambio, aplica el mismo factor a todas las posiciones.

En la práctica, temperatura y top-p se usan juntos:

```python
# Configuración típica para agentes MoA
configs_agentes = {
    "conservador" : {"temperature": 0.2, "top_p": 0.85},  # preciso, poco vocabulario
    "balanceado"  : {"temperature": 0.7, "top_p": 0.92},  # uso general
    "exploratorio": {"temperature": 1.1, "top_p": 0.97},  # máxima diversidad
    "agregador"   : {"temperature": 0.1, "top_p": 0.80},  # determinista
}
```

#### Mecanismo 3 — Prompt de sistema

La temperatura y top-p controlan la *aleatoriedad* de la generación.  El **prompt de sistema** controla la *perspectiva* desde la que el  agente aborda el problema. Es el mecanismo más poderoso de los tres porque opera al nivel semántico, no estadístico.

Un prompt de sistema bien diseñado para MoA tiene tres componentes:

El formato estructurado en el prompt de sistema no es un detalle menor: 
**le da al agregador un esquema predecible** para extraer información de 
cada propuesta. Sin formato, el agregador necesita primero inferir la 
estructura de cada respuesta antes de poder sintetizarlas —trabajo 
cognitivo que consume parte de su ventana de contexto y puede introducir 
errores de interpretación.

#### Los tres mecanismos juntos

La combinación óptima de los tres mecanismos depende de la tarea:

| Tipo de tarea | Temperatura | Top-p | Prompt |
|:---|:---|:---|:---|
| Análisis técnico riguroso | 0.2–0.4 | 0.85 | Rol específico + proceso explícito |
| Generación creativa | 0.8–1.2 | 0.95 | Rol con sesgo deliberado |
| Síntesis / agregación | 0.1–0.2 | 0.80 | Árbitro estructurado |
| Exploración de hipótesis | 1.0–1.3 | 0.97 | Abogado del diablo |

> **Nota de diseño:** más diversidad no siempre mejora el sistema. Si los 
> agentes son demasiado distintos entre sí, el agregador recibe propuestas 
> incompatibles y su síntesis puede ser peor que cualquier propuesta individual. 
> La diversidad óptima es la mínima necesaria para cubrir los puntos ciegos 
> del modelo base —no la máxima posible. Este es el NFL aplicado al diseño 
> de MoA.

#### Una dimensión perpendicular: diversidad por modelo

Los tres mecanismos anteriores operan dentro de un mismo modelo base: 
controlan qué región de su distribución aprendida se explora en cada 
generación. La diversidad que producen es **intra-modelo** —distintas 
salidas del mismo conjunto de pesos.

Hay una forma de diversidad cualitativamente distinta: usar **modelos 
con arquitecturas y corpus de entrenamiento diferentes**. Dos modelos 
con la misma temperatura y el mismo prompt pueden llegar a respuestas 
distintas no por aleatoriedad, sino porque sus representaciones internas 
del problema son estructuralmente diferentes. Es diversidad **inter-modelo**, 
y opera en una dimensión perpendicular a las anteriores.

El espacio de diseño completo de un MoA tiene por tanto dos ejes:

| | Intra-modelo (mismo base) | Inter-modelo (distintos modelos) |
|:---|:---|:---|
| **Fuente de diversidad** | Aleatoriedad de muestreo + sesgo de prompt | Sesgos de arquitectura y entrenamiento |
| **Control** | Temperatura, top-p, prompt de sistema | Elección de modelos |
| **Costo** | Bajo (una sola API) | Alto (múltiples APIs o modelos locales) |
| **Tipo de diferencia** | Misma representación, distintas salidas | Representaciones distintas del problema |
| **Cuándo aporta más** | Tareas con una solución correcta pero múltiples caminos | Tareas donde el sesgo de entrenamiento importa |

**¿Cuándo vale la pena el costo?**

La diversidad inter-modelo aporta valor cuando los modelos tienen 
**puntos ciegos sistemáticamente distintos** —no solo salidas aleatorias 
distintas. Esto ocurre en tres situaciones concretas:

1. **Dominios con sesgo de corpus:** un modelo entrenado con más texto 
   científico y otro con más texto periodístico llegan a conclusiones 
   distintas sobre la misma evidencia. El agregador captura ambas 
   perspectivas.

2. **Tareas con múltiples paradigmas válidos:** un modelo que razona 
   inductivamente y otro que razona deductivamente pueden cubrir el 
   espacio de soluciones más completamente que cualquiera solo.

3. **Detección de alucinaciones:** cuando dos modelos independientes 
   coinciden en un hecho, la probabilidad de que ambos alucinen 
   coordinadamente es mucho menor que la de un modelo solo. El 
   desacuerdo entre modelos es una señal de incertidumbre genuina.

```python
# Configuración de un MoA con diversidad en ambas dimensiones
agentes_moa = [
    # Eje inter-modelo: tres bases distintas
    # Eje intra-modelo: temperatura y prompt diferenciados por rol

    {
        "nombre"     : "Analítico",

    },
    {

        "modelo"     : "claude-sonnet-4-20250514",   # mismo modelo,
        "temperature": 1.1,                           # distinta temperatura
        "top_p"      : 0.97,
        "sistema"    : prompts_sistema["optimista"]
    },
    {
        "nombre"     : "Alternativo",
        "modelo"     : "gpt-4o",                     # modelo distinto
        "temperature": 0.7,
        "top_p"      : 0.92,

                                                      # → diferencia viene del modelo
    },
]

agregador = {
    "nombre"     : "Agregador",
    "modelo"     : "claude-sonnet-4-20250514",

}
```

Nótese que el tercer agente usa el mismo prompt que el primero pero un 
modelo distinto. Eso aísla la contribución de la diversidad inter-modelo: 
si sus respuestas difieren, la diferencia viene de los sesgos de 
entrenamiento, no del prompt ni de la temperatura.

> **Nota de diseño:** en la práctica, la mayoría de sistemas MoA 
> productivos usan un solo modelo base con distintos prompts y 
> temperaturas. La diversidad inter-modelo multiplica el costo de 
> inferencia y añade complejidad operacional (múltiples APIs, distintos 
> límites de contexto, distintos formatos de respuesta). Vale la pena 
> cuando la tarea requiere cobertura de sesgos sistemáticos distintos 
> —no como default. El NFL, de nuevo: no hay una estrategia de diversidad 
> que gane en todos los contextos.

### Por qué la diversidad es una propiedad de diseño

El beneficio principal de MoA no es tener más capacidad computacional —es tener **perspectivas genuinamente distintas**. Para que eso ocurra, los agentes proponentes deben ser diseñados para disentir.

Hay tres mecanismos para inducir diversidad:

**Diversidad por temperatura:** el mismo modelo con distintos valores de temperatura de muestreo produce respuestas más o menos conservadoras.

```python
# Mismo modelo, perspectivas distintas
agente_conservador = {"model": "claude-sonnet-4-20250514", "temperature": 0.2}
agente_creativo    = {"model": "claude-sonnet-4-20250514", "temperature": 0.9}
```

**Diversidad por prompt de sistema:** cada agente recibe instrucciones que 
definen su rol y sesgo deliberado.

```python
prompts_sistema = {
    "critico":   "Eres un revisor escéptico. Tu función es identificar 
                  fallas, suposiciones no justificadas y casos borde 
                  que la respuesta no considera.",

    "sintetico": "Eres un sintetizador. Tu función es encontrar el núcleo 
                  de verdad en cualquier argumento y expresarlo con la 
                  máxima claridad y economía.",

    "abogado":   "Eres un abogado del diablo. Tu función es construir 
                  el mejor argumento posible para la posición contraria 
                  a la que se propone, sin importar tu opinión personal."
}
```

**Diversidad por modelo:** agentes construidos sobre arquitecturas distintas 
tienen sesgos de entrenamiento distintos —lo que uno tiende a sobre-estimar, 
otro tiende a corregir.

```python
agentes_heterogeneos = [
    {"model": "claude-sonnet-4-20250514"},   # fuerte en razonamiento
    {"model": "gpt-4o"},                      # fuerte en síntesis
    {"model": "gemini-1.5-pro"},              # fuerte en contexto largo
]
```

La tercera forma es la más poderosa pero también la más cara. En práctica, 
la combinación de prompt de sistema + temperatura produce diversidad suficiente 
con un solo modelo de base.

### 3.4 Implementación MoA con la API de Claude

Implementemos un MoA mínimo pero funcional: tres agentes proponentes con 
perspectivas distintas y un agregador que razona sobre sus propuestas.

```python
import anthropic
import concurrent.futures
import time
from typing import Optional

cliente = anthropic.Anthropic()

# ─────────────────────────────────────────────────────────────
# CONFIGURACIÓN DE AGENTES
# ─────────────────────────────────────────────────────────────
AGENTES_PROPONENTES = [
    {
        "nombre"     : "Analítico",
        "temperatura": 0.3,
        "sistema"    : """Eres un analista riguroso. Cuando respondas:
1. Descompón el problema en componentes verificables.
2. Cita solo lo que puedas justificar con evidencia o razonamiento.
3. Señala explícitamente lo que NO sabes o no puedes confirmar.
4. Sé conciso: máximo 150 palabras."""
    },
    {
        "nombre"     : "Crítico",
        "temperatura": 0.4,
        "sistema"    : """Eres un revisor escéptico. Cuando respondas:
1. Identifica las suposiciones más débiles del problema planteado.
2. Señala casos borde o excepciones que complican la respuesta simple.
3. Propón la respuesta más honesta aunque sea incómoda.
4. Sé conciso: máximo 150 palabras."""
    },
    {
        "nombre"     : "Sintetizador",
        "temperatura": 0.6,
        "sistema"    : """Eres un pensador de síntesis. Cuando respondas:
1. Busca el patrón de fondo que unifica los aspectos del problema.
2. Usa analogías o ejemplos concretos para anclar la respuesta.
3. Prioriza claridad sobre exhaustividad.
4. Sé conciso: máximo 150 palabras."""
    }
]

SISTEMA_AGREGADOR = """Eres un árbitro epistémico. Recibirás propuestas de 
tres agentes especialistas sobre la misma pregunta. Tu función es:

1. DETECTAR consenso: ¿en qué coinciden los tres?
2. DETECTAR disenso: ¿dónde difieren y por qué?
3. PONDERAR: ¿qué propuesta es más sólida y por qué?
4. SINTETIZAR: produce una respuesta final que incorpore lo mejor 
   de cada perspectiva y sea más completa que cualquiera individual.

Estructura tu respuesta así:
- **Consenso:** [lo que todos coinciden]
- **Tensiones:** [donde difieren y tu evaluación]
- **Síntesis:** [tu respuesta integrada]

Máximo 300 palabras."""

# ─────────────────────────────────────────────────────────────
# LLAMADA A UN AGENTE
# ─────────────────────────────────────────────────────────────
def consultar_agente(agente: dict, pregunta: str) -> dict:
    """Consulta a un agente proponente y registra timestamps."""
    t_inicio = time.time()
    print(f"  🔄 [{agente['nombre']}] iniciando...")

    respuesta = cliente.messages.create(
        model      = "claude-sonnet-4-20250514",
        max_tokens = 300,
        temperature= agente["temperatura"],
        system     = agente["sistema"],
        messages   = [{"role": "user", "content": pregunta}]
    )

    t_fin    = time.time()
    duracion = t_fin - t_inicio
    texto    = respuesta.content[0].text

    print(f"  ✅ [{agente['nombre']}] completado en {duracion:.2f}s")
    return {
        "nombre"  : agente["nombre"],
        "respuesta": texto,
        "duracion" : duracion,
        "tokens"   : respuesta.usage.output_tokens
    }

# ─────────────────────────────────────────────────────────────
# AGREGADOR
# ─────────────────────────────────────────────────────────────
def agregar_propuestas(pregunta: str, propuestas: list[dict]) -> dict:
    """El agregador razona sobre las propuestas de los proponentes."""
    t_inicio = time.time()
    print(f"\n  🎯 [Agregador] sintetizando {len(propuestas)} propuestas...")

    # Construir el contexto con todas las propuestas
    contexto = f"**Pregunta original:** {pregunta}\n\n"
    for p in propuestas:
        contexto += f"---\n**Propuesta del agente {p['nombre']}** "
        contexto += f"({p['tokens']} tokens, {p['duracion']:.1f}s):\n"
        contexto += f"{p['respuesta']}\n\n"

    respuesta = cliente.messages.create(
        model      = "claude-sonnet-4-20250514",
        max_tokens = 600,
        temperature= 0.2,     # agregador conservador: minimiza alucinaciones
        system     = SISTEMA_AGREGADOR,
        messages   = [{"role": "user", "content": contexto}]
    )

    t_fin    = time.time()
    duracion = t_fin - t_inicio
    texto    = respuesta.content[0].text

    print(f"  ✅ [Agregador] completado en {duracion:.2f}s")
    return {
        "respuesta": texto,
        "duracion" : duracion,
        "tokens"   : respuesta.usage.output_tokens
    }

# ─────────────────────────────────────────────────────────────
# SISTEMA MoA COMPLETO
# ─────────────────────────────────────────────────────────────
def moa(pregunta: str, paralelo: bool = True) -> dict:
    """
    Ejecuta el pipeline MoA completo.
    paralelo=True: proponentes en threads simultáneos (más rápido)
    paralelo=False: secuencial (más fácil de depurar)
    """
    print("═" * 65)
    print(f"  SISTEMA MoA — {len(AGENTES_PROPONENTES)} proponentes + 1 agregador")
    print(f"  Pregunta: {pregunta[:60]}...")
    print("═" * 65 + "\n")

    t0 = time.time()

    # ── Fase 1: Proponentes ───────────────────────────────────
    print("  FASE 1: Consulta a proponentes\n")
    if paralelo:
        # Los proponentes corren en paralelo — como el Blackboard
        with concurrent.futures.ThreadPoolExecutor(
                max_workers=len(AGENTES_PROPONENTES)) as executor:
            futuros = {
                executor.submit(consultar_agente, agente, pregunta): agente
                for agente in AGENTES_PROPONENTES
            }
            propuestas = []
            for futuro in concurrent.futures.as_completed(futuros):
                propuestas.append(futuro.result())
    else:
        propuestas = [consultar_agente(a, pregunta)
                      for a in AGENTES_PROPONENTES]

    # ── Fase 2: Agregación ────────────────────────────────────
    print("\n  FASE 2: Agregación\n")
    dictamen = agregar_propuestas(pregunta, propuestas)

    t_total = time.time() - t0

    # ── Tabla resumen ─────────────────────────────────────────
    print("\n" + "═" * 65)
    print("  PIZARRA FINAL")
    print("═" * 65)
    print(f"  {'Agente':<18} {'Duración':>10} {'Tokens':>8}")
    print(f"  {'─' * 40}")
    for p in sorted(propuestas, key=lambda x: x["duracion"]):
        print(f"  {p['nombre']:<18} {p['duracion']:>9.2f}s "
              f"{p['tokens']:>8}")
    print(f"  {'Agregador':<18} {dictamen['duracion']:>9.2f}s "
          f"{dictamen['tokens']:>8}")
    print(f"  {'─' * 40}")
    print(f"  {'TOTAL':<18} {t_total:>9.2f}s")
    print("═" * 65)

    return {
        "pregunta"  : pregunta,
        "propuestas": propuestas,
        "dictamen"  : dictamen,
        "t_total"   : t_total
    }
```

### 3.5 MoE vs MoA — cuándo usar cada uno

Ahora que tenemos ambos conceptos formalizados, la pregunta práctica es cuál 
elegir. La respuesta depende de tres ejes:

| Criterio | MoE | MoA |
|:---|:---|:---|
| **Nivel de operación** | Parámetros / capas | Modelos completos |
| **Entrenamiento** | End-to-end con gradientes | No requerido (prompting) |
| **Latencia** | Baja (un solo forward pass) | Alta (llamadas seriales o paralelas) |
| **Costo por inferencia** | Bajo (solo k expertos activos) | Alto (N llamadas a API) |
| **Diversidad de perspectivas** | Limitada por arquitectura común | Alta (modelos heterogéneos) |
| **Adaptabilidad sin reentrenamiento** | No | Sí (cambiar prompts de sistema) |
| **Interpretabilidad** | Baja (¿qué aprendió cada experto?) | Alta (podemos leer cada propuesta) |
| **Cuándo usarlo** | Producción a escala, latencia crítica | Tareas complejas, calidad prioritaria |

La distinción más importante para el practicante:

> **MoE** es la respuesta cuando el cuello de botella es el **cómputo**.  
> **MoA** es la respuesta cuando el cuello de botella es la **calidad de 
> razonamiento**.

En sistemas de producción modernos, ambos coexisten: los agentes de un MoA 
pueden ser ellos mismos modelos con arquitectura MoE internamente. GPT-4 
(posiblemente MoE internamente) orquestado como agente en un sistema MoA 
es el estado del arte en 2025.
